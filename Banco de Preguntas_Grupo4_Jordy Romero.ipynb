```python
# Primero, asegur√©monos de que tenemos las bibliotecas necesarias
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, Not, Or, And, Implies, Equivalent
from sympy.logic.boolalg import Or, And, Not

# Crear una funci√≥n para visualizar los conjuntos y los intervalos
def plot_set_operations():
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))

    # Mostrar la intersecci√≥n de conjuntos
    A = {1, 2, 3}
    B = {2, 3, 4}
    intersection = A & B
    ax[0].set_title('Intersecci√≥n A ‚à© B')
    ax[0].barh(['A ‚à© B'], [len(intersection)], color='skyblue')
    ax[0].set_xlim(0, 4)
    ax[0].set_xlabel('N√∫mero de elementos')
    ax[0].set_yticks([])

    # Mostrar el complemento
    U = {1, 2, 3, 4, 5}
    A = {1, 2, 3}
    complement = U - A
    ax[1].set_title('Complemento de A en U')
    ax[1].barh(['U - A'], [len(complement)], color='salmon')
    ax[1].set_xlim(0, 4)
    ax[1].set_xlabel('N√∫mero de elementos')
    ax[1].set_yticks([])

    plt.show()

plot_set_operations()

# Pregunta 1: Proposici√≥n l√≥gica
print("Pregunta 1: ¬øCu√°l de las siguientes es una proposici√≥n l√≥gica?")
print("a) ¬øCu√°l es tu nombre?")
print("b) ¬°Corre m√°s r√°pido!")
print("c) La Tierra es redonda.")
print("d) ¬°Qu√© d√≠a tan hermoso!")
print("e) ¬øPodr√≠as ayudarme?")
print("Respuesta Correcta: c) La Tierra es redonda.")
print("Explicaci√≥n: Las proposiciones son declaraciones que pueden ser verdaderas o falsas. 'La Tierra es redonda' es una declaraci√≥n que puede ser evaluada como verdadera o falsa.\n")

# Pregunta 2: Equivalencia de implicaci√≥n
p, q = symbols('p q')
implication = Implies(p, q)
equiv = Equivalent(implication, Or(Not(p), q))
print("Pregunta 2: ¬øQu√© es equivalente a p‚Üíq?")
print("a) p‚à®q")
print("b) ¬¨p‚à®q")
print("c) ¬¨p‚àßq")
print("d) p‚àß¬¨q")
print("e) p‚Üîq")
print(f"Respuesta Correcta: b) ¬¨p‚à®q")
print(f"Explicaci√≥n: La implicaci√≥n p‚Üíq es equivalente a ¬¨p‚à®q seg√∫n la definici√≥n l√≥gica.\n")

# Pregunta 3: Negaci√≥n de conjunci√≥n
neg_conj = Not(And(p, q))
neg_conj_de_morgan = Or(Not(p), Not(q))
print("Pregunta 3: ¬øCu√°l es la negaci√≥n de p‚àßq?")
print("a) ¬¨p‚àßq")
print("b) p‚à®q")
print("c) ¬¨p‚à®¬¨q")
print("d) ¬¨p‚à®q")
print("e) p‚àß¬¨q")
print(f"Respuesta Correcta: c) ¬¨p‚à®¬¨q")
print(f"Explicaci√≥n: Seg√∫n las leyes de De Morgan, la negaci√≥n de p‚àßq es ¬¨p‚à®¬¨q.\n")

# Pregunta 4: Complemento de la uni√≥n de conjuntos
A, B = symbols('A B')
complement_union = And(Not(A), Not(B))
print("Pregunta 4: ¬øCu√°l es el complemento de A‚à™B?")
print("a) ¬¨A‚à©¬¨B")
print("b) ¬¨A‚à™B")
print("c) A‚à©B")
print("d) ¬¨(A‚à©B)")
print("e) A‚à™¬¨B")
print(f"Respuesta Correcta: a) ¬¨A‚à©¬¨B")
print(f"Explicaci√≥n: Seg√∫n la ley de De Morgan, el complemento de A‚à™B es ¬¨A‚à©¬¨B.\n")

# Pregunta 5: Intersecci√≥n de conjuntos
A = {1, 2, 3}
B = {2, 3, 4}
intersection = A & B
print("Pregunta 5: ¬øQu√© elementos contiene A‚à©B si A={1,2,3} y B={2,3,4}?")
print("a) {1,2,3,4}")
print("b) {2,3}")
print("c) {1,4}")
print("d) {1,2}")
print("e) {3,4}")
print(f"Respuesta Correcta: b) {intersection}")
print(f"Explicaci√≥n: La intersecci√≥n A‚à©B contiene solo los elementos que est√°n en ambos conjuntos A y B.\n")

# Pregunta 6: Complemento en el universo
U = {1, 2, 3, 4, 5}
A = {1, 2, 3}
complement = U - A
print("Pregunta 6: Si el universo es U={1,2,3,4,5} y A={1,2,3}, ¬øcu√°l es el complemento de A?")
print("a) {1,2}")
print("b) {3,4}")
print("c) {4,5}")
print("d) {2,3,4}")
print("e) {1,2,3,4}")
print(f"Respuesta Correcta: c) {complement}")
print(f"Explicaci√≥n: El complemento de A en el universo U son los elementos de U que no est√°n en A.\n")

# Pregunta 7: Notaci√≥n de intervalos cerrados
print("Pregunta 7: ¬øCu√°l es la notaci√≥n correcta para un intervalo cerrado que incluye 2 y 5?")
print("a) (2,5)")
print("b) [2,5)")
print("c) (2,5]")
print("d) [2,5]")
print("e) (2,4)")
print("Respuesta Correcta: d) [2,5]")
print("Explicaci√≥n: Un intervalo cerrado incluye ambos extremos, lo que se denota por corchetes.\n")

# Pregunta 8: Significado del intervalo (3,7)
print("Pregunta 8: ¬øQu√© significa el intervalo (3,7)?")
print("a) Incluye 3 y 7.")
print("b) Incluye 3 pero no 7.")
print("c) Incluye 7 pero no 3.")
print("d) No incluye 3 ni 7.")
print("e) Incluye 4 y 6.")
print("Respuesta Correcta: d) No incluye 3 ni 7.")
print("Explicaci√≥n: Un intervalo abierto (3,7) no incluye los extremos 3 y 7.\n")

# Pregunta 9: Conjunto Z de enteros
print("Pregunta 9: ¬øCu√°l de los siguientes conjuntos representa Z?")
print("a) {1,2,3,‚Ä¶}")
print("b) {0,1,2,3,‚Ä¶}")
print("c) {‚Ä¶,‚àí2,‚àí1,0,1,2,‚Ä¶}")
print("d) {2,4,6,‚Ä¶}")
print("e) {3,6,9,‚Ä¶}")
print("Respuesta Correcta: c) {‚Ä¶,‚àí2,‚àí1,0,1,2,‚Ä¶}")
print("Explicaci√≥n: Z incluye todos los enteros positivos, negativos y el cero.\n")

# Pregunta 10: Elemento neutro aditivo
print("Pregunta 10: ¬øCu√°l es el elemento neutro aditivo en los enteros?")
print("a) 1")
print("b) -1")
print("c) 0")
print("d) 2")
print("e) -2")
print("Respuesta Correcta: c) 0")
print("Explicaci√≥n: El elemento neutro aditivo es aquel que, al sumarlo a cualquier entero, no cambia su valor. En este caso, es 0.\n")

# Pregunta 11: Propiedad de los enteros
print("Pregunta 11: ¬øCu√°l es la propiedad que garantiza que la suma de dos enteros siempre resulta en otro entero?")
print("a) Asociatividad")
print("b) Conmutatividad")
print("c) Cerradura")
print("d) Elemento neutro")
print("e) Inverso")
print("Respuesta Correcta: c) Cerradura")
print("Explicaci√≥n: La propiedad de cerradura garantiza que la suma de dos enteros siempre resulta en otro entero.\n")

# Pregunta 12: Uni√≥n de intervalos
print("Pregunta 12: ¬øCu√°l es la uni√≥n de los conjuntos [1,2] y [2,3]?")
print("a) (1,3)")
print("b) [1,2]")
print("c) [1,3]")
print("d) [2,3]")
print("e) (2,3]")
print("Respuesta Correcta: c) [1,3]")
print("Explicaci√≥n: La uni√≥n de [1,2] y [2,3] incluye todos los elementos en cualquiera de los dos intervalos, lo cual es [1,3].\n")
```
