# Ejercicio en Python ‚ùó
Vamos a calcular la uni√≥n de intervalos con extremos abiertos y cerrados.

```python
def unir_intervalos(intervalos):
    # Ordenar intervalos por el inicio
    intervalos_ordenados = sorted(intervalos)
    unificados = []
    inicio_actual, fin_actual = intervalos_ordenados[0]

    for inicio, fin in intervalos_ordenados[1:]:
        if inicio <= fin_actual:  # Hay solapamiento
            fin_actual = max(fin_actual, fin)
        else:
            unificados.append((inicio_actual, fin_actual))
            inicio_actual, fin_actual = inicio, fin

    unificados.append((inicio_actual, fin_actual))
    return unificados

# Definir intervalos
intervalos = [(1, 2), (2, 3), (5, 6), (4, 5)]

# Unir los intervalos
intervalos_unidos = unir_intervalos(intervalos)

# Mostrar los intervalos unidos
print("Intervalos unidos:")
for intervalo in intervalos_unidos:
    print(f"[{intervalo[0]}, {intervalo[1]}]")
```

o	Una funci√≥n f es inyectiva si diferentes elementos en E se mapean a diferentes elementos en F.

3.	Sobreyectividad:

Una funci√≥n f es sobreyectiva si cada elemento en F es imagen de alg√∫n elemento en E.
o	Formalmente: Para todo y‚ààFy \in Fy‚ààF, existe x‚ààEx \in Ex‚ààE tal que f(x)=yf(x) = yf(x)=y.
Ejemplo:
o	f(x)=x2f(x) = x^2f(x)=x2 mapeada de R\mathbb{R}R a [0,‚àû)[0, \infty)[0,‚àû) es sobreyectiva porque cada y‚â•0y \geq 0y‚â•0 tiene un x‚ààRx \in \mathbb{R}x‚ààR tal que f(x)=yf(x) = yf(x)=y.


```python
def f(x):
    return x**2

def es_sobreyectiva(func, dominio, codominio):
    imagen = {func(x) for x in dominio}
    return imagen >= codominio

dominio = range(-10, 11)
codominio = set(range(0, 101))
print(es_sobreyectiva(f, dominio, codominio))  # Deber√≠a ser True
```

4.	Biyectividad:

o	f(x)=x+1f(x) = x + 1f(x)=x+1 es biyectiva porque es inyectiva (no mapea dos valores distintos a la misma imagen) y sobreyectiva (cubre todo R\ R).

A continuaci√≥n, un ejercicio para calcular la distancia euclidiana entre dos puntos en un espacio de dos dimensiones:

```python
import math

def distancia_euclidiana(punto1, punto2):
    """
    Calcula la distancia euclidiana entre dos puntos en un espacio 2D.

    Args:
    punto1 (tuple): Coordenadas del primer punto (x1, y1).
    punto2 (tuple): Coordenadas del segundo punto (x2, y2).

    Returns:
    float: Distancia euclidiana entre los dos puntos.
    """
    x1, y1 = punto1
    x2, y2 = punto2
    distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distancia

# Ejemplo de uso
punto_a = (3, 4)
punto_b = (7, 1)
print(f"La distancia euclidiana entre {punto_a} y {punto_b} es {distancia_euclidiana(punto_a, punto_b):.2f}")
```

Este c√≥digo define una funci√≥n para calcular la distancia euclidiana entre dos puntos en un plano bidimensional, usa la f√≥rmula (x2‚àíx1)2+(y2‚àíy1)2\{(x_2 - x_1)^2 + (y_2 - y_1)^2}(x2‚àíx1)2+(y2‚àíy1)2, y proporciona un ejemplo de uso.
