##**FunciÃ³n Lineal**
```python
# DefiniciÃ³n de la funciÃ³n lineal
def funcion_lineal(x, m, b):
    """
    Calcula el valor de una funciÃ³n lineal f(x) = mx + b.

    Args:
    x (float): Valor de la variable independiente.
    m (float): Pendiente de la funciÃ³n.
    b (float): Ordenada al origen de la funciÃ³n.

    Returns:
    float: Valor de la funciÃ³n lineal en x.
    """
    return m * x + b

# Ejemplo de uso
x = 2   # Valor de x
m = 1.5 # Pendiente
b = 3   # Ordenada al origen

# Calcular f(x) usando la funciÃ³n lineal definida
resultado = funcion_lineal(x, m, b)
print(f"Para x = {x}, la funciÃ³n lineal con m = {m} y b = {b} da como resultado: {resultado}")
```
##**FunciÃ³n CuadrÃ¡tica**
```python
# DefiniciÃ³n de la funciÃ³n cuadrÃ¡tica
def funcion_cuadratica(x, a, b, c):
    """
    Calcula el valor de una funciÃ³n cuadrÃ¡tica f(x) = ax^2 + bx + c.

    Args:
    x (float): Valor de la variable independiente.
    a (float): Coeficiente cuadrÃ¡tico.
    b (float): Coeficiente lineal.
    c (float): TÃ©rmino constante.

    Returns:
    float: Valor de la funciÃ³n cuadrÃ¡tica en x.
    """
    return a * x**2 + b * x + c

# Ejemplo de uso
x = 2   # Valor de x
a = 1   # Coeficiente cuadrÃ¡tico
b = -3  # Coeficiente lineal
c = 2   # TÃ©rmino constante

# Calcular f(x) usando la funciÃ³n cuadrÃ¡tica definida
resultado = funcion_cuadratica(x, a, b, c)
print(f"Para x = {x}, la funciÃ³n cuadrÃ¡tica con a = {a}, b = {b}, y c = {c} da como resultado: {resultado}")
```
##**FunciÃ³n Racional**
```python
# DefiniciÃ³n de la funciÃ³n racional
def funcion_racional(x, P, Q):
    """
    Calcula el valor de una funciÃ³n racional f(x) = P(x) / Q(x).

    Args:
    x (float): Valor de la variable independiente.
    P (callable): FunciÃ³n polinÃ³mica en el numerador.
    Q (callable): FunciÃ³n polinÃ³mica en el denominador.

    Returns:
    float: Valor de la funciÃ³n racional en x.
    """
    try:
        return P(x) / Q(x)
    except ZeroDivisionError:
        return float('inf')  # Devuelve infinito si el denominador es cero

# DefiniciÃ³n de los polinomios P(x) y Q(x)
def P(x):
    return 2 * x**2 + 3 * x + 1  # Ejemplo de polinomio en el numerador

def Q(x):
    return x - 1  # Ejemplo de polinomio en el denominador

# Ejemplo de uso
x = 2  # Valor de x

# Calcular f(x) usando la funciÃ³n racional definida
resultado = funcion_racional(x, P, Q)
print(f"Para x = {x}, la funciÃ³n racional con P(x) = 2x^2 + 3x + 1 y Q(x) = x - 1 da como resultado: {resultado}")
```
##**FunciÃ³n Irracional**
```python
import math

# DefiniciÃ³n de la funciÃ³n irracional
def funcion_irracional(x):
    """
    Calcula el valor de una funciÃ³n irracional f(x) = sqrt(x^2 + 1).

    Args:
    x (float): Valor de la variable independiente.

    Returns:
    float: Valor de la funciÃ³n irracional en x.
    """
    return math.sqrt(x**2 + 1)

# Ejemplo de uso
x = 2  # Valor de x

# Calcular f(x) usando la funciÃ³n irracional definida
resultado = funcion_irracional(x)
print(f"Para x = {x}, la funciÃ³n irracional f(x) = sqrt(x^2 + 1) da como resultado: {resultado}")
```
##**FunciÃ³n Exponencial**
```python
import math

# DefiniciÃ³n de la funciÃ³n exponencial
def funcion_exponencial(x, a, b):
    """
    Calcula el valor de una funciÃ³n exponencial f(x) = a * e^(bx).

    Args:
    x (float): Valor de la variable independiente.
    a (float): Coeficiente que escala la funciÃ³n.
    b (float): Coeficiente del exponente.

    Returns:
    float: Valor de la funciÃ³n exponencial en x.
    """
    return a * math.exp(b * x)

# Ejemplo de uso
x = 2   # Valor de x
a = 1   # Coeficiente que escala la funciÃ³n
b = 0.5 # Coeficiente del exponente

# Calcular f(x) usando la funciÃ³n exponencial definida
resultado = funcion_exponencial(x, a, b)
print(f"Para x = {x}, la funciÃ³n exponencial con a = {a} y b = {b} da como resultado: {resultado}")
```
##**FunciÃ³n LogarÃ­tmica**
```python
import math

# DefiniciÃ³n de la funciÃ³n logarÃ­tmica
def funcion_logaritmica(x, a, b):
    """
    Calcula el valor de una funciÃ³n logarÃ­tmica f(x) = a * log_b(x).

    Args:
    x (float): Valor de la variable independiente.
    a (float): Coeficiente que escala la funciÃ³n.
    b (float): Base del logaritmo.

    Returns:
    float: Valor de la funciÃ³n logarÃ­tmica en x.
    """
    # Usamos el cambio de base para calcular el logaritmo en una base diferente de e
    return a * (math.log(x) / math.log(b))

# Ejemplo de uso
x = 2   # Valor de x
a = 1   # Coeficiente que escala la funciÃ³n
b = 10  # Base del logaritmo

# Calcular f(x) usando la funciÃ³n logarÃ­tmica definida
resultado = funcion_logaritmica(x, a, b)
print(f"Para x = {x}, la funciÃ³n logarÃ­tmica con a = {a} y b = {b} da como resultado: {resultado}")
```
##**FunciÃ³n TrigonomÃ©trica**
```python
import math

# DefiniciÃ³n de la funciÃ³n trigonomÃ©trica
def funcion_trigonometrica(x, a, b, c):
    """
    Calcula el valor de una funciÃ³n trigonomÃ©trica f(x) = a * sin(bx + c).

    Args:
    x (float): Valor de la variable independiente.
    a (float): Amplitud de la funciÃ³n.
    b (float): Frecuencia angular.
    c (float): Desplazamiento de fase.

    Returns:
    float: Valor de la funciÃ³n trigonomÃ©trica en x.
    """
    return a * math.sin(b * x + c)

# Ejemplo de uso
x = math.pi / 2  # Valor de x
a = 1            # Amplitud
b = 1            # Frecuencia angular
c = 0            # Desplazamiento de fase

# Calcular f(x) usando la funciÃ³n trigonomÃ©trica definida
resultado = funcion_trigonometrica(x, a, b, c)
print(f"Para x = {x}, la funciÃ³n trigonomÃ©trica con a = {a}, b = {b}, y c = {c} da como resultado: {resultado}")
```
