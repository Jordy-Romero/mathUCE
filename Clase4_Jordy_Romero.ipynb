###**Origen de las Funciones**
El concepto de función como un objeto matemático independiente, susceptible de ser estudiado por sí solo, no apareció hasta los inicios del cálculo en el siglo XVII. René Descartes, Isaac Newton y Gottfried Leibniz establecieron la idea de función como dependencia entre dos cantidades variables. Leibniz en particular acuñó los términos «función», «variable», «constante» y «parámetro». La notación f(x) fue utilizada por primera vez por A.C. Clairaut, y porLeonhard Euler en su obra Commentarii de San petersburgo en 1736.

![Texto alternativo](https://encrypted-tbn3.gstatic.com/licensed-image?q=tbn:ANd9GcSSaPS1whNliQyei3dU1WQ8l187xwK9IW7XBAeavNtloh0HEdro8maVdQbTSQliEjEY1G-sgr6IJeWLkRg)

Inicialmente, una función se identificaba a efectos prácticos con una expresión analítica que permitía calcular sus valores. Sin embargo, esta definición tenía algunas limitaciones: expresiones distintas pueden arrojar los mismos valores, y no todas las «dependencias» entre dos cantidades pueden expresarse de esta manera. En 1837 Dirichlet propuso la definición moderna de función numérica como una correspondencia cualquiera entre dos conjuntos de números, que asocia a cada número en el primer conjunto un único número del segundo.

###**Concepto de Funciones**
En matemáticas, una función de un conjunto X a un conjunto Y asigna a cada elemento de X exactamente un elemento de Y. El conjunto X se denomina dominio de la función y el conjunto Y se denomina codominio de la función. Las funciones eran originalmente la idealización de cómo una cantidad variable depende de otra cantidad.

Es una regla de correspondencia entre dos conjuntos de tal manera que a cada elemento del primer conjunto le corresponde uno y sólo un elemento del segundo conjunto. Al primer conjunto (el conjunto D) se le da el nombre de dominio.

###**Conjunto de partida o Preimágen**
El conjunto de partida en una función es lo que comúnmente se llama dominio de una función, y es el conjunto que se utiliza como entrada en la misma; el conjunto de llegada se le suele llamar codominio, imagen o recorrido y es el conjunto que se obtiene como salida de la función.

Al conjunto inicial o de partida donde están los valores de la variable independiente se le llama dominio que se abrevia **Dom (f)** y al conjunto final o de llegada donde están las imágenes se llama codominio que se abrevia **Codom (f)**.

###**Conjunto de llegada o Imágen**
En matemáticas, un codominio o conjunto de destino de una función es un conjunto en el que está obligada a caer toda la salida de la función. Es el conjunto Y en la notación f: X → Y. El término rango a veces se usa de manera ambigua para referirse al codominio o a la imagen de una función.
$$f(x) A → B$$
$$Dom f(x) = A$$
$$Ran f(x) = B$$

#**Tipos de Funciones**
##**Función Lineal**
Una función lineal es una función polinómica de primer grado. Es decir, tiene la siguiente forma:
$$f(x)=mx+b$$
Donde:

**m: es la pendiente de la función**

**n: es la ordenada (en el origen) de la función**

La gráfica de una función lineal es siempre una recta.

![Texto alternativo](https://www.problemasyecuaciones.com/funciones/lineales/T2.png)

##**Pendiente de la recta**
La pendiente es el coeficiente de la variable, es decir, m.

Geométricamente, cuanto mayor es la pendiente, más inclinada es la recta. Es decir, más rápido crece la función.

**Pendiente positiva:** Al aumentar los valores de la variable independiente “X” aumentar los valores de la variable dependiente “y”, por lo tanto la función es creciente.

$$m>0$$

**Pendiente negativa:** Al aumentar los valores de la variable dependiente “X”, los valores de la variable dependiente “Y” disminuyen, por lo tanto la función es decreciente.

$$m<0$$

**Pendiente nula:** Al aumentar los valores de la variable dependiente “x” los valores de la variable dependiente “y” no varían. Por lo tanto la función es constante.

$$m=0$$

$$m = \frac{y_2 - y_1}{x_2 - x_1}$$

##**Función Cuadrática**
La gráfica de una función cuadrática es una parábola , un tipo de curva de 2 dimensiones.

##**Raíces de la función cuadrática**
Las raíces o ceros de una función son los valores de x donde esta se anula, es decir, para los cuales f(x)=0. Geométricamente, las raíces son los puntos donde la gráfica de la función corta al eje x.

Sustituyendo f(x) por 0 en una función cuadrática se obtiene la ecuación ax² + bx + c = 0. Esta puede ser resuelta mediante la fórmula resolvente para ecuaciones cuadráticas, también llamada fórmula general o de Bhaskara:

La fórmula general para resolver una ecuación cuadrática es:

$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

##**Disriminante**
Habiendo una raíz cuadrada en esta fórmula, sabemos que solo existe si su radicando es no negativo, siendo en caso contrario un número imaginario. Este radicando recibe el nombre de discriminante, y se lo denota con la letra griega delta mayúscula:
$$Δ = b²- 4ac$$

La importancia del discriminante radica en que nos dirá si la función cuadrática tiene o no raíces reales. Pueden darse tres casos:

1) Si el discriminante es positivo, la función tiene dos raíces reales diferentes.
$$X_1 \neq X_2$$
2) Si el discriminante es cero, la función tiene una raíz real.
$$X_1=X_2$$
3) Si el discriminante es negativo, la función no tiene raíces reales.
$$X=i$$

##**Función Trigonométrica**
###**Función Seno**
 Esta función puede ser evaluada para cualquier número real, la función seno está definida para todos los números reales. El periodo de esta función puede ser observado claramente en su gráfica, ya que es la distancia entre puntos «equivalentes».
$$y=sen(x)$$
###**Periodo de la función Seno**
  Dado que la gráfica de $y=sin(x)$ se ve como un único patrón que se repite una y otra vez, podemos pensar en el periodo como la distancia en el eje x antes de que la gráfica empiece a repetirse.
  Observando a la gráfica, vemos que la gráfica se repite después de #2π#. Esto significa que la función es periódica con un periodo de $2π$. En el círculo unitario, $2π$ equivale a una vuelta completa alrededor del círculo.

  Cualquier cantidad mayor que $2π$ significa que estamos repitiendo la vuelta. Esta es la razón por la que el valor de la función es el mismo cada $2π$.

##**Encontrar el periodo de una función seno**
Para encontrar el periodo de una función seno, debemos considerar al coeficiente de la x que está dentro de la función. Podemos usar la B para representar a este coeficiente. Entonces, si es que tenemos una ecuación en la forma $y=sin(Bx)$, tenemos la siguiente fórmula:
$$Periodo = \frac{|B|}{2\pi}$$

En el denominador, tenemos |B|. Esto significa que tomamos el valor absoluto de B. Entonces, si es que B es un número negativo, simplemente tomamos la versión positiva del número.

###**Función Coseno**
 La función coseno es una función trigonométrica que es periódica. Una función periódica es una función que se repite a sí misma una y otra vez en ambas direcciones. El periodo de la función coseno es 2π, por lo tanto, el valor de la función es equivalente a cada 2π unidades.
###**Periodo de la función coseno**
El periodo de la función coseno en su forma básica, $y=cos(x)$, es $2π$. Este periodo puede ser modificado al multiplicar a la variable x por una constante.

Podemos reducir el periodo de la función al multiplicar a x por un número mayor que 1. Esto producirá que la función sea «acelerada» y el periodo se haga más pequeño.

Esto significa que la función ocurrirá más rápidamente y tomará menos para que empiece a repetirse. Por ejemplo, en la función $y=cos(2x)$, el periodo es $π$, lo cual es la mitad del periodo de la función original.
###**Determinar el periodo de una función coseno**
Podemos determinar el periodo de una función coseno al usar al coeficiente de la variable x. Este coeficiente es usualmente representado por la letra B. Entonces, la forma estándar de la función coseno es $y=sin(Bx)$. Usando esta forma, podemos obtener la siguiente fórmula:
$$Periodo = \frac{2π}{|B|}$$
Esto significa que para obtener el periodo, simplemente tenemos que dividir a $2π$ por $|B|$, en donde, |B| es el valor absoluto de B. Para encontrar el valor absoluto, solo tenemos que tomar la versión positiva del número. Por ejemplo, si tenemos $-2$, su valor absoluto es $2$.

Podemos usar esta fórmula incluso cuando tenemos otras variaciones de la función coseno. Por ejemplo, si tenemos la función

$y=2cos(2x+5)$, solo tomamos al coeficiente de la variable x:
$$Periodo = \frac{2π}{|B|}$$
$$Periodo = \frac{2π}{2}$$
$$Periodo =2$$

###**Función Tangente**
La gráfica de la tangente se ve diferente en comparación con las gráficas del seno y del coseno. A diferencia del dominio del seno y del coseno, el dominio de la tangente no es igual a todos los números reales. La tangente es equivalente al seno sobre el coseno. Por lo tanto, cuando los valores del coseno son iguales a $0$, la tangente tiende a infinito. Esto significa que la tangente tiene asíntotas cuando los valores del coseno son $0$. Por otra parte, el rango de la tangente sí es igual a todos los números reales.
$$tan(x) = \frac{\cos(x)}{\sin(x)}$$

##**Códigos Python**
##**Función Lineal**
```python
# Definición de la función lineal
def funcion_lineal(x, m, b):
    """
    Calcula el valor de una función lineal f(x) = mx + b.

    Args:
    x (float): Valor de la variable independiente.
    m (float): Pendiente de la función.
    b (float): Ordenada al origen de la función.

    Returns:
    float: Valor de la función lineal en x.
    """
    return m * x + b

# Ejemplo de uso
x = 2   # Valor de x
m = 1.5 # Pendiente
b = 3   # Ordenada al origen

# Calcular f(x) usando la función lineal definida
resultado = funcion_lineal(x, m, b)
print(f"Para x = {x}, la función lineal con m = {m} y b = {b} da como resultado: {resultado}")
```
##**Función Cuadrática**
```python
# Definición de la función cuadrática
def funcion_cuadratica(x, a, b, c):
    """
    Calcula el valor de una función cuadrática f(x) = ax^2 + bx + c.

    Args:
    x (float): Valor de la variable independiente.
    a (float): Coeficiente cuadrático.
    b (float): Coeficiente lineal.
    c (float): Término constante.

    Returns:
    float: Valor de la función cuadrática en x.
    """
    return a * x**2 + b * x + c

# Ejemplo de uso
x = 2   # Valor de x
a = 1   # Coeficiente cuadrático
b = -3  # Coeficiente lineal
c = 2   # Término constante

# Calcular f(x) usando la función cuadrática definida
resultado = funcion_cuadratica(x, a, b, c)
print(f"Para x = {x}, la función cuadrática con a = {a}, b = {b}, y c = {c} da como resultado: {resultado}")
```
##**Función Racional**
```python
# Definición de la función racional
def funcion_racional(x, P, Q):
    """
    Calcula el valor de una función racional f(x) = P(x) / Q(x).

    Args:
    x (float): Valor de la variable independiente.
    P (callable): Función polinómica en el numerador.
    Q (callable): Función polinómica en el denominador.

    Returns:
    float: Valor de la función racional en x.
    """
    try:
        return P(x) / Q(x)
    except ZeroDivisionError:
        return float('inf')  # Devuelve infinito si el denominador es cero

# Definición de los polinomios P(x) y Q(x)
def P(x):
    return 2 * x**2 + 3 * x + 1  # Ejemplo de polinomio en el numerador

def Q(x):
    return x - 1  # Ejemplo de polinomio en el denominador

# Ejemplo de uso
x = 2  # Valor de x

# Calcular f(x) usando la función racional definida
resultado = funcion_racional(x, P, Q)
print(f"Para x = {x}, la función racional con P(x) = 2x^2 + 3x + 1 y Q(x) = x - 1 da como resultado: {resultado}")
```
##**Función Irracional**
```python
import math

# Definición de la función irracional
def funcion_irracional(x):
    """
    Calcula el valor de una función irracional f(x) = sqrt(x^2 + 1).

    Args:
    x (float): Valor de la variable independiente.

    Returns:
    float: Valor de la función irracional en x.
    """
    return math.sqrt(x**2 + 1)

# Ejemplo de uso
x = 2  # Valor de x

# Calcular f(x) usando la función irracional definida
resultado = funcion_irracional(x)
print(f"Para x = {x}, la función irracional f(x) = sqrt(x^2 + 1) da como resultado: {resultado}")
```
##**Función Exponencial**
```python
import math

# Definición de la función exponencial
def funcion_exponencial(x, a, b):
    """
    Calcula el valor de una función exponencial f(x) = a * e^(bx).

    Args:
    x (float): Valor de la variable independiente.
    a (float): Coeficiente que escala la función.
    b (float): Coeficiente del exponente.

    Returns:
    float: Valor de la función exponencial en x.
    """
    return a * math.exp(b * x)

# Ejemplo de uso
x = 2   # Valor de x
a = 1   # Coeficiente que escala la función
b = 0.5 # Coeficiente del exponente

# Calcular f(x) usando la función exponencial definida
resultado = funcion_exponencial(x, a, b)
print(f"Para x = {x}, la función exponencial con a = {a} y b = {b} da como resultado: {resultado}")
```
##**Función Logarítmica**
```python
import math

# Definición de la función logarítmica
def funcion_logaritmica(x, a, b):
    """
    Calcula el valor de una función logarítmica f(x) = a * log_b(x).

    Args:
    x (float): Valor de la variable independiente.
    a (float): Coeficiente que escala la función.
    b (float): Base del logaritmo.

    Returns:
    float: Valor de la función logarítmica en x.
    """
    # Usamos el cambio de base para calcular el logaritmo en una base diferente de e
    return a * (math.log(x) / math.log(b))

# Ejemplo de uso
x = 2   # Valor de x
a = 1   # Coeficiente que escala la función
b = 10  # Base del logaritmo

# Calcular f(x) usando la función logarítmica definida
resultado = funcion_logaritmica(x, a, b)
print(f"Para x = {x}, la función logarítmica con a = {a} y b = {b} da como resultado: {resultado}")
```
##**Función Trigonométrica**
```python
import math

# Definición de la función trigonométrica
def funcion_trigonometrica(x, a, b, c):
    """
    Calcula el valor de una función trigonométrica f(x) = a * sin(bx + c).

    Args:
    x (float): Valor de la variable independiente.
    a (float): Amplitud de la función.
    b (float): Frecuencia angular.
    c (float): Desplazamiento de fase.

    Returns:
    float: Valor de la función trigonométrica en x.
    """
    return a * math.sin(b * x + c)

# Ejemplo de uso
x = math.pi / 2  # Valor de x
a = 1            # Amplitud
b = 1            # Frecuencia angular
c = 0            # Desplazamiento de fase

# Calcular f(x) usando la función trigonométrica definida
resultado = funcion_trigonometrica(x, a, b, c)
print(f"Para x = {x}, la función trigonométrica con a = {a}, b = {b}, y c = {c} da como resultado: {resultado}")
```
