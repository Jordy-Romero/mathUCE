### 🔍 **¿Qué es el Límite de una sucesión?** 🔍
![Texto alternativo](https://media.tenor.com/MVK4B9LYZGAAAAAi/ijustboughtlandinthemetaverse.gif)

Formalmente, una sucesión es una función que asigna números naturales a un conjunto de números, comúnmente denotados como $\{a_n\}_{n=1}^{\infty}$. El límite de una sucesión, si existe, es el valor al que se aproximan los términos de la sucesión a medida que el índice $n$ tiende a infinito. Se representa matemáticamente como:

$$ \lim_{n \to \infty} a_n = L $$

donde $L$ es el límite. Intuitivamente, esto significa que para cualquier valor arbitrariamente pequeño $\epsilon > 0$, existe un índice natural $N$ tal que para todo $n > N$, se cumple la desigualdad:

$$ |a_n - L| < \epsilon $$


```python
import matplotlib.pyplot as plt
import numpy as np

# Definición de la sucesión a_n (en este caso, una sucesión que converge a 0.5)
def sucesion(n):
    return 1 / (n + 1)

# Definición del límite L
L = 0.5

# Valores de n para la gráfica
n_values = np.arange(1, 21)  # Valores de n desde 1 hasta 20

# Términos de la sucesión a_n
sucesion_values = [sucesion(n) for n in n_values]

# Gráfica de la sucesión y su límite
plt.figure(figsize=(10, 6))
plt.plot(n_values, sucesion_values, marker='o', linestyle='-', color='b', label='$a_n$')
plt.hlines(L, n_values[0], n_values[-1], colors='r', linestyles='dashed', label='$L$')
plt.xlabel('n')
plt.ylabel('$a_n$')
plt.title('Aproximación de una sucesión a su límite')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 🌟 **Propiedades de los límites de sucesiones** 🌟

Las propiedades básicas que gobiernan los límites de sucesiones son las siguientes:

* **Propiedad aditiva:** Si $\lim_{n \to \infty} a_n = L$ y $\lim_{n \to \infty} b_n = M$, entonces se cumple que $\lim_{n \to \infty} (a_n + b_n) = L + M$.
* **Propiedad multiplicativa:** Si $\lim_{n \to \infty} a_n = L$ y $c$ es un número real, entonces se cumple que $\lim_{n \to \infty} c \cdot a_n = c \cdot L$.
* **Propiedad del cociente:** Si $\lim_{n \to \infty} a_n = L$ y $\lim_{n \to \infty} b_n = M$, donde $M \neq 0$, entonces se cumple que $\lim_{n \to \infty} \frac{a_n}{b_n} = \frac{L}{M}$.
* **Teorema del límite:** Si $f$ es una función continua en un intervalo cerrado $[a, b]$ y $x_0 \in [a, b]$, entonces se cumple que $\lim_{n \to \infty} f(x_n) = f(x_0)$, donde $\{x_n\}_{n=1}^{\infty}$ es una sucesión de números reales que converge a $x_0$.

---


```python
import matplotlib.pyplot as plt
import numpy as np

# Definición de la sucesión a_n y b_n para las propiedades
def sucesion_a(n):
    return 1 / (n + 1)

def sucesion_b(n):
    return 0.5 / (n + 1)

# Valores de n para la gráfica
n_values = np.arange(1, 21)  # Valores de n desde 1 hasta 20

# Términos de las sucesiones a_n y b_n
sucesion_a_values = [sucesion_a(n) for n in n_values]
sucesion_b_values = [sucesion_b(n) for n in n_values]

# Valores de las propiedades de los límites
L = 0.5
M = 0.2
c = 2

# Gráfica de las propiedades de los límites
plt.figure(figsize=(12, 8))

# Propiedad aditiva: a_n + b_n
plt.subplot(2, 2, 1)
plt.plot(n_values, sucesion_a_values, marker='o', linestyle='-', color='b', label='$a_n$')
plt.plot(n_values, sucesion_b_values, marker='s', linestyle='-', color='g', label='$b_n$')
plt.plot(n_values, np.add(sucesion_a_values, sucesion_b_values), marker='d', linestyle='-', color='r', label='$a_n + b_n$')
plt.hlines(L + M, n_values[0], n_values[-1], colors='m', linestyles='dashed', label='$L + M$')
plt.xlabel('n')
plt.ylabel('Valores')
plt.title('Propiedad aditiva')
plt.legend()
plt.grid(True)

# Propiedad multiplicativa: c * a_n
plt.subplot(2, 2, 2)
plt.plot(n_values, sucesion_a_values, marker='o', linestyle='-', color='b', label='$a_n$')
plt.plot(n_values, np.multiply(c, sucesion_a_values), marker='s', linestyle='-', color='g', label='$c \cdot a_n$')
plt.hlines(c * L, n_values[0], n_values[-1], colors='m', linestyles='dashed', label='$c \cdot L$')
plt.xlabel('n')
plt.ylabel('Valores')
plt.title('Propiedad multiplicativa')
plt.legend()
plt.grid(True)

# Propiedad del cociente: a_n / b_n
plt.subplot(2, 2, 3)
plt.plot(n_values, sucesion_a_values, marker='o', linestyle='-', color='b', label='$a_n$')
plt.plot(n_values, sucesion_b_values, marker='s', linestyle='-', color='g', label='$b_n$')
plt.plot(n_values, np.divide(sucesion_a_values, sucesion_b_values), marker='d', linestyle='-', color='r', label='$\\frac{a_n}{b_n}$')
plt.hlines(L / M, n_values[0], n_values[-1], colors='m', linestyles='dashed', label='$\\frac{L}{M}$')
plt.xlabel('n')
plt.ylabel('Valores')
plt.title('Propiedad del cociente')
plt.legend()
plt.grid(True)

# Teorema del límite: f(x_n) -> f(x_0)
plt.subplot(2, 2, 4)
x_values = np.linspace(0, 2, 100)
y_values = np.sin(np.pi * x_values)  # Función seno como ejemplo
x_n = np.array([1.5, 1.75, 1.9])  # Ejemplo de sucesión x_n
plt.plot(x_values, y_values, color='b', label='$f(x)$')
plt.scatter(x_n, np.sin(np.pi * x_n), color='r', label='$f(x_n)$')
plt.hlines(np.sin(np.pi * 1.5), 0, 2, colors='m', linestyles='dashed', label='$f(x_0)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Teorema del límite')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### 🔄 **Convergencia y divergencia de sucesiones** 🔄

Una sucesión se considera convergente si posee un límite finito. Si el límite es infinito, la sucesión se clasifica como divergente. También existe el caso en que la sucesión no tenga un límite, en cuyo caso se denomina oscilante.

---

```python
import matplotlib.pyplot as plt
import numpy as np

# Función para generar una sucesión convergente
def sucesion_convergente(n):
    return 1 / (n + 1)

# Función para generar una sucesión divergente
def sucesion_divergente(n):
    return n + 1

# Función para generar una sucesión oscilante
def sucesion_oscilante(n):
    return (-1) ** n

# Valores de n para la gráfica
n_values = np.arange(1, 21)  # Valores de n desde 1 hasta 20

# Términos de las sucesiones
sucesion_convergente_values = [sucesion_convergente(n) for n in n_values]
sucesion_divergente_values = [sucesion_divergente(n) for n in n_values]
sucesion_oscilante_values = [sucesion_oscilante(n) for n in n_values]

# Gráfico de convergencia y divergencia de sucesiones
plt.figure(figsize=(10, 6))

# Sucesión convergente
plt.plot(n_values, sucesion_convergente_values, marker='o', linestyle='-', color='b', label='Convergente')

# Sucesión divergente
plt.plot(n_values, sucesion_divergente_values, marker='s', linestyle='-', color='g', label='Divergente')

# Sucesión oscilante
plt.plot(n_values, sucesion_oscilante_values, marker='d', linestyle='-', color='r', label='Oscilante')

plt.xlabel('n')
plt.ylabel('Valores')
plt.title('Convergencia y Divergencia de Sucesiones')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

### ➕ **Series** ➕

![Texto alternativo](https://content.gnoss.ws/imagenes/Usuarios/ImagenesCKEditor/e39dea24-3293-4ccc-971a-48f63fc6c932/dc92efa2-ac10-480d-9027-c424f5eba151.gif)


Una serie se define como la suma de los términos de una sucesión. Se representa matemáticamente como:

$$ \sum_{n=1}^{\infty} a_n $$

donde $a_n$ es el término n-ésimo de la sucesión. La serie converge si la sucesión de sus sumas parciales converge. En otras palabras, si la sucesión $\{S_n\}_{n=1}^{\infty}$, donde $S_n = \sum_{k=1}^{n} a_k$, converge a un límite $L$, entonces la serie converge a $L$. Se representa como:

$$ \sum_{n=1}^{\infty} a_n = L $$

#### 📏 **Criterios de convergencia de series** 📏

Existen diversos criterios matemáticos para determinar si una serie converge o diverge. Entre los más comunes se encuentran:

* **Criterio de la razón:** La serie $\sum_{n=1}^{\infty} a_n$ converge si y solo si $ \lim_{n \to \infty} \left| \frac{a_{n+1}}{a_n} \right| < 1$.
* **Criterio de la raíz:** La serie $\sum_{n=1}^{\infty} a_n$ converge si y solo si $ \lim_{n \to \infty} \sqrt[n]{|a_n|} < 1$.
* **Criterio de la prueba:** La serie $\sum_{n=1}^{\infty} a_n$ converge si existe una sucesión positiva convergente $\{b_n\}_{n=1}^{\infty}$ tal que para todo $n$, se cumple la desigualdad:

$$ |a_n| \leq |b_n| $$

Los límites y series son conceptos fundamentales en el análisis matemático con un sinfín de aplicaciones dentro de la propia disciplina. A continuación, se exploran algunos ejemplos:


### ✏️ **1. Cálculo:** ✏️

* **Teoremas de convergencia uniforme:** Estos teoremas establecen condiciones bajo las cuales la convergencia de una serie es uniforme, lo que tiene importantes implicaciones matemáticas.

```python
from graphviz import Digraph

# Operaciones aritméticas
a = 10
b = 5
suma = a + b
resta = a - b
multiplicacion = a * b
division = a / b
division_entera = a // b
modulo = a % b
potencia = a ** b

# Crear el diagrama de flujo
dot = Digraph()

# Nodo inicial
dot.node('A', 'Inicio\n(a=10, b=5)', color='black', fillcolor='lightblue', style='filled', shape='ellipse')

# Operaciones
dot.node('B', 'Suma\na + b = {}'.format(suma), color='black', fillcolor='lightgreen', style='filled')
dot.node('C', 'Resta\na - b = {}'.format(resta), color='black', fillcolor='lightcoral', style='filled')
dot.node('D', 'Multiplicación\na * b = {}'.format(multiplicacion), color='black', fillcolor='lightsalmon', style='filled')
dot.node('E', 'División\na / b = {:.2f}'.format(division), color='black', fillcolor='lightpink', style='filled')
dot.node('F', 'División entera\na // b = {}'.format(division_entera), color='black', fillcolor='lightyellow', style='filled')
dot.node('G', 'Módulo\na % b = {}'.format(modulo), color='black', fillcolor='lightgoldenrodyellow', style='filled')
dot.node('H', 'Potenciación\na ** b = {}'.format(potencia), color='black', fillcolor='lightcyan', style='filled')

# Enlaces
dot.edge('A', 'B', color='blue')
dot.edge('A', 'C', color='green')
dot.edge('A', 'D', color='red')
dot.edge('A', 'E', color='purple')
dot.edge('A', 'F', color='orange')
dot.edge('A', 'G', color='pink')
dot.edge('A', 'H', color='cyan')

# Renderizar el gráfico
dot.render('operaciones_aritmeticas_colores', format='png', cleanup=True)

# Mostrar el gráfico en la notebook
from IPython.display import Image
Image(filename='operaciones_aritmeticas_colores.png')
```

## **🧊 Organizador Gráfico - Operaciones Lógicas 🧊**

```python
from graphviz import Digraph

# Operaciones lógicas
p = True
q = False
resultado_and = p and q
resultado_or = p or q
resultado_not = not p

# Imprimir resultados
print("AND lógico:", resultado_and)  # Y lógico
print("OR lógico:", resultado_or)  # O lógico
print("NOT lógico:", resultado_not)  # Negación lógico

# Crear el diagrama de flujo
dot = Digraph()

# Nodo inicial
dot.node('A', 'Inicio\n(p=True, q=False)', color='black', fillcolor='lightblue', style='filled', shape='ellipse')

# Operaciones
dot.node('B', 'AND lógico\np and q = {}'.format(resultado_and), color='black', fillcolor='lightgreen', style='filled')
dot.node('C', 'OR lógico\np or q = {}'.format(resultado_or), color='black', fillcolor='lightcoral', style='filled')
dot.node('D', 'NOT lógico\nnot p = {}'.format(resultado_not), color='black', fillcolor='lightpink', style='filled')

# Enlaces
dot.edge('A', 'B', color='blue')
dot.edge('A', 'C', color='green')
dot.edge('A', 'D', color='red')

# Renderizar el gráfico
dot.render('operaciones_logicas_colores', format='png', cleanup=True)

# Mostrar el gráfico en la notebook
from IPython.display import Image
Image(filename='operaciones_logicas_colores.png')
```
