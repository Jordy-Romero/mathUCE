# **Introducci√≥n a las Derivadas** ‚úèÔ∏è

## **Definiciones de derivada** üîç

### **Derivada en un punto a partir de cocientes diferenciales**
La derivada de una funci√≥n $f$ en el punto $a$ es la pendiente de la recta tangente a la gr√°fica de $f$ en el punto $a$. El valor de esta pendiente ser√° aproximadamente igual a la pendiente de una recta secante a la gr√°fica que pase por el punto $(a, f(a))$ y por un punto cercano $(x, f(x))$; por conveniencia suele expresarse $x = a + h$, donde $h$ es un n√∫mero cercano a 0. A partir de estos dos puntos se calcula la pendiente de la recta secante como:

$$
\frac{f(x) - f(a)}{x - a} = \frac{f(a + h) - f(a)}{h}.
$$

(Esta expresi√≥n se denomina ¬´cociente diferencial¬ª o ¬´cociente de Newton¬ª.) A medida que el n√∫mero $h$ se acerca a cero, el valor de esta pendiente se aproximar√° mejor al de la recta tangente. Esto permite definir la derivada de la funci√≥n $f$ en el punto $a$, denotada como $f'(a)$, como el l√≠mite de estos cocientes cuando $h$ tiende a cero:

$$
f'(a) = \lim_{h \to 0} \frac{f(a + h) - f(a)}{h}.
$$

No obstante, esta definici√≥n s√≥lo es v√°lida cuando el l√≠mite es un n√∫mero real: en los puntos $a$ donde el l√≠mite no existe, la funci√≥n $f$ no tiene derivada.

## **Derivada de una funci√≥n**

Dada una funci√≥n $f$, se puede definir una nueva funci√≥n que, en cada punto $x$, toma el valor de la derivada $f'(x)$. Esta funci√≥n se denota $f'$ y se denomina funci√≥n derivada de $f$ o simplemente derivada de $f$. Esto es, la derivada de $f$ es la funci√≥n dada por:

$$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}.
$$

Esta funci√≥n s√≥lo est√° definida en los puntos del dominio de $f$ donde el l√≠mite existe; en otras palabras, el dominio de $f'$ est√° contenido en el de $f$.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

# Definir la funci√≥n cuadr√°tica que queremos derivar
def f(x):
    return x**2 + 2*x + 1  # Funci√≥n cuadr√°tica: x^2 + 2x + 1

# Crear un rango de valores x
x = np.linspace(-10, 10, 400)  # Genera 400 puntos entre -10 y 10
y = f(x)  # Evaluar la funci√≥n f(x) en los puntos x

# Calcular la derivada utilizando scipy
def df(x):
    # Calcula la derivada en el punto x usando el m√©todo de diferencias finitas
    return derivative(f, x, dx=1e-6)

# Crear un rango de valores x para la derivada
x_deriv = np.linspace(-10, 10, 400)  # Genera 400 puntos entre -10 y 10
y_deriv = np.array([df(xi) for xi in x_deriv])  # Evaluar la derivada en los puntos x_deriv

# Crear una figura con dos subplots
plt.figure(figsize=(12, 6))  # Tama√±o de la figura: 12x6 pulgadas

# Graficar la funci√≥n original
plt.subplot(1, 2, 1)  # Crear el primer subplot en una figura de 1 fila y 2 columnas
plt.plot(x, y, label='f(x) = x^2 + 2x + 1')  # Graficar la funci√≥n original
plt.title('Funci√≥n Cuadr√°tica')  # T√≠tulo del gr√°fico
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('f(x)')  # Etiqueta del eje y
plt.grid(True)  # Mostrar la cuadr√≠cula
plt.legend()  # Mostrar la leyenda

# Graficar la derivada
plt.subplot(1, 2, 2)  # Crear el segundo subplot en una figura de 1 fila y 2 columnas
plt.plot(x_deriv, y_deriv, label="f'(x) = 2x + 2")  # Graficar la derivada de la funci√≥n
plt.title('Derivada')  # T√≠tulo del gr√°fico
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel("f'(x)")  # Etiqueta del eje y
plt.grid(True)  # Mostrar la cuadr√≠cula
plt.legend()  # Mostrar la leyenda

plt.tight_layout()  # Ajustar el dise√±o para evitar superposici√≥n
plt.show()  # Mostrar la figura
````

# **Notaci√≥n** üìù

Existen diversas formas para nombrar a la derivada. Siendo $f$ una funci√≥n, se escribe la derivada de la funci√≥n $f$ respecto al valor $x$ en varios modos.



## **Notaci√≥n de Lagrange** üí°

La notaci√≥n m√°s simple para diferenciaci√≥n, en uso actual, se debe a Lagrange, y consiste en denotar la derivada de una funci√≥n $f(x)$ como $f'(x)$: se lee ¬´f prima de x¬ª. Esta notaci√≥n se extiende a derivadas de orden superior, dando lugar a $f''(x)$ (¬´f segunda de x¬ª o ¬´f dos prima de x¬ª) para la derivada segunda, y a $f'''(x)$ para la derivada tercera.

La derivada cuarta y siguientes se pueden denotar de dos formas:

- Con n√∫meros romanos: $f^{\text{IV}}(x)$.
- Con n√∫meros entre par√©ntesis: $f^{(4)}(x)$.

Esta √∫ltima opci√≥n da lugar tambi√©n a la notaci√≥n $f^{(n)}(x)$ para denotar la derivada n-√©sima de $f$.


````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la variable simb√≥lica y la funci√≥n
x = sp.Symbol('x')
f = sp.exp(-x**2)

# Calcular las derivadas
f_prime = sp.diff(f, x, 1)  # Primera derivada
f_double_prime = sp.diff(f, x, 2)  # Segunda derivada
f_triple_prime = sp.diff(f, x, 3)  # Tercera derivada
f_quadruple_prime = sp.diff(f, x, 4)  # Cuarta derivada

# Convertir las derivadas a funciones num√©ricas
f_prime_func = sp.lambdify(x, f_prime, 'numpy')
f_double_prime_func = sp.lambdify(x, f_double_prime, 'numpy')
f_triple_prime_func = sp.lambdify(x, f_triple_prime, 'numpy')
f_quadruple_prime_func = sp.lambdify(x, f_quadruple_prime, 'numpy')

# Crear un rango de valores x para graficar
x_vals = np.linspace(-3, 3, 400)
y = np.exp(-x_vals**2)
y_prime = f_prime_func(x_vals)
y_double_prime = f_double_prime_func(x_vals)
y_triple_prime = f_triple_prime_func(x_vals)
y_quadruple_prime = f_quadruple_prime_func(x_vals)

# Crear una figura con cinco subplots
plt.figure(figsize=(15, 12))

# Graficar la funci√≥n original
plt.subplot(3, 2, 1)
plt.plot(x_vals, y, label='f(x) = e^(-x^2)')
plt.title('Funci√≥n Original')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Graficar la primera derivada
plt.subplot(3, 2, 2)
plt.plot(x_vals, y_prime, label="f'(x)", color='orange')
plt.title('Primera Derivada')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.grid(True)
plt.legend()

# Graficar la segunda derivada
plt.subplot(3, 2, 3)
plt.plot(x_vals, y_double_prime, label="f''(x)", color='green')
plt.title('Segunda Derivada')
plt.xlabel('x')
plt.ylabel("f''(x)")
plt.grid(True)
plt.legend()

# Graficar la tercera derivada
plt.subplot(3, 2, 4)
plt.plot(x_vals, y_triple_prime, label="f'''(x)", color='red')
plt.title('Tercera Derivada')
plt.xlabel('x')
plt.ylabel("f'''(x)")
plt.grid(True)
plt.legend()

# Graficar la cuarta derivada
plt.subplot(3, 2, 5)
plt.plot(x_vals, y_quadruple_prime, label="f''''(x)", color='purple')
plt.title('Cuarta Derivada')
plt.xlabel('x')
plt.ylabel("f''''(x)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````

## **Notaci√≥n de Leibniz** üìö

Otra notaci√≥n com√∫n para la diferenciaci√≥n es debida a Leibniz. Para la funci√≥n derivada de $f$, se escribe:

$$
\frac{\mathrm{d}\left(f(x)\right)}{\mathrm{d}x}
$$

Tambi√©n puede encontrarse como $ \frac{\mathrm{d}y}{\mathrm{d}x}$, $ \frac{\mathrm{d}f}{\mathrm{d}x}$ o $ \frac{\mathrm{d}}{\mathrm{d}x} f(x)$. Se lee ¬´derivada de $y$ o $f$ con respecto a $x$¬ª. Esta notaci√≥n tiene la ventaja de sugerir la derivada de una funci√≥n con respecto a otra como un cociente de diferenciales.

Con esta notaci√≥n, se puede escribir la derivada de $f$ en el punto $a$ de dos modos diferentes:

$$
\frac{\mathrm{d}f}{\mathrm{d}x}\bigg|_{x=a} = \left(\frac{\mathrm{d}\left(f(x)\right)}{\mathrm{d}x}\right)(a)
$$

Si $y = f(x)$, se puede escribir la derivada como

$$
\frac{\mathrm{d}y}{\mathrm{d}x}
$$

Las derivadas sucesivas se expresan como

$$
\frac{\mathrm{d}^n f}{\mathrm{d}x^n} \text{ o } \frac{\mathrm{d}^n y}{\mathrm{d}x^n}
$$


para la en√©sima derivada de $f$ o de $y$, respectivamente. Hist√≥ricamente, esto viene del hecho que, por ejemplo, la tercera derivada es

$$
\frac{\mathrm{d} \left(\frac{\mathrm{d} \left( \frac{\mathrm{d} \left(f(x)\right)}{\mathrm{d}x}\right)}{\mathrm{d}x}\right)}{\mathrm{d}x}
$$

la cual se puede escribir como

$$
\left(\frac{\mathrm{d}}{\mathrm{d}x}\right)^3 \left(f(x)\right) = \frac{\mathrm{d}^3}{\left(\mathrm{d}x\right)^3} \left(f(x)\right).
$$

La notaci√≥n de Leibniz es muy √∫til, por cuanto permite especificar la variable de diferenciaci√≥n (en el denominador); lo cual es pertinente en caso de diferenciaci√≥n parcial. Tambi√©n facilita recordar la regla de la cadena, porque los t√©rminos ¬´d¬ª parecen cancelarse simb√≥licamente:

$$
\frac{\mathrm{d}y}{\mathrm{d}x} = \frac{\mathrm{d}y}{\mathrm{d}u} \cdot \frac{\mathrm{d}u}{\mathrm{d}x}.
$$

En la formulaci√≥n popular del c√°lculo mediante l√≠mites, los t√©rminos ¬´d¬ª no pueden cancelarse literalmente, porque por s√≠ mismos son indefinidos; son definidos solamente cuando se usan juntos para expresar una derivada. En an√°lisis no-est√°ndar, no obstante, se puede ver como n√∫meros infinitesimales que se cancelan.

Ciertamente, Leibniz consider√≥ la derivada $ \frac{\mathrm{d}y}{\mathrm{d}x} $ como el cociente de dos ¬´infinit√©simos¬ª $ \mathrm{d}y $ y $ \mathrm{d}x $, llamados ¬´diferenciales¬ª. Estos infinit√©simos no eran n√∫meros sino cantidades m√°s peque√±as que cualquier n√∫mero positivo.

## **Notaci√≥n de Newton** üìñ

La notaci√≥n de Newton para la diferenciaci√≥n respecto al tiempo era poner un punto arriba del nombre de la funci√≥n:

$$
\dot{x} = x'(t)
$$

$$
\ddot{x} = x''(t)
$$

## **Regla de la Constante** üî¢
   
   Si tienes una funci√≥n que es una constante $c$, la derivada de una constante es siempre 0.

   $$
   \frac{d}{dx}(c) = 0
   $$

   **Explicaci√≥n**: La derivada mide el cambio de una funci√≥n con respecto a $x$. Como una constante no cambia, su derivada es 0.

## **Regla de la Potencia** üìà
   
   Para una funci√≥n de la forma $x^n$, donde $n$ es un n√∫mero real, la derivada es:

   $$
   \frac{d}{dx}(x^n) = nx^{n-1}
   $$

   **Explicaci√≥n**: Multiplicamos el exponente $n$ por la base $x$ elevada a $n-1$. Esto sigue la regla general para derivar potencias de $x$.

## **Regla del Producto** üìâ
   
   Si tienes un producto de dos funciones $u(x)$ y $v(x)$, la derivada es:

   $$
   \frac{d}{dx}[u(x) \cdot v(x)] = u'(x) \cdot v(x) + u(x) \cdot v'(x)
   $$

   **Explicaci√≥n**: La derivada del producto de dos funciones es la derivada de la primera funci√≥n multiplicada por la segunda funci√≥n m√°s la primera funci√≥n multiplicada por la derivada de la segunda funci√≥n.

## **Regla del Cociente** ‚ûó
   
   Para una funci√≥n que es el cociente de dos funciones $u(x)$ y $v(x)$, donde $v(x) \neq 0$, la derivada es:

   $$
   \frac{d}{dx}\left(\frac{u(x)}{v(x)}\right) = \frac{u'(x) \cdot v(x) - u(x) \cdot v'(x)}{[v(x)]^2}
   $$

   **Explicaci√≥n**: Derivamos el numerador y el denominador por separado, y luego aplicamos la f√≥rmula para combinar los resultados en un solo cociente.

## **Regla de la Cadena** üîó
   
   Si tienes una funci√≥n compuesta $g(h(x))$, la derivada es:

   $$
   \frac{d}{dx}[g(h(x))] = g'(h(x)) \cdot h'(x)
   $$

   **Explicaci√≥n**: Derivamos la funci√≥n exterior $g$ en el punto $h(x)$ y multiplicamos por la derivada de la funci√≥n interior $h(x)$.

````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la variable simb√≥lica
x = sp.symbols('x')

# Funciones a derivar
# 1. Regla de la Constante: La derivada de una constante es 0
f1 = 5  # Funci√≥n constante

# 2. Regla de la Potencia: Derivada de x^n es n*x^(n-1)
f2 = x**3 - 4*x**2 + 2*x  # Funci√≥n polin√≥mica

# 3. Regla del Producto: Derivada de (u*v) es u'*v + u*v'
f3 = x**2 * sp.sin(x)     # Funci√≥n producto

# 4. Regla del Cociente: Derivada de (u/v) es (u'*v - u*v') / v^2
f4 = (x**2 + 1) / (x - 1)  # Funci√≥n cociente

# 5. Regla de la Cadena: Derivada de f(g(x)) es f'(g(x)) * g'(x)
f5 = sp.sin(x**2)         # Funci√≥n compuesta

# Calcular las derivadas simb√≥licas
f1_prime = sp.diff(f1, x)
f2_prime = sp.diff(f2, x)
f3_prime = sp.diff(f3, x)
f4_prime = sp.diff(f4, x)
f5_prime = sp.diff(f5, x)

# Convertir las funciones simb√≥licas a funciones num√©ricas
f1_l = sp.lambdify(x, f1, 'numpy')
f1_prime_l = sp.lambdify(x, f1_prime, 'numpy')

f2_l = sp.lambdify(x, f2, 'numpy')
f2_prime_l = sp.lambdify(x, f2_prime, 'numpy')

f3_l = sp.lambdify(x, f3, 'numpy')
f3_prime_l = sp.lambdify(x, f3_prime, 'numpy')

f4_l = sp.lambdify(x, f4, 'numpy')
f4_prime_l = sp.lambdify(x, f4_prime, 'numpy')

f5_l = sp.lambdify(x, f5, 'numpy')
f5_prime_l = sp.lambdify(x, f5_prime, 'numpy')

# Crear un rango de valores para x
x_values = np.linspace(-2, 2, 400)  # Rango general para funciones
x_values_nonzero = np.linspace(0.1, 2, 400)  # Para evitar divisi√≥n por cero en la funci√≥n cociente

# Crear la gr√°fica
plt.figure(figsize=(14, 12))

# 1. Regla de la Constante
plt.subplot(3, 2, 1)
plt.plot(x_values, f1_l(x_values) * np.ones_like(x_values), label='$f(x) = 5$', color='blue')
plt.plot(x_values, f1_prime_l(x_values) * np.ones_like(x_values), label="$f'(x) = 0$", color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regla de la Constante')
plt.legend()
plt.grid(True)

# 2. Regla de la Potencia
plt.subplot(3, 2, 2)
plt.plot(x_values, f2_l(x_values), label='$f(x) = x^3 - 4x^2 + 2x$', color='blue')
plt.plot(x_values, f2_prime_l(x_values), label="$f'(x) = 3x^2 - 8x + 2$", color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regla de la Potencia')
plt.legend()
plt.grid(True)

# 3. Regla del Producto
plt.subplot(3, 2, 3)
plt.plot(x_values, f3_l(x_values), label='$f(x) = x^2 \sin(x)$', color='blue')
plt.plot(x_values, f3_prime_l(x_values), label="$f'(x) = x^2 \cos(x) + 2x \sin(x)$", color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regla del Producto')
plt.legend()
plt.grid(True)

# 4. Regla del Cociente
plt.subplot(3, 2, 4)
plt.plot(x_values_nonzero, f4_l(x_values_nonzero), label='$f(x) = \\frac{x^2 + 1}{x - 1}$', color='blue')
plt.plot(x_values_nonzero, f4_prime_l(x_values_nonzero), label="$f'(x) = \\frac{2x(x - 1) - (x^2 + 1)}{(x - 1)^2}$", color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regla del Cociente')
plt.legend()
plt.grid(True)

# 5. Regla de la Cadena
plt.subplot(3, 2, 5)
plt.plot(x_values, f5_l(x_values), label='$f(x) = \sin(x^2)$', color='blue')
plt.plot(x_values, f5_prime_l(x_values), label="$f'(x) = 2x \cos(x^2)$", color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regla de la Cadena')
plt.legend()
plt.grid(True)

# Ajustar el dise√±o y mostrar la gr√°fica
plt.tight_layout()
plt.show()
````

## **Derivadas de Funciones Trigonom√©tricas** üåê
   
   Las derivadas b√°sicas de funciones trigonom√©tricas son:

   $$
   \frac{d}{dx} (\sin x) = \cos x
   $$

   $$
   \frac{d}{dx} (\cos x) = -\sin x
   $$

   $$
   \frac{d}{dx} (\tan x) = \sec^2 x
   $$

   $$
   \frac{d}{dx} (\cot x) = -\csc^2 x
   $$

   $$
   \frac{d}{dx} (\sec x) = \sec x \cdot \tan x
   $$

   $$
   \frac{d}{dx} (\csc x) = -\csc x \cdot \cot x
   $$

   **Explicaci√≥n**: Estas derivadas son derivadas est√°ndar para funciones trigonom√©tricas, obtenidas usando identidades y propiedades trigonom√©tricas.

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores para x
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Funciones trigonom√©tricas
sin_x = np.sin(x)  # Seno de x
cos_x = np.cos(x)  # Coseno de x
tan_x = np.tan(x)  # Tangente de x
cot_x = 1 / tan_x  # Cotangente de x
sec_x = 1 / cos_x  # Secante de x
csc_x = 1 / sin_x  # Cosecante de x

# Derivadas
d_sin_x = cos_x  # Derivada de sin(x)
d_cos_x = -sin_x  # Derivada de cos(x)
d_tan_x = sec_x ** 2  # Derivada de tan(x)
d_cot_x = -csc_x ** 2  # Derivada de cot(x)
d_sec_x = sec_x * tan_x  # Derivada de sec(x)
d_csc_x = -csc_x * cot_x  # Derivada de csc(x)

# Crear el gr√°fico
fig, axs = plt.subplots(3, 2, figsize=(14, 12))  # Crear una figura con 3 filas y 2 columnas
fig.suptitle('Funciones Trigonom√©tricas y sus Derivadas')  # T√≠tulo general

# Graficar sin(x) y su derivada
axs[0, 0].plot(x, sin_x, label='sin(x)')
axs[0, 0].plot(x, d_sin_x, label='cos(x)', linestyle='--')  # Derivada en l√≠nea discontinua
axs[0, 0].set_title('sin(x) y su derivada')  # T√≠tulo del gr√°fico
axs[0, 0].legend()  # Mostrar leyenda

# Graficar cos(x) y su derivada
axs[0, 1].plot(x, cos_x, label='cos(x)')
axs[0, 1].plot(x, d_cos_x, label='-sin(x)', linestyle='--')  # Derivada en l√≠nea discontinua
axs[0, 1].set_title('cos(x) y su derivada')  # T√≠tulo del gr√°fico
axs[0, 1].legend()  # Mostrar leyenda

# Graficar tan(x) y su derivada
axs[1, 0].plot(x, tan_x, label='tan(x)')
axs[1, 0].plot(x, d_tan_x, label='sec^2(x)', linestyle='--')  # Derivada en l√≠nea discontinua
axs[1, 0].set_title('tan(x) y su derivada')  # T√≠tulo del gr√°fico
axs[1, 0].set_ylim(-10, 10)  # Limitar el rango y mejorar la visualizaci√≥n
axs[1, 0].legend()  # Mostrar leyenda

# Graficar cot(x) y su derivada
axs[1, 1].plot(x, cot_x, label='cot(x)')
axs[1, 1].plot(x, d_cot_x, label='-csc^2(x)', linestyle='--')  # Derivada en l√≠nea discontinua
axs[1, 1].set_title('cot(x) y su derivada')  # T√≠tulo del gr√°fico
axs[1, 1].set_ylim(-10, 10)  # Limitar el rango y mejorar la visualizaci√≥n
axs[1, 1].legend()  # Mostrar leyenda

# Graficar sec(x) y su derivada
axs[2, 0].plot(x, sec_x, label='sec(x)')
axs[2, 0].plot(x, d_sec_x, label='sec(x) * tan(x)', linestyle='--')  # Derivada en l√≠nea discontinua
axs[2, 0].set_title('sec(x) y su derivada')  # T√≠tulo del gr√°fico
axs[2, 0].legend()  # Mostrar leyenda

# Graficar csc(x) y su derivada
axs[2, 1].plot(x, csc_x, label='csc(x)')
axs[2, 1].plot(x, d_csc_x, label='-csc(x) * cot(x)', linestyle='--')  # Derivada en l√≠nea discontinua
axs[2, 1].set_title('csc(x) y su derivada')  # T√≠tulo del gr√°fico
axs[2, 1].legend()  # Mostrar leyenda

# Ajustar el espacio entre gr√°ficos
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Ajustar el dise√±o para que no se sobrepongan
plt.show()  # Mostrar el gr√°fico
````

# **Derivadas de Funciones Exponenciales y Logar√≠tmicas** üå±
  
   Para funciones exponenciales y logar√≠tmicas:

   $$
   \frac{d}{dx} (e^x) = e^x
   $$

   $$
   \frac{d}{dx} (\ln x) = \frac{1}{x}
   $$

   **Explicaci√≥n**: La derivada de $e^x$ es la misma funci√≥n $e^x$, y la derivada del logaritmo natural $\ln x$ es $\frac{1}{x}$.

````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la variable simb√≥lica para la derivada
x = sp.symbols('x')

# Definir las funciones para las derivadas
# Funci√≥n exponencial
f_exp = sp.exp(x)  # e^x

# Funci√≥n logar√≠tmica
f_log = sp.ln(x)   # ln(x)

# Calcular las derivadas de las funciones
# La derivada de e^x es e^x
f_exp_prime = sp.diff(f_exp, x)

# La derivada de ln(x) es 1/x
f_log_prime = sp.diff(f_log, x)

# Convertir las funciones simb√≥licas a funciones num√©ricas
# Estas funciones se pueden usar para graficar
f_exp_l = sp.lambdify(x, f_exp, 'numpy')
f_exp_prime_l = sp.lambdify(x, f_exp_prime, 'numpy')

f_log_l = sp.lambdify(x, f_log, 'numpy')
f_log_prime_l = sp.lambdify(x, f_log_prime, 'numpy')

# Crear un rango de valores para x
# Para la funci√≥n exponencial, se usa un rango amplio
x_exp_values = np.linspace(-1, 2, 400)

# Para la funci√≥n logar√≠tmica, se evita valores negativos y cero
x_log_values = np.linspace(0.1, 2, 400)

# Crear la gr√°fica
plt.figure(figsize=(14, 6))  # Tama√±o de la figura

# Gr√°fico para la funci√≥n exponencial y su derivada
plt.subplot(1, 2, 1)
plt.plot(x_exp_values, f_exp_l(x_exp_values), label='$f(x) = e^x$', color='blue')  # Funci√≥n exponencial
plt.plot(x_exp_values, f_exp_prime_l(x_exp_values), label="$f'(x) = e^x$", color='red', linestyle='--')  # Derivada de la funci√≥n exponencial
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('y')  # Etiqueta del eje y
plt.title('Funci√≥n Exponencial y su Derivada')  # T√≠tulo del gr√°fico
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar la cuadr√≠cula

# Gr√°fico para la funci√≥n logar√≠tmica y su derivada
plt.subplot(1, 2, 2)
plt.plot(x_log_values, f_log_l(x_log_values), label='$f(x) = \ln(x)$', color='blue')  # Funci√≥n logar√≠tmica
plt.plot(x_log_values, f_log_prime_l(x_log_values), label="$f'(x) = \\frac{1}{x}$", color='red', linestyle='--')  # Derivada de la funci√≥n logar√≠tmica
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('y')  # Etiqueta del eje y
plt.title('Funci√≥n Logar√≠tmica y su Derivada')  # T√≠tulo del gr√°fico
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar la cuadr√≠cula

# Ajustar el dise√±o de los subgr√°ficos para que no se solapen
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()
````

# **Derivadas de Funciones Inversas** üîÑ
   
   Las derivadas de las funciones inversas b√°sicas son:

   $$
   \frac{d}{dx} (\arcsin x) = \frac{1}{\sqrt{1 - x^2}}
   $$

   $$
   \frac{d}{dx} (\arccos x) = -\frac{1}{\sqrt{1 - x^2}}
   $$

   $$
   \frac{d}{dx} (\arctan x) = \frac{1}{1 + x^2}
   $$

   **Explicaci√≥n**: Estas derivadas se derivan usando identidades trigonom√©tricas y las definiciones de las funciones inversas.

````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la variable simb√≥lica para los c√°lculos
x = sp.symbols('x')

# Definir las funciones inversas
# Funci√≥n arco seno (arcsin(x))
f_asin = sp.asin(x)

# Funci√≥n arco coseno (arccos(x))
f_acos = sp.acos(x)

# Funci√≥n arco tangente (arctan(x))
f_atan = sp.atan(x)

# Calcular las derivadas de las funciones inversas
# La derivada de arcsin(x) es 1 / sqrt(1 - x^2)
f_asin_prime = sp.diff(f_asin, x)

# La derivada de arccos(x) es -1 / sqrt(1 - x^2)
f_acos_prime = sp.diff(f_acos, x)

# La derivada de arctan(x) es 1 / (1 + x^2)
f_atan_prime = sp.diff(f_atan, x)

# Convertir las funciones simb√≥licas a funciones num√©ricas para graficar
# Esto permite evaluar las funciones con valores num√©ricos
f_asin_l = sp.lambdify(x, f_asin, 'numpy')
f_asin_prime_l = sp.lambdify(x, f_asin_prime, 'numpy')

f_acos_l = sp.lambdify(x, f_acos, 'numpy')
f_acos_prime_l = sp.lambdify(x, f_acos_prime, 'numpy')

f_atan_l = sp.lambdify(x, f_atan, 'numpy')
f_atan_prime_l = sp.lambdify(x, f_atan_prime, 'numpy')

# Crear un rango de valores para x
# Para arcsin y arccos, el dominio es [-1, 1]
x_values = np.linspace(-1, 1, 400)

# Para arctan, se puede usar un rango m√°s amplio para ver mejor el comportamiento
x_atan_values = np.linspace(-10, 10, 400)

# Crear la gr√°fica
plt.figure(figsize=(14, 12))  # Tama√±o de la figura

# Gr√°fico para la funci√≥n arco seno y su derivada
plt.subplot(3, 2, 1)
plt.plot(x_values, f_asin_l(x_values), label='$f(x) = \\arcsin(x)$', color='blue')  # Funci√≥n arco seno
plt.plot(x_values, f_asin_prime_l(x_values), label="$f'(x) = \\frac{1}{\\sqrt{1 - x^2}}$", color='red', linestyle='--')  # Derivada de arco seno
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('y')  # Etiqueta del eje y
plt.title('Funci√≥n Arco Seno y su Derivada')  # T√≠tulo del gr√°fico
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar la cuadr√≠cula

# Gr√°fico para la funci√≥n arco coseno y su derivada
plt.subplot(3, 2, 2)
plt.plot(x_values, f_acos_l(x_values), label='$f(x) = \\arccos(x)$', color='blue')  # Funci√≥n arco coseno
plt.plot(x_values, f_acos_prime_l(x_values), label="$f'(x) = -\\frac{1}{\\sqrt{1 - x^2}}$", color='red', linestyle='--')  # Derivada de arco coseno
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('y')  # Etiqueta del eje y
plt.title('Funci√≥n Arco Coseno y su Derivada')  # T√≠tulo del gr√°fico
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar la cuadr√≠cula

# Gr√°fico para la funci√≥n arco tangente y su derivada
plt.subplot(3, 2, 3)
plt.plot(x_atan_values, f_atan_l(x_atan_values), label='$f(x) = \\arctan(x)$', color='blue')  # Funci√≥n arco tangente
plt.plot(x_atan_values, f_atan_prime_l(x_atan_values), label="$f'(x) = \\frac{1}{1 + x^2}$", color='red', linestyle='--')  # Derivada de arco tangente
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('y')  # Etiqueta del eje y
plt.title('Funci√≥n Arco Tangente y su Derivada')  # T√≠tulo del gr√°fico
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar la cuadr√≠cula

# Ajustar el dise√±o de los subgr√°ficos para que no se solapen
plt.tight_layout()

# Mostrar la gr√°fica final
plt.show()
````

# **Ejemplos Explicados** ‚úÖ

### **1. Regla de la Constante**

**Ejemplo:**
Sea $f(x) = 5$. Queremos encontrar $\frac{d}{dx}(5)$.

**Proceso:**
La derivada de una constante es siempre cero. Por lo tanto:
$$
\frac{d}{dx}(5) = 0
$$

### **2. Regla de la Potencia**

**Ejemplo:**
Sea $f(x) = x^3$. Queremos encontrar $\frac{d}{dx}(x^3)$.

**Proceso:**
Aplicamos la regla de la potencia con $n = 3$:
$$
\frac{d}{dx}(x^3) = 3 x^{3-1} = 3 x^2
$$

### **3. Regla del Producto**

**Ejemplo:**
Sea $f(x) = (3x) \cdot (x^2)$. Queremos encontrar $\frac{d}{dx}[(3x) \cdot (x^2)]$.

**Proceso:**
Definimos $u(x) = 3x$ y $v(x) = x^2$.

- La derivada de $u(x) = 3x$ es $u'(x) = 3$.
- La derivada de $v(x) = x^2$ es $v'(x) = 2x$.

Aplicamos la regla del producto:
$$
\frac{d}{dx}[(3x) \cdot (x^2)] = 3 \cdot (x^2) + (3x) \cdot (2x) = 3x^2 + 6x^2 = 9x^2
$$

### **4. Regla del Cociente**

**Ejemplo:**
Sea $f(x) = \frac{4x^2}{x^3}$. Queremos encontrar $\frac{d}{dx}\left(\frac{4x^2}{x^3}\right)$.

**Proceso:**
Definimos $u(x) = 4x^2$ y $v(x) = x^3$.

- La derivada de $u(x) = 4x^2$ es $u'(x) = 8x$.
- La derivada de $v(x) = x^3$ es $v'(x) = 3x^2$.

Aplicamos la regla del cociente:
$$
\frac{d}{dx}\left(\frac{4x^2}{x^3}\right) = \frac{(8x) \cdot (x^3) - (4x^2) \cdot (3x^2)}{(x^3)^2} = \frac{8x^4 - 12x^4}{x^6} = \frac{-4x^4}{x^6} = -\frac{4}{x^2}
$$

### **5. Regla de la Cadena**

**Ejemplo:**
Sea $f(x) = \cos(x^2)$. Queremos encontrar $\frac{d}{dx}[\cos(x^2)]$.

**Proceso:**
Definimos $g(u) = \cos(u)$ con $u = h(x) = x^2$.

- La derivada de $g(u) = \cos(u)$ es $g'(u) = -\sin(u)$.
- La derivada de $h(x) = x^2$ es $h'(x) = 2x$.

Aplicamos la regla de la cadena:
$$
\frac{d}{dx}[\cos(x^2)] = -\sin(x^2) \cdot 2x = -2x \sin(x^2)
$$

La suma hasta un t√©rmino espec√≠fico cualquiera se llama suma parcial.

**2. Series Geom√©tricas:** Una serie geom√©trica es la suma de los t√©rminos de una secuencia geom√©trica, en la cual la raz√≥n entre t√©rminos consecutivos es constante.

F√≥rmula de la serie geom√©trica:

![Texto alternativo](https://mathcracker.com/images/legacy/geometric-series.png)


````python
# Funci√≥n para calcular la suma de una serie geom√©trica
def suma_serie_geometrica(a, r, n):
    """
    Calcula la suma de los primeros n t√©rminos de una serie geom√©trica.

    Par√°metros:
    a (float): Primer t√©rmino de la serie.
    r (float): Raz√≥n com√∫n de la serie.
    n (int): N√∫mero de t√©rminos a sumar.

    Retorna:
    float: Suma de los primeros n t√©rminos de la serie geom√©trica.
    """
    if r == 1:
        return a * n
    else:
        return a * (1 - r**n) / (1 - r)

# Ejemplo de uso
a = 2  # Primer t√©rmino
r = 3  # Raz√≥n com√∫n
n = 4  # N√∫mero de t√©rminos

suma = suma_serie_geometrica(a, r, n)
print(f"La suma de los primeros {n} t√©rminos de la serie geom√©trica es: {suma}")
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Par√°metros de la serie geom√©trica
a = 1  # primer t√©rmino
r = 0.5  # raz√≥n com√∫n

# N√∫mero de t√©rminos en la serie finita
n_terms = 20

# Crear un rango de t√©rminos para la serie finita
n = np.arange(1, n_terms + 1)
terms = a * r**(n - 1)

# Calcular la suma acumulativa de la serie finita
cumulative_sum = np.cumsum(terms)

# Crear un rango de valores para la serie infinita
x = np.linspace(0, n_terms, 100)
y = a / (1 - r) * np.ones_like(x)

# Graficar la serie geom√©trica finita
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(n, cumulative_sum, marker='o', linestyle='-', color='b', label='Suma acumulativa')
plt.title('Serie Geom√©trica Finita')
plt.xlabel('N√∫mero de t√©rminos')
plt.ylabel('Suma acumulativa')
plt.grid(True)
plt.legend()

# Graficar la serie geom√©trica infinita
plt.subplot(1, 2, 2)
plt.plot(x, y, color='r', linestyle='--', label='Suma de la serie infinita')
plt.title('Serie Geom√©trica Infinita')
plt.xlabel('N√∫mero de t√©rminos')
plt.ylabel('Suma')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````

**3. Serie Telesc√≥pica:** Una serie telesc√≥pica es aquella en la que todos sus t√©rminos se cancelan excepto el primero y el √∫ltimo Esto hace a tales series sencillas de analizar.

![Texto alternativo](https://matematicaoltre.altervista.org/wp-content/uploads/2021/09/serie-numeriche-esercizi-svolti-es1-4.png)




````python
# Importamos las librer√≠as necesarias
import numpy as np
import matplotlib.pyplot as plt

# Definimos el n√∫mero de t√©rminos
n_terms = 20

# Definimos la serie telesc√≥pica
def series_telescopica(n):
    return 1/n - 1/(n+1)

# Calculamos los t√©rminos de la serie
terms = [series_telescopica(n) for n in range(1, n_terms+1)]
partial_sums = np.cumsum(terms)

# Visualizamos los t√©rminos de la serie
plt.figure(figsize=(10, 6))
plt.plot(range(1, n_terms+1), terms, 'o-', label='T√©rminos de la Serie Telesc√≥pica')
plt.xlabel('N√∫mero de T√©rmino')
plt.ylabel('Valor del T√©rmino')
plt.title('Serie Telesc√≥pica: T√©rminos')
plt.grid(True)
plt.legend()
plt.show()

# Visualizamos la suma parcial de los t√©rminos de la serie
plt.figure(figsize=(10, 6))
plt.plot(range(1, n_terms+1), partial_sums, 'o-', label='Suma Parcial de la Serie Telesc√≥pica')
plt.xlabel('N√∫mero de T√©rmino')
plt.ylabel('Suma Parcial')
plt.title('Serie Telesc√≥pica: Suma Parcial')
plt.grid(True)
plt.legend()
plt.show()

# Suma total despu√©s de n t√©rminos
suma_total = partial_sums[-1]
print(f'Suma total despu√©s de {n_terms} t√©rminos: {suma_total}')
````

## Noci√≥n y definici√≥n de l√≠mite üìú

El matem√°tico franc√©s Augustine Louis Cauchy (1789-1857) fue el primero en desarrollar una definici√≥n rigurosa de l√≠mite (aunque ya era usado este concepto desde los antig√ºos griegos para el c√°lculo de √°reas) de la siguiente manera:

*Cuando los valores atribuidos sucesivamente a una variable se aproximan indefinidamente a un valor fijo para llegar por √∫ltimo a diferir de ese valor en una cantidad tan peque√±a como se desee, entonces dicho valor fijo recibe el nombre de l√≠mite de todos los dem√°s valores.*

En palabras m√°s llanas decimos que: el l√≠mite de una funci√≥n $f(x)$ en el punto $x_0$, es obtener el valor al que se va aproximando dicha funci√≥n cuando $x$ tiende a $x_0$, pero sin llegar a ese punto, pues puede ser que $f(x_0)$ no est√© definida en ese punto; durante este proceso nos vamos fijando en los valores que toma $f(x)$ para *puntos extremadamente cercanos* a $x_0$.

La sintaxis matem√°tica del l√≠mite es:

$$\displaystyle \lim_{x \to x_0} f(x) = L $$

donde $L$ es el valor del l√≠mite.

Para aclarar lo anterior veamos un primer ejemplo:

### Ejemplo 1

¬øCu√°l es el l√≠mite de la funci√≥n $f(x) = x^2 + 2x +1$ cuando $x$ tiende a 2?

Checa la siguiente gr√°fica:

````python
#Definimos la funci√≥n f(x)
f = lambda x: x**2 + 2*x + 1

#Definimos el dominio y el codominio de la funci√≥n f(x)
x = np.linspace(-2,4,100) # Dominio
y = f(x)                  # Codiminio

#Definimos espacio de los puntos para l√≠mite izquierdo y derecho
pxi = np.arange(1,2,0.1)
pyi = f(pxi)

pxd = np.arange(3,2,-0.1)
pyd = f(pxd)

#Graficamos especificando el titulo y nombre de los ejes
par = [{'title':'$f(x)=x^2 + 2x + 1$',
        'xlabel':'$x$',
        'ylabel':'$f(x)$'}]

graf = vis.planoCartesiano(par=par)
graf.plot(x=x, y=f(x))       #Gr√°fica de la funci√≥n
graf.scatter(x=pxi, y=pyi,   #Gr√°fica del l√≠mite izquierdo
             par={'label':'L√≠mite izquierdo','color':'black', 'zorder':5})
graf.scatter(x=pxd, y=pyd,   #Gr√°fica del l√≠mite derecho
             par={'label':'L√≠mite derecho','color':'blue', 'zorder':5})
graf.scatter(x=2, y=f(2),    #Gr√°fica del l√≠mite
             par={'label':'f(2) = {}'.format(f(2)),'color':'red', 'zorder':5})
graf.legend()
graf.show()
````

Podemos acercarnos a $x = 2$ por la izquierda (puntos negros) o por la derecha (puntos azules), hasta llegar a puntos muy cercanos a $2$. En este caso, la funci√≥n si est√° definida en $x=2$ y tiene un valor de $f(2)=9$ (punto rojo), por lo que tenemos:

$$\displaystyle \lim_{x \to 2} x^2 + 2x +1 = 9 $$

Lo anterior se puede comprobar sustituyendo $x=2$ en la funci√≥n (esto es posible por que $f(x)$ si est√° definida en $x=2$):

$$ \displaystyle \lim_{x \to 2} x^2 + 2x +1 = (2)^2 + 2(2) + 1 = 9 $$

Tambi√©n se puede usar la funci√≥n `limit` de la biblioteca $sympy$ de $Python$:

````python
#Calculando l√≠mite con Python
#Denotamos x como una variable
x = Symbol('x')
#Definimos la funci√≥n f(x)
y = x*x + 2*x +1
#Calculamos el l√≠mite de f(x)=y cuando x tiende a 2
limit(y, x, 2)
````

Como hemos visto que se pueden calcular l√≠mites por la izquierda y por la derecha. Cuando encontramos que esos l√≠mites, tambi√©n llamados *l√≠mites laterales*, son diferentes, entonces podemos decir que tenemos una discontinuidad.

## Propiedades de los l√≠mites üîç
Algunas propiedades matem√°ticas de los l√≠mites pueden facilitar en algunos casos los c√°lculos en funciones m√°s complejas. Considerando dos funciones definidas en un mismo intervalo.

* **Unicidad del l√≠mite:** El l√≠mite de una funci√≥n ser√° √∫nico en caso de su existencia.
* **L√≠mite de una constante:** El l√≠mite de una funci√≥n constante $f(x) = k$ ser√° igual a la constante $k$.
$$\lim_{x \to a} k = k $$
* **Suma y resta de l√≠mites:** El l√≠mite de la suma ser√° la suma de los l√≠mites.
$$\lim_{x \to a} [f(x) \pm g(x) ] = \lim_{x \to a} f(x) \pm \lim_{x \to a} g(x) $$
* **Producto de l√≠mites:** El l√≠mite del producto es el producto de los l√≠mites
$$\lim_{x \to a} [f(x) \cdot g(x) ] = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x) $$
* **Factor del l√≠mite:** El l√≠mite del producto de una constante por una funci√≥n ser√° la constante por el l√≠mite de la funci√≥n.
$$\lim_{x \to a} [k \cdot f(x) ] = k \cdot \lim_{x \to a} f(x) $$
* **Cociente de l√≠mites:**  El cociente de los l√≠mites de dos funciones ser√° el l√≠mite del cociente de dichas funciones.
$$ \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)} =  \lim_{x \to a} \frac{f(x)}{g(x)} $$
* **Potencia del l√≠mite:** El l√≠mite de una funci√≥n potencial es la potencia del l√≠mite de la base elevado al exponente.
$$ \lim_{x \to a} [f(x)^k]=[\lim_{x \to a} f(x)]^k $$
* **Exponencial de un l√≠mite:** El l√≠mite de una funci√≥n exponencial es la potencia de la base elevada al l√≠mite de la funci√≥n exponente.
$$ \lim_{x \to a} k^{f(x)}=k^{\lim_{x \to a} f(x)} $$
* **Ra√≠z de un l√≠mite**: El l√≠mite de una ra√≠z, es la ra√≠z del l√≠mite.
$$ \lim_{x \to a} \sqrt[n]{f(x)}= \sqrt[n]{\lim_{x \to a} f(x)} $$
* **Logar√≠tmo de un l√≠mite:** El l√≠mite del logaritmo es el logaritmo del l√≠mite.
$$ \lim_{x \to a} \log_k f(x)= \log_k \lim_{x \to a} f(x) $$

### Ejemplo 2

> Agregar bloque entrecomillado



Calcular el l√≠mite de $f(x)$ cuando $x$ tiende a $4$ para $f(x) = \sqrt{2x + 1}$.


Primero definimos $f_0(x) = 2x+1$ y calculamos:

$$ \lim_{x \to 4} f_0(x) = \lim_{x \to 4}  (2x + 1) = 9$$

Ahora podemos usar la siguiente propiedad:

$$ \lim_{x \to a} \sqrt{2x + 1} = \lim_{x \to a} \sqrt{f_0(x)} = \sqrt{\lim_{x \to a} f_0(x)}$$

Entonces el l√≠mite que se busca es:

$$ \sqrt{\lim_{x \to a} f_0(x)} = \sqrt{9} = 3 $$

Verifique lo anterior usando la funci√≥n `limit` como en el ejemplo 1.

Veamos lo anterior gr√°ficamente:

````python
#Definimos la funci√≥n f(x) y f0(x)
f0 = lambda x: 2 * x + 1
f = lambda x: np.sqrt(f0(x))

#Definimos el dominio y el codominio de la funci√≥n f(x)
x = np.linspace(0,8,100)  # Dominio
y0 = f0(x)                # Codominio de f0
yf = f(x)                 # Codominio de f

#Graficamos especificando el titulo y nombre de los ejes
par = [{'title':'L√≠mite de la ra√≠z cuadrada de una funci√≥n',
        'xlabel':'$x$',
        'ylabel':'$y$'}]

graf = vis.planoCartesiano(par=par)
graf.plot(x=x, y=y0, par={'label':'$f0(x)$'})           #Gr√°fica de la funci√≥n f0(x)
graf.plot(x=x, y=yf, par={'label':'$f(x)$'})            #Gr√°fica de la funci√≥n f(x)

x0 = 4

graf.scatter(x=x0, y=f0(x0),        #L√≠mite de f0(x)
             par={'label':'L√≠mite de $f0({})$ = {}'.format(x0,f0(x0)), 's':100, 'marker':'s','color':'C0', 'zorder':5})

graf.scatter(x=x0, y=f(x0),         #L√≠mite de f(x)
             par={'label':'L√≠mite de $f({})$ = {}'.format(x0,f(x0)), 's':100, 'marker':'s','color':'C1', 'zorder':5})

graf.legend()
graf.show()
````

**¬øC√≥mo obtener la continuidad de una funci√≥n?**

Para determinar la continuidad de una funci√≥n en un punto ‚Äúa‚Äù, puedes seguir los siguientes pasos:

Verificar si la funci√≥n est√° definida en el punto de inter√©s: Comprueba si el valor de x en el punto dado est√° dentro del dominio de la funci√≥n. Si la funci√≥n no est√° definida en ese punto, no se puede hablar de continuidad en ese punto.

Evaluar los l√≠mites laterales: Calcula el l√≠mite de la funci√≥n cuando x se aproxima al punto de inter√©s desde la izquierda y desde la derecha. Si ambos l√≠mites laterales existen y son iguales, es decir, si los l√≠mites laterales coinciden, se cumple la primera condici√≥n para la continuidad.

Evaluar el valor de la funci√≥n en el punto: Comprueba si el valor de la funci√≥n en el punto de inter√©s est√° definido. Si no est√° definido, la funci√≥n no es continua en ese punto.

Comparar el l√≠mite con el valor de la funci√≥n: Verifica si el l√≠mite de la funci√≥n cuando x se acerca al punto de inter√©s es igual al valor de la funci√≥n en ese punto. Si ambos valores coinciden, se cumple la tercera condici√≥n para la continuidad.

Si todos estos pasos se cumplen, entonces la funci√≥n es continua en el punto ‚Äúa‚Äù.

Para determinar si la funci√≥n es continua en todo su dominio, debes repetir estos pasos para cada punto dentro del dominio. Como estudiar la continuidad en infinitos puntos es materialmente imposible, para estudiar la continuidad en el dominio de una funci√≥n debemos aprender reglas para las diferentes funciones para saber directamente en qu√© puntos podemos o no encontrar discontinuidades, y estudiar solo estos puntos.

Los m√°ximos y m√≠nimos de una funci√≥n son puntos donde la funci√≥n alcanza sus valores m√°s altos o m√°s bajos, respectivamente, en un intervalo. En c√°lculo, se utilizan las derivadas para encontrar y clasificar estos puntos.

**M√°ximos y M√≠nimos Locales**

Un m√°ximo local de una funci√≥n $f(x)$ es un punto $x=c$ tal que $f(c)$ es mayor que los valores de $f(x)$ en un intervalo cercano a $c$. De manera similar,un m√≠nimo local es un punto $x=c$ tal que $f(c)$ es menor que los valores de $f(x)$ en un intervalo cercano a c.

**M√°ximos y M√≠nimos Globales**

* Un m√°ximo global es el valor m√°s alto que la funci√≥n alcanza en todo su dominio.
* Un m√≠nimo global es el valor m√°s bajo que la funci√≥n alcanza en todo su dominio.


# Pasos a seguir üìã :

 **Criterio de la Primera Derivada**

Para encontrar los m√°ximos y m√≠nimos locales de una funci√≥n, utilizamos la primera derivada $f'(x)$:
1. **Puntos cr√≠ticos**:Los puntos cr√≠ticos son los valores de $x$ donde la derivada primera $f'(x) =0$ o no existe.Para encontrarlos,resolvemos la ecuacion:

$f'(x)=0$  
2. **Clasifica los puntos cr√≠ticos**: Para determinar si un punto cr√≠tico es un m√°ximo, un m√≠nimo o un punto de inflexi√≥n, analizamos el signo de $f'(x)$ alrededor de esos puntos.
 * Si $f'(x)$ cambia de positivo a negativo al pasar por un punto critico $c$,enotnces $f(c)$ es un m√°ximo local.
 * Si $f'(x)$ cambia de negativo a positivo al pasar por un punto critico $c$ entonces $f(c)$ es un m√≠nimo local.

**Criterio de la Segunda Derivada**: ‚úç

Otra forma de clasificar los puntos cr√≠ticos es utilizando la segunda derivada $f''(x)$

1. Calcula la segunda derivada $f''(x)$
2. Evalua la sgunda derivada en los puntos criticos.
 * Si $f''(c)>0$,entonces f(c) es un minimo local (Convexa:la concavidad es hacia arriba).
 * sI $f''(C)<0$, entonces $f(c)$ es un maximo local(c√≥ncava:la concavidad es hacia abajo).
 * Si $f''(c)=0$,el criterio de la segunda derivada no porporciona informacion concluyente,y puede ser necesario usar otros metodos o el criterio de la primera derivada.


````python
# prompt: Imagen del criterio de la segunda derivada

# Import the SymPy library
import sympy as sp

# Define la funci√≥n simb√≥lica
x = sp.Symbol('x')
f = x**3 - 3*x**2 + 2

# Calcula la primera y segunda derivada
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)

# Encuentra los puntos cr√≠ticos
critical_points = sp.solve(f_prime, x)

# Crea el gr√°fico
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.linspace(-2, 3, 200)
y_vals = sp.lambdify(x, f)(x_vals)
y_prime_vals = sp.lambdify(x, f_prime)(x_vals)
y_double_prime_vals = sp.lambdify(x, f_double_prime)(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$f(x)$') # Fixed typo here, closing the label string
plt.plot(x_vals, y_prime_vals, label="$f'(x)$")
plt.plot(x_vals, y_double_prime_vals, label="$f''(x)$")

# Marca los puntos cr√≠ticos
for point in critical_points:
    y_val = sp.lambdify(x, f)(point)
    y_double_prime_val = sp.lambdify(x, f_double_prime)(point)
    if y_double_prime_val > 0:
        plt.plot(point, y_val, 'go', label='M√≠nimo local')
    elif y_double_prime_val < 0:
        plt.plot(point, y_val, 'ro', label='M√°ximo local')
    else:
        plt.plot(point, y_val, 'bo', label='Punto de inflexi√≥n')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Criterio de la Segunda Derivada')
plt.legend()
plt.grid(True)
plt.show()
````

# **Optimizaci√≥n en Econom√≠a** üíπ üí≤

Las derivadas se utilizan para maximizar beneficios o minimizar costos.

**Ejemplo**

Una empresa tiene una funci√≥n de costo $C(x)=x^3-6x^2+15x+100$

Encuentra el nivel de produccion que minimiza el costo.

* Solucionamos Puntos criticos
$f'(x)=3x^2-12x+15$

$3x^2-12x+15=0$

$x=2$

*Evaluamos la segunda derivada para confirmar que es un m√≠nimo:

$C''(x)=6x-12$

$C''(2)=0$ $(Punto de inflexi√≥n)$

```
# Tiene formato de c√≥digo
```

````python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la funci√≥n de costo
def C(x):
    return x**3 - 6*x**2 + 15*x + 100

# Definimos su derivada primera
def C_prime(x):
    return 3*x**2 - 12*x + 15

# Definimos su derivada segunda
def C_double_prime(x):
    return 6*x - 12

# Generamos valores de x para graficar
x = np.linspace(-1, 5, 400)
y = C(x)
y_prime = C_prime(x)
y_double_prime = C_double_prime(x)

# Encontramos los puntos cr√≠ticos
x_criticos = [2]

# Graficamos la funci√≥n y sus derivadas
plt.figure(figsize=(10, 6))

# Gr√°fica de la funci√≥n de costo
plt.plot(x, y, label='$C(x) = x^3 - 6x^2 + 15x + 100$', color='blue')
plt.scatter(x_criticos, [C(xi) for xi in x_criticos], color='red', zorder=5)

# Anotaciones de los puntos cr√≠ticos
for xi in x_criticos:
    plt.annotate(f'({xi}, {C(xi):.2f})', (xi, C(xi)), textcoords="offset points", xytext=(0,10), ha='center', color='red')

# Detalles de la gr√°fica
plt.title('Funci√≥n de Costo y su Derivada')
plt.xlabel('x')
plt.ylabel('Costo')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
````
