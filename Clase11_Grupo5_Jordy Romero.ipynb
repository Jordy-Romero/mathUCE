## **Newton y Leibniz: Síntesis y Formalización** 🧠

1. **Isaac Newton (1643-1727)**
   - En Cambridge, Newton desarrolló su propio método para el cálculo de tangentes. En 1665, descubrió un algoritmo para derivar funciones algebraicas, coincidiendo con el método de Fermat.

   - Newton utilizó sus métodos de fluxiones para desarrollar leyes de movimiento y gravitación, aplicando el cálculo diferencial e integral a problemas físicos.

# **Introducción a las Derivadas** ✏️

## **Definiciones de derivada** 🔍

### **Derivada en un punto a partir de cocientes diferenciales**
La derivada de una función $f$ en el punto $a$ es la pendiente de la recta tangente a la gráfica de $f$ en el punto $a$. El valor de esta pendiente será aproximadamente igual a la pendiente de una recta secante a la gráfica que pase por el punto $(a, f(a))$ y por un punto cercano $(x, f(x))$; por conveniencia suele expresarse $x = a + h$, donde $h$ es un número cercano a 0. A partir de estos dos puntos se calcula la pendiente de la recta secante como:

$$
\frac{f(x) - f(a)}{x - a} = \frac{f(a + h) - f(a)}{h}.
$$

(Esta expresión se denomina «cociente diferencial» o «cociente de Newton».) A medida que el número $h$ se acerca a cero, el valor de esta pendiente se aproximará mejor al de la recta tangente. Esto permite definir la derivada de la función $f$ en el punto $a$, denotada como $f'(a)$, como el límite de estos cocientes cuando $h$ tiende a cero:

$$
f'(a) = \lim_{h \to 0} \frac{f(a + h) - f(a)}{h}.
$$

No obstante, esta definición sólo es válida cuando el límite es un número real: en los puntos $a$ donde el límite no existe, la función $f$ no tiene derivada.

## **Derivada de una función**

Dada una función $f$, se puede definir una nueva función que, en cada punto $x$, toma el valor de la derivada $f'(x)$. Esta función se denota $f'$ y se denomina función derivada de $f$ o simplemente derivada de $f$. Esto es, la derivada de $f$ es la función dada por:

$$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}.
$$

Esta función sólo está definida en los puntos del dominio de $f$ donde el límite existe; en otras palabras, el dominio de $f'$ está contenido en el de $f$.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

# Definir la función cuadrática que queremos derivar
def f(x):
    return x**2 + 2*x + 1  # Función cuadrática: x^2 + 2x + 1

# Crear un rango de valores x
x = np.linspace(-10, 10, 400)  # Genera 400 puntos entre -10 y 10
y = f(x)  # Evaluar la función f(x) en los puntos x

# Calcular la derivada utilizando scipy
def df(x):
    # Calcula la derivada en el punto x usando el método de diferencias finitas
    return derivative(f, x, dx=1e-6)

# Crear un rango de valores x para la derivada
x_deriv = np.linspace(-10, 10, 400)  # Genera 400 puntos entre -10 y 10
y_deriv = np.array([df(xi) for xi in x_deriv])  # Evaluar la derivada en los puntos x_deriv

# Crear una figura con dos subplots
plt.figure(figsize=(12, 6))  # Tamaño de la figura: 12x6 pulgadas

# Graficar la función original
plt.subplot(1, 2, 1)  # Crear el primer subplot en una figura de 1 fila y 2 columnas
plt.plot(x, y, label='f(x) = x^2 + 2x + 1')  # Graficar la función original
plt.title('Función Cuadrática')  # Título del gráfico
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('f(x)')  # Etiqueta del eje y
plt.grid(True)  # Mostrar la cuadrícula
plt.legend()  # Mostrar la leyenda

# Graficar la derivada
plt.subplot(1, 2, 2)  # Crear el segundo subplot en una figura de 1 fila y 2 columnas
plt.plot(x_deriv, y_deriv, label="f'(x) = 2x + 2")  # Graficar la derivada de la función
plt.title('Derivada')  # Título del gráfico
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel("f'(x)")  # Etiqueta del eje y
plt.grid(True)  # Mostrar la cuadrícula
plt.legend()  # Mostrar la leyenda

plt.tight_layout()  # Ajustar el diseño para evitar superposición
plt.show()  # Mostrar la figura
````

# **Notación** 📝

Existen diversas formas para nombrar a la derivada. Siendo $f$ una función, se escribe la derivada de la función $f$ respecto al valor $x$ en varios modos.



## **Notación de Lagrange** 💡

La notación más simple para diferenciación, en uso actual, se debe a Lagrange, y consiste en denotar la derivada de una función $f(x)$ como $f'(x)$: se lee «f prima de x». Esta notación se extiende a derivadas de orden superior, dando lugar a $f''(x)$ («f segunda de x» o «f dos prima de x») para la derivada segunda, y a $f'''(x)$ para la derivada tercera.

La derivada cuarta y siguientes se pueden denotar de dos formas:

- Con números romanos: $f^{\text{IV}}(x)$.
- Con números entre paréntesis: $f^{(4)}(x)$.

Esta última opción da lugar también a la notación $f^{(n)}(x)$ para denotar la derivada n-ésima de $f$.


````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la variable simbólica y la función
x = sp.Symbol('x')
f = sp.exp(-x**2)

# Calcular las derivadas
f_prime = sp.diff(f, x, 1)  # Primera derivada
f_double_prime = sp.diff(f, x, 2)  # Segunda derivada
f_triple_prime = sp.diff(f, x, 3)  # Tercera derivada
f_quadruple_prime = sp.diff(f, x, 4)  # Cuarta derivada

# Convertir las derivadas a funciones numéricas
f_prime_func = sp.lambdify(x, f_prime, 'numpy')
f_double_prime_func = sp.lambdify(x, f_double_prime, 'numpy')
f_triple_prime_func = sp.lambdify(x, f_triple_prime, 'numpy')
f_quadruple_prime_func = sp.lambdify(x, f_quadruple_prime, 'numpy')

# Crear un rango de valores x para graficar
x_vals = np.linspace(-3, 3, 400)
y = np.exp(-x_vals**2)
y_prime = f_prime_func(x_vals)
y_double_prime = f_double_prime_func(x_vals)
y_triple_prime = f_triple_prime_func(x_vals)
y_quadruple_prime = f_quadruple_prime_func(x_vals)

# Crear una figura con cinco subplots
plt.figure(figsize=(15, 12))

# Graficar la función original
plt.subplot(3, 2, 1)
plt.plot(x_vals, y, label='f(x) = e^(-x^2)')
plt.title('Función Original')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Graficar la primera derivada
plt.subplot(3, 2, 2)
plt.plot(x_vals, y_prime, label="f'(x)", color='orange')
plt.title('Primera Derivada')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.grid(True)
plt.legend()

# Graficar la segunda derivada
plt.subplot(3, 2, 3)
plt.plot(x_vals, y_double_prime, label="f''(x)", color='green')
plt.title('Segunda Derivada')
plt.xlabel('x')
plt.ylabel("f''(x)")
plt.grid(True)
plt.legend()

# Graficar la tercera derivada
plt.subplot(3, 2, 4)
plt.plot(x_vals, y_triple_prime, label="f'''(x)", color='red')
plt.title('Tercera Derivada')
plt.xlabel('x')
plt.ylabel("f'''(x)")
plt.grid(True)
plt.legend()

# Graficar la cuarta derivada
plt.subplot(3, 2, 5)
plt.plot(x_vals, y_quadruple_prime, label="f''''(x)", color='purple')
plt.title('Cuarta Derivada')
plt.xlabel('x')
plt.ylabel("f''''(x)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````

## **Notación de Leibniz** 📚

Otra notación común para la diferenciación es debida a Leibniz. Para la función derivada de $f$, se escribe:

$$
\frac{\mathrm{d}\left(f(x)\right)}{\mathrm{d}x}
$$

También puede encontrarse como $ \frac{\mathrm{d}y}{\mathrm{d}x}$, $ \frac{\mathrm{d}f}{\mathrm{d}x}$ o $ \frac{\mathrm{d}}{\mathrm{d}x} f(x)$. Se lee «derivada de $y$ o $f$ con respecto a $x$». Esta notación tiene la ventaja de sugerir la derivada de una función con respecto a otra como un cociente de diferenciales.

Con esta notación, se puede escribir la derivada de $f$ en el punto $a$ de dos modos diferentes:

$$
\frac{\mathrm{d}f}{\mathrm{d}x}\bigg|_{x=a} = \left(\frac{\mathrm{d}\left(f(x)\right)}{\mathrm{d}x}\right)(a)
$$

Si $y = f(x)$, se puede escribir la derivada como

$$
\frac{\mathrm{d}y}{\mathrm{d}x}
$$

Las derivadas sucesivas se expresan como

$$
\frac{\mathrm{d}^n f}{\mathrm{d}x^n} \text{ o } \frac{\mathrm{d}^n y}{\mathrm{d}x^n}
$$


para la enésima derivada de $f$ o de $y$, respectivamente. Históricamente, esto viene del hecho que, por ejemplo, la tercera derivada es

$$
\frac{\mathrm{d} \left(\frac{\mathrm{d} \left( \frac{\mathrm{d} \left(f(x)\right)}{\mathrm{d}x}\right)}{\mathrm{d}x}\right)}{\mathrm{d}x}
$$

la cual se puede escribir como

$$
\left(\frac{\mathrm{d}}{\mathrm{d}x}\right)^3 \left(f(x)\right) = \frac{\mathrm{d}^3}{\left(\mathrm{d}x\right)^3} \left(f(x)\right).
$$

La notación de Leibniz es muy útil, por cuanto permite especificar la variable de diferenciación (en el denominador); lo cual es pertinente en caso de diferenciación parcial. También facilita recordar la regla de la cadena, porque los términos «d» parecen cancelarse simbólicamente:

$$
\frac{\mathrm{d}y}{\mathrm{d}x} = \frac{\mathrm{d}y}{\mathrm{d}u} \cdot \frac{\mathrm{d}u}{\mathrm{d}x}.
$$

En la formulación popular del cálculo mediante límites, los términos «d» no pueden cancelarse literalmente, porque por sí mismos son indefinidos; son definidos solamente cuando se usan juntos para expresar una derivada. En análisis no-estándar, no obstante, se puede ver como números infinitesimales que se cancelan.

Ciertamente, Leibniz consideró la derivada $ \frac{\mathrm{d}y}{\mathrm{d}x} $ como el cociente de dos «infinitésimos» $ \mathrm{d}y $ y $ \mathrm{d}x $, llamados «diferenciales». Estos infinitésimos no eran números sino cantidades más pequeñas que cualquier número positivo.

## **Notación de Newton** 📖

La notación de Newton para la diferenciación respecto al tiempo era poner un punto arriba del nombre de la función:

$$
\dot{x} = x'(t)
$$

$$
\ddot{x} = x''(t)
$$

Se lee «punto $x$ » o « $x$ punto». Actualmente está en desuso en Matemáticas puras, sin embargo, se sigue usando en áreas de la física como la mecánica, donde otras notaciones de la derivada se pueden confundir con la notación de velocidad relativa. Se usa para definir la derivada temporal de una variable.

Esta notación de Newton se usa principalmente en mecánica, normalmente para derivadas de tiempo tales como velocidad y aceleración, y en teoría de ecuaciones diferenciales ordinarias. Usualmente solo se usa para las primeras y segundas derivadas.

## **Regla de la Constante** 🔢
   
   Si tienes una función que es una constante $c$, la derivada de una constante es siempre 0.

   $$
   \frac{d}{dx}(c) = 0
   $$

   **Explicación**: La derivada mide el cambio de una función con respecto a $x$. Como una constante no cambia, su derivada es 0.

## **Regla de la Potencia** 📈
   
   Para una función de la forma $x^n$, donde $n$ es un número real, la derivada es:

   $$
   \frac{d}{dx}(x^n) = nx^{n-1}
   $$

   **Explicación**: Multiplicamos el exponente $n$ por la base $x$ elevada a $n-1$. Esto sigue la regla general para derivar potencias de $x$.

## **Regla del Producto** 📉
   
   Si tienes un producto de dos funciones $u(x)$ y $v(x)$, la derivada es:

   $$
   \frac{d}{dx}[u(x) \cdot v(x)] = u'(x) \cdot v(x) + u(x) \cdot v'(x)
   $$

   **Explicación**: La derivada del producto de dos funciones es la derivada de la primera función multiplicada por la segunda función más la primera función multiplicada por la derivada de la segunda función.

## **Regla del Cociente** ➗
   
   Para una función que es el cociente de dos funciones $u(x)$ y $v(x)$, donde $v(x) \neq 0$, la derivada es:

   $$
   \frac{d}{dx}\left(\frac{u(x)}{v(x)}\right) = \frac{u'(x) \cdot v(x) - u(x) \cdot v'(x)}{[v(x)]^2}
   $$

   **Explicación**: Derivamos el numerador y el denominador por separado, y luego aplicamos la fórmula para combinar los resultados en un solo cociente.

## **Regla de la Cadena** 🔗
   
   Si tienes una función compuesta $g(h(x))$, la derivada es:

   $$
   \frac{d}{dx}[g(h(x))] = g'(h(x)) \cdot h'(x)
   $$

   **Explicación**: Derivamos la función exterior $g$ en el punto $h(x)$ y multiplicamos por la derivada de la función interior $h(x)$.

````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la variable simbólica
x = sp.symbols('x')

# Funciones a derivar
# 1. Regla de la Constante: La derivada de una constante es 0
f1 = 5  # Función constante

# 2. Regla de la Potencia: Derivada de x^n es n*x^(n-1)
f2 = x**3 - 4*x**2 + 2*x  # Función polinómica

# 3. Regla del Producto: Derivada de (u*v) es u'*v + u*v'
f3 = x**2 * sp.sin(x)     # Función producto

# 4. Regla del Cociente: Derivada de (u/v) es (u'*v - u*v') / v^2
f4 = (x**2 + 1) / (x - 1)  # Función cociente

# 5. Regla de la Cadena: Derivada de f(g(x)) es f'(g(x)) * g'(x)
f5 = sp.sin(x**2)         # Función compuesta

# Calcular las derivadas simbólicas
f1_prime = sp.diff(f1, x)
f2_prime = sp.diff(f2, x)
f3_prime = sp.diff(f3, x)
f4_prime = sp.diff(f4, x)
f5_prime = sp.diff(f5, x)

# Convertir las funciones simbólicas a funciones numéricas
f1_l = sp.lambdify(x, f1, 'numpy')
f1_prime_l = sp.lambdify(x, f1_prime, 'numpy')

f2_l = sp.lambdify(x, f2, 'numpy')
f2_prime_l = sp.lambdify(x, f2_prime, 'numpy')

f3_l = sp.lambdify(x, f3, 'numpy')
f3_prime_l = sp.lambdify(x, f3_prime, 'numpy')

f4_l = sp.lambdify(x, f4, 'numpy')
f4_prime_l = sp.lambdify(x, f4_prime, 'numpy')

f5_l = sp.lambdify(x, f5, 'numpy')
f5_prime_l = sp.lambdify(x, f5_prime, 'numpy')

# Crear un rango de valores para x
x_values = np.linspace(-2, 2, 400)  # Rango general para funciones
x_values_nonzero = np.linspace(0.1, 2, 400)  # Para evitar división por cero en la función cociente

# Crear la gráfica
plt.figure(figsize=(14, 12))

# 1. Regla de la Constante
plt.subplot(3, 2, 1)
plt.plot(x_values, f1_l(x_values) * np.ones_like(x_values), label='$f(x) = 5$', color='blue')
plt.plot(x_values, f1_prime_l(x_values) * np.ones_like(x_values), label="$f'(x) = 0$", color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regla de la Constante')
plt.legend()
plt.grid(True)

# 2. Regla de la Potencia
plt.subplot(3, 2, 2)
plt.plot(x_values, f2_l(x_values), label='$f(x) = x^3 - 4x^2 + 2x$', color='blue')
plt.plot(x_values, f2_prime_l(x_values), label="$f'(x) = 3x^2 - 8x + 2$", color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regla de la Potencia')
plt.legend()
plt.grid(True)

# 3. Regla del Producto
plt.subplot(3, 2, 3)
plt.plot(x_values, f3_l(x_values), label='$f(x) = x^2 \sin(x)$', color='blue')
plt.plot(x_values, f3_prime_l(x_values), label="$f'(x) = x^2 \cos(x) + 2x \sin(x)$", color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regla del Producto')
plt.legend()
plt.grid(True)

# 4. Regla del Cociente
plt.subplot(3, 2, 4)
plt.plot(x_values_nonzero, f4_l(x_values_nonzero), label='$f(x) = \\frac{x^2 + 1}{x - 1}$', color='blue')
plt.plot(x_values_nonzero, f4_prime_l(x_values_nonzero), label="$f'(x) = \\frac{2x(x - 1) - (x^2 + 1)}{(x - 1)^2}$", color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regla del Cociente')
plt.legend()
plt.grid(True)

# 5. Regla de la Cadena
plt.subplot(3, 2, 5)
plt.plot(x_values, f5_l(x_values), label='$f(x) = \sin(x^2)$', color='blue')
plt.plot(x_values, f5_prime_l(x_values), label="$f'(x) = 2x \cos(x^2)$", color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regla de la Cadena')
plt.legend()
plt.grid(True)

# Ajustar el diseño y mostrar la gráfica
plt.tight_layout()
plt.show()
````

## **Derivadas de Funciones Trigonométricas** 🌐
   
   Las derivadas básicas de funciones trigonométricas son:

   $$
   \frac{d}{dx} (\sin x) = \cos x
   $$

   $$
   \frac{d}{dx} (\cos x) = -\sin x
   $$

   $$
   \frac{d}{dx} (\tan x) = \sec^2 x
   $$

   $$
   \frac{d}{dx} (\cot x) = -\csc^2 x
   $$

   $$
   \frac{d}{dx} (\sec x) = \sec x \cdot \tan x
   $$

   $$
   \frac{d}{dx} (\csc x) = -\csc x \cdot \cot x
   $$

   **Explicación**: Estas derivadas son derivadas estándar para funciones trigonométricas, obtenidas usando identidades y propiedades trigonométricas.

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores para x
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Funciones trigonométricas
sin_x = np.sin(x)  # Seno de x
cos_x = np.cos(x)  # Coseno de x
tan_x = np.tan(x)  # Tangente de x
cot_x = 1 / tan_x  # Cotangente de x
sec_x = 1 / cos_x  # Secante de x
csc_x = 1 / sin_x  # Cosecante de x

# Derivadas
d_sin_x = cos_x  # Derivada de sin(x)
d_cos_x = -sin_x  # Derivada de cos(x)
d_tan_x = sec_x ** 2  # Derivada de tan(x)
d_cot_x = -csc_x ** 2  # Derivada de cot(x)
d_sec_x = sec_x * tan_x  # Derivada de sec(x)
d_csc_x = -csc_x * cot_x  # Derivada de csc(x)

# Crear el gráfico
fig, axs = plt.subplots(3, 2, figsize=(14, 12))  # Crear una figura con 3 filas y 2 columnas
fig.suptitle('Funciones Trigonométricas y sus Derivadas')  # Título general

# Graficar sin(x) y su derivada
axs[0, 0].plot(x, sin_x, label='sin(x)')
axs[0, 0].plot(x, d_sin_x, label='cos(x)', linestyle='--')  # Derivada en línea discontinua
axs[0, 0].set_title('sin(x) y su derivada')  # Título del gráfico
axs[0, 0].legend()  # Mostrar leyenda

# Graficar cos(x) y su derivada
axs[0, 1].plot(x, cos_x, label='cos(x)')
axs[0, 1].plot(x, d_cos_x, label='-sin(x)', linestyle='--')  # Derivada en línea discontinua
axs[0, 1].set_title('cos(x) y su derivada')  # Título del gráfico
axs[0, 1].legend()  # Mostrar leyenda

# Graficar tan(x) y su derivada
axs[1, 0].plot(x, tan_x, label='tan(x)')
axs[1, 0].plot(x, d_tan_x, label='sec^2(x)', linestyle='--')  # Derivada en línea discontinua
axs[1, 0].set_title('tan(x) y su derivada')  # Título del gráfico
axs[1, 0].set_ylim(-10, 10)  # Limitar el rango y mejorar la visualización
axs[1, 0].legend()  # Mostrar leyenda

# Graficar cot(x) y su derivada
axs[1, 1].plot(x, cot_x, label='cot(x)')
axs[1, 1].plot(x, d_cot_x, label='-csc^2(x)', linestyle='--')  # Derivada en línea discontinua
axs[1, 1].set_title('cot(x) y su derivada')  # Título del gráfico
axs[1, 1].set_ylim(-10, 10)  # Limitar el rango y mejorar la visualización
axs[1, 1].legend()  # Mostrar leyenda

# Graficar sec(x) y su derivada
axs[2, 0].plot(x, sec_x, label='sec(x)')
axs[2, 0].plot(x, d_sec_x, label='sec(x) * tan(x)', linestyle='--')  # Derivada en línea discontinua
axs[2, 0].set_title('sec(x) y su derivada')  # Título del gráfico
axs[2, 0].legend()  # Mostrar leyenda

# Graficar csc(x) y su derivada
axs[2, 1].plot(x, csc_x, label='csc(x)')
axs[2, 1].plot(x, d_csc_x, label='-csc(x) * cot(x)', linestyle='--')  # Derivada en línea discontinua
axs[2, 1].set_title('csc(x) y su derivada')  # Título del gráfico
axs[2, 1].legend()  # Mostrar leyenda

# Ajustar el espacio entre gráficos
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Ajustar el diseño para que no se sobrepongan
plt.show()  # Mostrar el gráfico
````

# **Derivadas de Funciones Exponenciales y Logarítmicas** 🌱
  
   Para funciones exponenciales y logarítmicas:

   $$
   \frac{d}{dx} (e^x) = e^x
   $$

   $$
   \frac{d}{dx} (\ln x) = \frac{1}{x}
   $$

   **Explicación**: La derivada de $e^x$ es la misma función $e^x$, y la derivada del logaritmo natural $\ln x$ es $\frac{1}{x}$.

````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la variable simbólica para la derivada
x = sp.symbols('x')

# Definir las funciones para las derivadas
# Función exponencial
f_exp = sp.exp(x)  # e^x

# Función logarítmica
f_log = sp.ln(x)   # ln(x)

# Calcular las derivadas de las funciones
# La derivada de e^x es e^x
f_exp_prime = sp.diff(f_exp, x)

# La derivada de ln(x) es 1/x
f_log_prime = sp.diff(f_log, x)

# Convertir las funciones simbólicas a funciones numéricas
# Estas funciones se pueden usar para graficar
f_exp_l = sp.lambdify(x, f_exp, 'numpy')
f_exp_prime_l = sp.lambdify(x, f_exp_prime, 'numpy')

f_log_l = sp.lambdify(x, f_log, 'numpy')
f_log_prime_l = sp.lambdify(x, f_log_prime, 'numpy')

# Crear un rango de valores para x
# Para la función exponencial, se usa un rango amplio
x_exp_values = np.linspace(-1, 2, 400)

# Para la función logarítmica, se evita valores negativos y cero
x_log_values = np.linspace(0.1, 2, 400)

# Crear la gráfica
plt.figure(figsize=(14, 6))  # Tamaño de la figura

# Gráfico para la función exponencial y su derivada
plt.subplot(1, 2, 1)
plt.plot(x_exp_values, f_exp_l(x_exp_values), label='$f(x) = e^x$', color='blue')  # Función exponencial
plt.plot(x_exp_values, f_exp_prime_l(x_exp_values), label="$f'(x) = e^x$", color='red', linestyle='--')  # Derivada de la función exponencial
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('y')  # Etiqueta del eje y
plt.title('Función Exponencial y su Derivada')  # Título del gráfico
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar la cuadrícula

# Gráfico para la función logarítmica y su derivada
plt.subplot(1, 2, 2)
plt.plot(x_log_values, f_log_l(x_log_values), label='$f(x) = \ln(x)$', color='blue')  # Función logarítmica
plt.plot(x_log_values, f_log_prime_l(x_log_values), label="$f'(x) = \\frac{1}{x}$", color='red', linestyle='--')  # Derivada de la función logarítmica
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('y')  # Etiqueta del eje y
plt.title('Función Logarítmica y su Derivada')  # Título del gráfico
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar la cuadrícula

# Ajustar el diseño de los subgráficos para que no se solapen
plt.tight_layout()

# Mostrar la gráfica
plt.show()
````

# **Derivadas de Funciones Inversas** 🔄
   
   Las derivadas de las funciones inversas básicas son:

   $$
   \frac{d}{dx} (\arcsin x) = \frac{1}{\sqrt{1 - x^2}}
   $$

   $$
   \frac{d}{dx} (\arccos x) = -\frac{1}{\sqrt{1 - x^2}}
   $$

   $$
   \frac{d}{dx} (\arctan x) = \frac{1}{1 + x^2}
   $$

   **Explicación**: Estas derivadas se derivan usando identidades trigonométricas y las definiciones de las funciones inversas.

````python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la variable simbólica para los cálculos
x = sp.symbols('x')

# Definir las funciones inversas
# Función arco seno (arcsin(x))
f_asin = sp.asin(x)

# Función arco coseno (arccos(x))
f_acos = sp.acos(x)

# Función arco tangente (arctan(x))
f_atan = sp.atan(x)

# Calcular las derivadas de las funciones inversas
# La derivada de arcsin(x) es 1 / sqrt(1 - x^2)
f_asin_prime = sp.diff(f_asin, x)

# La derivada de arccos(x) es -1 / sqrt(1 - x^2)
f_acos_prime = sp.diff(f_acos, x)

# La derivada de arctan(x) es 1 / (1 + x^2)
f_atan_prime = sp.diff(f_atan, x)

# Convertir las funciones simbólicas a funciones numéricas para graficar
# Esto permite evaluar las funciones con valores numéricos
f_asin_l = sp.lambdify(x, f_asin, 'numpy')
f_asin_prime_l = sp.lambdify(x, f_asin_prime, 'numpy')

f_acos_l = sp.lambdify(x, f_acos, 'numpy')
f_acos_prime_l = sp.lambdify(x, f_acos_prime, 'numpy')

f_atan_l = sp.lambdify(x, f_atan, 'numpy')
f_atan_prime_l = sp.lambdify(x, f_atan_prime, 'numpy')

# Crear un rango de valores para x
# Para arcsin y arccos, el dominio es [-1, 1]
x_values = np.linspace(-1, 1, 400)

# Para arctan, se puede usar un rango más amplio para ver mejor el comportamiento
x_atan_values = np.linspace(-10, 10, 400)

# Crear la gráfica
plt.figure(figsize=(14, 12))  # Tamaño de la figura

# Gráfico para la función arco seno y su derivada
plt.subplot(3, 2, 1)
plt.plot(x_values, f_asin_l(x_values), label='$f(x) = \\arcsin(x)$', color='blue')  # Función arco seno
plt.plot(x_values, f_asin_prime_l(x_values), label="$f'(x) = \\frac{1}{\\sqrt{1 - x^2}}$", color='red', linestyle='--')  # Derivada de arco seno
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('y')  # Etiqueta del eje y
plt.title('Función Arco Seno y su Derivada')  # Título del gráfico
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar la cuadrícula

# Gráfico para la función arco coseno y su derivada
plt.subplot(3, 2, 2)
plt.plot(x_values, f_acos_l(x_values), label='$f(x) = \\arccos(x)$', color='blue')  # Función arco coseno
plt.plot(x_values, f_acos_prime_l(x_values), label="$f'(x) = -\\frac{1}{\\sqrt{1 - x^2}}$", color='red', linestyle='--')  # Derivada de arco coseno
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('y')  # Etiqueta del eje y
plt.title('Función Arco Coseno y su Derivada')  # Título del gráfico
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar la cuadrícula

# Gráfico para la función arco tangente y su derivada
plt.subplot(3, 2, 3)
plt.plot(x_atan_values, f_atan_l(x_atan_values), label='$f(x) = \\arctan(x)$', color='blue')  # Función arco tangente
plt.plot(x_atan_values, f_atan_prime_l(x_atan_values), label="$f'(x) = \\frac{1}{1 + x^2}$", color='red', linestyle='--')  # Derivada de arco tangente
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('y')  # Etiqueta del eje y
plt.title('Función Arco Tangente y su Derivada')  # Título del gráfico
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar la cuadrícula

# Ajustar el diseño de los subgráficos para que no se solapen
plt.tight_layout()

# Mostrar la gráfica final
plt.show()
````

# **Ejemplos Explicados** ✅

### **1. Regla de la Constante**

**Ejemplo:**
Sea $f(x) = 5$. Queremos encontrar $\frac{d}{dx}(5)$.

**Proceso:**
La derivada de una constante es siempre cero. Por lo tanto:
$$
\frac{d}{dx}(5) = 0
$$

### **2. Regla de la Potencia**

**Ejemplo:**
Sea $f(x) = x^3$. Queremos encontrar $\frac{d}{dx}(x^3)$.

**Proceso:**
Aplicamos la regla de la potencia con $n = 3$:
$$
\frac{d}{dx}(x^3) = 3 x^{3-1} = 3 x^2
$$

### **3. Regla del Producto**

**Ejemplo:**
Sea $f(x) = (3x) \cdot (x^2)$. Queremos encontrar $\frac{d}{dx}[(3x) \cdot (x^2)]$.

**Proceso:**
Definimos $u(x) = 3x$ y $v(x) = x^2$.

- La derivada de $u(x) = 3x$ es $u'(x) = 3$.
- La derivada de $v(x) = x^2$ es $v'(x) = 2x$.

Aplicamos la regla del producto:
$$
\frac{d}{dx}[(3x) \cdot (x^2)] = 3 \cdot (x^2) + (3x) \cdot (2x) = 3x^2 + 6x^2 = 9x^2
$$

### **4. Regla del Cociente**

**Ejemplo:**
Sea $f(x) = \frac{4x^2}{x^3}$. Queremos encontrar $\frac{d}{dx}\left(\frac{4x^2}{x^3}\right)$.

**Proceso:**
Definimos $u(x) = 4x^2$ y $v(x) = x^3$.

- La derivada de $u(x) = 4x^2$ es $u'(x) = 8x$.
- La derivada de $v(x) = x^3$ es $v'(x) = 3x^2$.

Aplicamos la regla del cociente:
$$
\frac{d}{dx}\left(\frac{4x^2}{x^3}\right) = \frac{(8x) \cdot (x^3) - (4x^2) \cdot (3x^2)}{(x^3)^2} = \frac{8x^4 - 12x^4}{x^6} = \frac{-4x^4}{x^6} = -\frac{4}{x^2}
$$

### **5. Regla de la Cadena**

**Ejemplo:**
Sea $f(x) = \cos(x^2)$. Queremos encontrar $\frac{d}{dx}[\cos(x^2)]$.

**Proceso:**
Definimos $g(u) = \cos(u)$ con $u = h(x) = x^2$.

- La derivada de $g(u) = \cos(u)$ es $g'(u) = -\sin(u)$.
- La derivada de $h(x) = x^2$ es $h'(x) = 2x$.

Aplicamos la regla de la cadena:
$$
\frac{d}{dx}[\cos(x^2)] = -\sin(x^2) \cdot 2x = -2x \sin(x^2)
$$

### **6. Derivadas de Funciones Trigonométricas**

**Ejemplo:**
Para $f(x) = \sin(x)$:

**Proceso:**
Aplicamos la fórmula:
$$
\frac{d}{dx}(\sin(x)) = \cos(x)
$$

### **7. Derivadas de Funciones Exponenciales y Logarítmicas**

**Ejemplo:**
Para $f(x) = e^x$:

**Proceso:**
Aplicamos la fórmula:
$$
\frac{d}{dx}(e^x) = e^x
$$

### **8. Derivadas de Funciones Inversas**

**Ejemplo:**
Para $f(x) = \arctan(x)$:

**Proceso:**
Aplicamos la fórmula:

\frac{d}{dx}(\arctan(x)) = \frac{1}{1 + x^2}

# 🌟 Series

##📚Introducción:
Una serie matemática es la expresión de la suma del número infinito de términos de una sucesión.

La suma hasta un término específico cualquiera se llama suma parcial.

Si las sumas parciales de una serie infinita se aproximan cada vez más a un numero k, de modo que tal que si la serie continúa, la suma puede acercarse a k tanto como se desee; a esto se le conoce como el límite de las sumas parciales o el límite de la serie infinita (suma al infinito, forma abreviada para designar a su límite pues realizar la suma infinita es imposible).


![Texto alternativo](https://www.lavanguardia.com/files/image_948_465/uploads/2018/10/24/5fa450fbd4775.jpeg)

## 📖 CLASIFICACIÓN DE LAS SERIES

### 🔗 Según la cantidad de términos

* Finitas: Una secuencia o serie finita, tiene un primer y último término bien definidos.

* Infinitas: Estas series requieren de herramientas del análisis
matemático para ser debidamente comprendidas y manipuladas.

### 📌 Según el orden de los elementos

* Ascendentes: van de un número menor a uno mayor. (Progresivas).

* Descendentes: van de un número mayor a uno menor (Regresivas).

* Alternadas: los términos se alternan, ya sea que uno crezca y el siguiente decrezca o que uno sea positivo y el siguiente negativo, o ambos cambios a la vez.

## 🔢 Principales tipos de Series

**1. Series Aritméticas:** Una serie aritmética es la suma de los términos de una secuencia aritmética, en la cual la diferencia entre términos consecutivos es constante.

Fórmula de la serie aritmética:
 ![Texto alternativo](https://www.neurochispas.com/wp-content/uploads/2022/11/Formula-de-la-suma-de-una-progresion-aritmetica-1024x692.png)

**2. Series Geométricas:** Una serie geométrica es la suma de los términos de una secuencia geométrica, en la cual la razón entre términos consecutivos es constante.

Fórmula de la serie geométrica:

![Texto alternativo](https://mathcracker.com/images/legacy/geometric-series.png)


````python
# Función para calcular la suma de una serie geométrica
def suma_serie_geometrica(a, r, n):
    """
    Calcula la suma de los primeros n términos de una serie geométrica.

    Parámetros:
    a (float): Primer término de la serie.
    r (float): Razón común de la serie.
    n (int): Número de términos a sumar.

    Retorna:
    float: Suma de los primeros n términos de la serie geométrica.
    """
    if r == 1:
        return a * n
    else:
        return a * (1 - r**n) / (1 - r)

# Ejemplo de uso
a = 2  # Primer término
r = 3  # Razón común
n = 4  # Número de términos

suma = suma_serie_geometrica(a, r, n)
print(f"La suma de los primeros {n} términos de la serie geométrica es: {suma}")
````

````python
# Importamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# Definimos el número de términos
n_terms = 20

# Definimos la serie telescópica
def series_telescopica(n):
    return 1/n - 1/(n+1)

# Calculamos los términos de la serie
terms = [series_telescopica(n) for n in range(1, n_terms+1)]
partial_sums = np.cumsum(terms)

# Visualizamos los términos de la serie
plt.figure(figsize=(10, 6))
plt.plot(range(1, n_terms+1), terms, 'o-', label='Términos de la Serie Telescópica')
plt.xlabel('Número de Término')
plt.ylabel('Valor del Término')
plt.title('Serie Telescópica: Términos')
plt.grid(True)
plt.legend()
plt.show()

# Visualizamos la suma parcial de los términos de la serie
plt.figure(figsize=(10, 6))
plt.plot(range(1, n_terms+1), partial_sums, 'o-', label='Suma Parcial de la Serie Telescópica')
plt.xlabel('Número de Término')
plt.ylabel('Suma Parcial')
plt.title('Serie Telescópica: Suma Parcial')
plt.grid(True)
plt.legend()
plt.show()

# Suma total después de n términos
suma_total = partial_sums[-1]
print(f'Suma total después de {n_terms} términos: {suma_total}')
````

## Noción y definición de límite 📜

El matemático francés Augustine Louis Cauchy (1789-1857) fue el primero en desarrollar una definición rigurosa de límite (aunque ya era usado este concepto desde los antigüos griegos para el cálculo de áreas) de la siguiente manera:

*Cuando los valores atribuidos sucesivamente a una variable se aproximan indefinidamente a un valor fijo para llegar por último a diferir de ese valor en una cantidad tan pequeña como se desee, entonces dicho valor fijo recibe el nombre de límite de todos los demás valores.*

En palabras más llanas decimos que: el límite de una función $f(x)$ en el punto $x_0$, es obtener el valor al que se va aproximando dicha función cuando $x$ tiende a $x_0$, pero sin llegar a ese punto, pues puede ser que $f(x_0)$ no esté definida en ese punto; durante este proceso nos vamos fijando en los valores que toma $f(x)$ para *puntos extremadamente cercanos* a $x_0$.

La sintaxis matemática del límite es:

$$\displaystyle \lim_{x \to x_0} f(x) = L $$

donde $L$ es el valor del límite.

Para aclarar lo anterior veamos un primer ejemplo:

## Propiedades de los límites 🔍
Algunas propiedades matemáticas de los límites pueden facilitar en algunos casos los cálculos en funciones más complejas. Considerando dos funciones definidas en un mismo intervalo.

* **Unicidad del límite:** El límite de una función será único en caso de su existencia.
* **Límite de una constante:** El límite de una función constante $f(x) = k$ será igual a la constante $k$.
$$\lim_{x \to a} k = k $$
* **Suma y resta de límites:** El límite de la suma será la suma de los límites.
$$\lim_{x \to a} [f(x) \pm g(x) ] = \lim_{x \to a} f(x) \pm \lim_{x \to a} g(x) $$
* **Producto de límites:** El límite del producto es el producto de los límites
$$\lim_{x \to a} [f(x) \cdot g(x) ] = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x) $$
* **Factor del límite:** El límite del producto de una constante por una función será la constante por el límite de la función.
$$\lim_{x \to a} [k \cdot f(x) ] = k \cdot \lim_{x \to a} f(x) $$
* **Cociente de límites:**  El cociente de los límites de dos funciones será el límite del cociente de dichas funciones.
$$ \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)} =  \lim_{x \to a} \frac{f(x)}{g(x)} $$
* **Potencia del límite:** El límite de una función potencial es la potencia del límite de la base elevado al exponente.
$$ \lim_{x \to a} [f(x)^k]=[\lim_{x \to a} f(x)]^k $$
* **Exponencial de un límite:** El límite de una función exponencial es la potencia de la base elevada al límite de la función exponente.
$$ \lim_{x \to a} k^{f(x)}=k^{\lim_{x \to a} f(x)} $$
* **Raíz de un límite**: El límite de una raíz, es la raíz del límite.
$$ \lim_{x \to a} \sqrt[n]{f(x)}= \sqrt[n]{\lim_{x \to a} f(x)} $$
* **Logarítmo de un límite:** El límite del logaritmo es el logaritmo del límite.
$$ \lim_{x \to a} \log_k f(x)= \log_k \lim_{x \to a} f(x) $$

**Tipos de Continuidad:**

1. **Continuidad en un punto:**

2. **Continuidad en un intervalo:**
   Una función \( f(x) \) es continua en un intervalo \( [a, b] \) si es continua en cada punto del intervalo \( (a, b) \) y también es continua en los extremos \( a \) y \( b \). Esto implica que no hay saltos bruscos ni agujeros en la gráfica de la función dentro del intervalo dado.

De manera similar, es continua por la derecha si el límite desde la derecha es igual a \( f(a) \).

- **Discontinuidad removible:** Cuando una función está indefinida o tiene un agujero en un punto, pero se puede "rellenar" para hacerla continua ajustando o redefiniendo la función en ese punto.
- **Discontinuidad de salto:** Cuando la función tiene un salto brusco en un punto, es decir, existe un cambio finito en el valor de la función de un lado del punto al otro.
- **Discontinuidad infinita:** Cuando la función tiene un comportamiento asintótico o se dispara a infinito en un punto.

**Discontinuidades no evitables**

Existen otros tipos de discontinuidades que no pueden resolverse, por lo que se llaman discontinuidades no evitables. Estas discontinuidades se clasifican en:

Discontinuidades de salto: cuando existen ambos límites laterales (por la derecha y por la izquierda), pero no coinciden.

En sentido genérico, se llama discontinuidad de segunda especie a la que tiene lugar cuando uno de los límites laterales es finito y el otro es infinito o no existe.

**¿Cuál es la relación entre el dominio y la continuidad de una función?**

El dominio de una función y su continuidad están estrechamente relacionados, pues para que una función sea continua en un punto, ese punto debe pertenecer al dominio de la función. Si un punto está excluido del dominio de la función, no se puede evaluar y, por lo tanto, no tiene sentido hablar de su continuidad en ese punto.

**¿Cómo obtener la continuidad de una función?**

Para determinar la continuidad de una función en un punto “a”, puedes seguir los siguientes pasos:

Verificar si la función está definida en el punto de interés: Comprueba si el valor de x en el punto dado está dentro del dominio de la función. Si la función no está definida en ese punto, no se puede hablar de continuidad en ese punto.

Evaluar los límites laterales: Calcula el límite de la función cuando x se aproxima al punto de interés desde la izquierda y desde la derecha. Si ambos límites laterales existen y son iguales, es decir, si los límites laterales coinciden, se cumple la primera condición para la continuidad.

Evaluar el valor de la función en el punto: Comprueba si el valor de la función en el punto de interés está definido. Si no está definido, la función no es continua en ese punto.

Comparar el límite con el valor de la función: Verifica si el límite de la función cuando x se acerca al punto de interés es igual al valor de la función en ese punto. Si ambos valores coinciden, se cumple la tercera condición para la continuidad.

Si todos estos pasos se cumplen, entonces la función es continua en el punto “a”.

Para determinar si la función es continua en todo su dominio, debes repetir estos pasos para cada punto dentro del dominio. Como estudiar la continuidad en infinitos puntos es materialmente imposible, para estudiar la continuidad en el dominio de una función debemos aprender reglas para las diferentes funciones para saber directamente en qué puntos podemos o no encontrar discontinuidades, y estudiar solo estos puntos.

**¿Se puede calcular la continuidad sin hacer los limites laterales?**

No, para determinar la continuidad de una función en un punto específico x=a, es necesario calcular los límites laterales.

**¿Qué tipos de discontinuidades hay?**

Cada tipo de discontinuidad implica un comportamiento particular en un punto o en un conjunto de puntos de una función, siendo los principales tipos los siguientes:

Discontinuidad evitable o punto removible: Ocurre cuando la función tiene un agujero o un hueco en un punto específico, pero se puede “rellenar” o corregir asignando un valor adecuado en ese punto. Matemáticamente, esto ocurre cuando los límites laterales coinciden, pero el valor de la función en el punto no.

Discontinuidad inevitable de salto finito: Ocurre cuando la función tiene un salto abrupto en un punto. En este tipo de discontinuidad, los límites laterales existen en el punto, pero no son iguales, provocando una interrupción o salto en la gráfica de la función.

Discontinuidad inevitable de salto infinito: Ocurre cuando la función tiene al menos un límite lateral infinito o menos infinito en un punto. Esto genera una discontinuidad vertical en la gráfica de la función y una posterior asíntota vertical.

Es importante tener en cuenta que, en la práctica, las funciones pueden tener combinaciones de estos tipos de discontinuidades y pueden presentar comportamientos más complicados.

Propiedades de las funciones continuas
Las operaciones algebraicas (suma, resta, multiplicación) de funciones continuas también resultan en funciones continuas. Por ejemplo, cada término individual en un polinomio es una función continua.

La multiplicación de una función continua por una constante también resulta en una función continua, pues la constante puede verse como una función continua en sí misma.

La composición de funciones continuas también es continua.

## **DERIVADAS**


La derivada de una función \( f(x) \) en un punto \( x = a \) representa la tasa de cambio instantánea de la función en ese punto. Geométricamente, la derivada es la pendiente de la recta tangente a la curva de la función en ese punto.

Formalmente, la derivada de \( f(x) \) en \( x = a \), denotada como \( f'(a) \) o \( \frac{df}{dx}(a) \), se define como el límite:
\[ f'(a) = \lim_{h \to 0} \frac{f(a + h) - f(a)}{h} \]
siempre que este límite exista.

La derivada nos proporciona información sobre cómo cambia una función en un punto específico y se utiliza ampliamente en la física, la economía, la ingeniería y otras disciplinas para modelar el cambio y la velocidad de cambio.

En resumen, la continuidad asegura que no hay interrupciones abruptas en una función, mientras que la derivada proporciona información sobre la tasa de cambio de la función en un punto dado. Ambos conceptos son esenciales para comprender y analizar el comportamiento de las funciones en matemáticas avanzadas y aplicadas.

**Conceptos relacionados con las derivadas**


Cálculo diferencial: estudia cómo cambian las funciones continuas según sus variables cambian de estado. La derivada es el principal objeto de estudio en el cálculo diferencial.

Variables independientes: representan insumos o causas, es decir, razones potenciales de variación. En un experimento, cualquier variable que el experimentador manipule puede denominarse variable independiente.

Límite: formaliza la noción intuitiva de aproximación hacia un punto concreto de una sucesión o una función, a medida que los parámetros de esa sucesión o función se acercan a un determinado valor.

Recta tangente: la tangente​ a una curva en un punto P es una recta que toca a la curva solo en dicho punto, llamado punto de tangencia.

Recta secante: es una recta que corta a una curva en 2 puntos. Conforme estos puntos se acercan y su distancia se reduce a cero, la recta adquiere el nombre de recta tangente.
}
Cálculo infinitesimal (o simplemente cálculo): es el estudio del cambio y la continuidad, en la misma manera que la geometría estudia el espacio


**Aplicaciones de las derivadas**


Las derivadas tienen aplicaciones muy variadas en una gran cantidad de áreas científicas, ya que es una herramienta de cálculo fundamental en los estudios de Física, Química y Biología, o en ciencias sociales como la Economía y la Sociología

Se aplican en los casos donde hay que medir la rapidez con la que se produce el cambio de una magnitud o situación.

Por ejemplo, cuando se refiere a la gráfica de dos dimensiones de f, se considera la derivada como la pendiente de la recta tangente del gráfico en el punto x. Se puede aproximar la pendiente de esta tangente como el límite cuando la distancia entre los dos puntos que determinan una recta secante tiende a cero, es decir, la recta secante se convierte en una recta tangente.

Con esta interpretación, se pueden identificar muchas propiedades geométricas de los gráficos de funciones, como por ejemplo la monotonía de una función (si es creciente o decreciente) y la concavidad o convexidad.

**Máximos y Mínimos**

Extrema_example_es.svg.png
1. En funciones

Los máximos y mínimos de una función son puntos donde la función alcanza sus valores más altos o más bajos, respectivamente, en un intervalo. En cálculo, se utilizan las derivadas para encontrar y clasificar estos puntos.

**Máximos y Mínimos Locales**

Un máximo local de una función $f(x)$ es un punto $x=c$ tal que $f(c)$ es mayor que los valores de $f(x)$ en un intervalo cercano a $c$. De manera similar,un mínimo local es un punto $x=c$ tal que $f(c)$ es menor que los valores de $f(x)$ en un intervalo cercano a c.

**Máximos y Mínimos Globales**

* Un máximo global es el valor más alto que la función alcanza en todo su dominio.
* Un mínimo global es el valor más bajo que la función alcanza en todo su dominio.


# Pasos a seguir 📋 :

 **Criterio de la Primera Derivada**

Para encontrar los máximos y mínimos locales de una función, utilizamos la primera derivada $f'(x)$:
1. **Puntos críticos**:Los puntos críticos son los valores de $x$ donde la derivada primera $f'(x) =0$ o no existe.Para encontrarlos,resolvemos la ecuacion:

$f'(x)=0$  
2. **Clasifica los puntos críticos**: Para determinar si un punto crítico es un máximo, un mínimo o un punto de inflexión, analizamos el signo de $f'(x)$ alrededor de esos puntos.
 * Si $f'(x)$ cambia de positivo a negativo al pasar por un punto critico $c$,enotnces $f(c)$ es un máximo local.
 * Si $f'(x)$ cambia de negativo a positivo al pasar por un punto critico $c$ entonces $f(c)$ es un mínimo local.

**Criterio de la Segunda Derivada**: ✍

Otra forma de clasificar los puntos críticos es utilizando la segunda derivada $f''(x)$

1. Calcula la segunda derivada $f''(x)$
2. Evalua la sgunda derivada en los puntos criticos.
 * Si $f''(c)>0$,entonces f(c) es un minimo local (Convexa:la concavidad es hacia arriba).
 * sI $f''(C)<0$, entonces $f(c)$ es un maximo local(cóncava:la concavidad es hacia abajo).
 * Si $f''(c)=0$,el criterio de la segunda derivada no porporciona informacion concluyente,y puede ser necesario usar otros metodos o el criterio de la primera derivada.


````python
# prompt: Imagen del criterio de la segunda derivada

# Import the SymPy library
import sympy as sp

# Define la función simbólica
x = sp.Symbol('x')
f = x**3 - 3*x**2 + 2

# Calcula la primera y segunda derivada
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)

# Encuentra los puntos críticos
critical_points = sp.solve(f_prime, x)

# Crea el gráfico
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.linspace(-2, 3, 200)
y_vals = sp.lambdify(x, f)(x_vals)
y_prime_vals = sp.lambdify(x, f_prime)(x_vals)
y_double_prime_vals = sp.lambdify(x, f_double_prime)(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$f(x)$') # Fixed typo here, closing the label string
plt.plot(x_vals, y_prime_vals, label="$f'(x)$")
plt.plot(x_vals, y_double_prime_vals, label="$f''(x)$")

# Marca los puntos críticos
for point in critical_points:
    y_val = sp.lambdify(x, f)(point)
    y_double_prime_val = sp.lambdify(x, f_double_prime)(point)
    if y_double_prime_val > 0:
        plt.plot(point, y_val, 'go', label='Mínimo local')
    elif y_double_prime_val < 0:
        plt.plot(point, y_val, 'ro', label='Máximo local')
    else:
        plt.plot(point, y_val, 'bo', label='Punto de inflexión')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Criterio de la Segunda Derivada')
plt.legend()
plt.grid(True)
plt.show()
````

$3x^2-12x+15=0$

````python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función de costo
def C(x):
    return x**3 - 6*x**2 + 15*x + 100

# Definimos su derivada primera
def C_prime(x):
    return 3*x**2 - 12*x + 15

# Definimos su derivada segunda
def C_double_prime(x):
    return 6*x - 12

# Generamos valores de x para graficar
x = np.linspace(-1, 5, 400)
y = C(x)
y_prime = C_prime(x)
y_double_prime = C_double_prime(x)

# Encontramos los puntos críticos
x_criticos = [2]

# Graficamos la función y sus derivadas
plt.figure(figsize=(10, 6))

# Gráfica de la función de costo
plt.plot(x, y, label='$C(x) = x^3 - 6x^2 + 15x + 100$', color='blue')
plt.scatter(x_criticos, [C(xi) for xi in x_criticos], color='red', zorder=5)

# Anotaciones de los puntos críticos
for xi in x_criticos:
    plt.annotate(f'({xi}, {C(xi):.2f})', (xi, C(xi)), textcoords="offset points", xytext=(0,10), ha='center', color='red')

# Detalles de la gráfica
plt.title('Función de Costo y su Derivada')
plt.xlabel('x')
plt.ylabel('Costo')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
````
