## 📚 ¿Quién inventó las derivadas?:

Las derivadas, en el contexto de cálculo matemático, no fueron inventadas por una sola persona, sino que se desarrollaron a lo largo de varios siglos con contribuciones de varios matemáticos importantes. # **Definición de Espacio Métrico** 📖

Un **espacio métrico** es un conjunto $M$ junto con una función $d: M \times M \rightarrow \mathbb{R}$ que define una distancia entre cada par de puntos en $M$. La función $d$ se llama **métrica** y debe satisfacer las siguientes propiedades para todos los $x, y, z \in M$:

1. **No negatividad**: $d(x, y) \geq 0$ y $d(x, y) = 0$ si y solo si $x = y$.
2. **Simetría**: $d(x, y) = d(y, x)$.
3. **Desigualdad triangular**: $d(x, z) \leq d(x, y) + d(y, z)$.

Matemáticamente, un espacio métrico se denota como $(M, d)$.

## **Ejemplos de Espacios Métricos**

1. **Espacio Euclidiano**: El conjunto $\mathbb{R}^n$ con la distancia euclidiana definida por:

   $$ d(x, y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2} $$

2. **Espacio Discreto**: El conjunto $M$ con la distancia $d$ definida por:

   $$ d(x, y) = \begin{cases}
      0 & \text{si } x = y \\
      1 & \text{si } x \neq y
   \end{cases} $$


# **Distancia y Propiedades de la Métrica**📏


La **distancia** en un espacio métrico $(M, d)$ es una función $d$ que cumple con las propiedades mencionadas anteriormente.

## **Propiedades de la Métrica**

1. **No negatividad 🚫➖**:

   $$ d(x, y) \geq 0 $$

   Esta propiedad asegura que la distancia entre dos puntos nunca es negativa.
   
2. **Identidad del Indiscernible 🧮**:

   $$ d(x, y) = 0 \iff x = y $$

   Esto implica que la distancia entre un punto y sí mismo es cero y, si la distancia entre dos puntos es cero, entonces los puntos son idénticos.
   
3. **Simetría**:

   $$ d(x, y) = d(y, x) $$

   La distancia desde $x$ a $y$ es la misma que desde $y$ a $x$.
   
4. **Desigualdad Triangular 📐**:

   $$ d(x, z) \leq d(x, y) + d(y, z) $$
   
   Esta propiedad establece que la distancia directa entre dos puntos $x$ y $z$ es menor o igual que la suma de las distancias desde $x$ a $y$ y de $y$ a $z$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos puntos para demostrar las propiedades
x = np.array([0, 0])
y = np.array([1, 1])
z = np.array([2, 0])

# Función de distancia euclidiana
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))

# Propiedades de la métrica
d_xy = euclidean_distance(x, y)
d_xz = euclidean_distance(x, z)
d_yz = euclidean_distance(y, z)

# Graficar los puntos
plt.figure(figsize=(10, 10))

# Graficar puntos y sus etiquetas
plt.scatter(*x, color='red')
plt.text(x[0], x[1], 'x', fontsize=12, ha='right')

plt.scatter(*y, color='blue')
plt.text(y[0], y[1], 'y', fontsize=12, ha='right')

plt.scatter(*z, color='green')
plt.text(z[0], z[1], 'z', fontsize=12, ha='right')

# Graficar las distancias
plt.plot([x[0], y[0]], [x[1], y[1]], 'k--', label=f'd(x, y) = {d_xy:.2f}')
plt.plot([x[0], z[0]], [x[1], z[1]], 'k--', label=f'd(x, z) = {d_xz:.2f}')
plt.plot([y[0], z[0]], [y[1], z[1]], 'k--', label=f'd(y, z) = {d_yz:.2f}')

# Añadir títulos y leyendas
plt.title('Propiedades de la Métrica')
plt.legend()
plt.grid(True)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Mostrar gráficos de propiedades
plt.show()

# Verificar las propiedades
print("Propiedades de la Métrica:")
print(f"1. No negatividad: d(x, y) = {d_xy} >= 0")
print(f"2. Identidad del Indiscernible: d(x, x) = {euclidean_distance(x, x)} == 0")
print(f"3. Simetría: d(x, y) = {d_xy} == d(y, x) = {euclidean_distance(y, x)}")
print(f"4. Desigualdad Triangular: d(x, z) = {d_xz} <= d(x, y) + d(y, z) = {d_xy + d_yz}")
```

# **Bolas Abiertas y Cerradas**

En un espacio métrico $(M, d)$, las bolas abiertas y cerradas son conjuntos que agrupan todos los puntos dentro de una cierta distancia de un punto central.

## **Bola Abierta**🌠
Sean $ \mathbf{a} = (a_1, a_2, \ldots, a_n) \in \mathbb{R}^n$ y $r > 0$.
Llamamos bola abierta de centro $\mathbf{a}$ y radio $r$ al conjunto:

Donde $d(x, y)$ es la distancia euclidiana en $\mathbb{R}^n:$

- Si $n = 1$,
$$d(x, a) = |x - a|$$

- Si $n = 2$,
$$d(x, a) = \sqrt{(x_1 - a_1)^2 + (x_2 - a_2)^2}$$

- Si $n = 3$,
$$d(x, a) = \sqrt{(x_1 - a_1)^2 + (x_2 - a_2)^2 + (x_3 - a_3)^2}$$


**Representación:**

En la recta real $n = 1$, la bola abierta $B_r(\mathbf{a})$ es el intervalo abierto $]a - r, a + r[:$

![Texto alternativo](https://www.matesfacil.com/topologia/abiertos/bola1.png)

En el plano real ($n = 2$, la bola abierta $B_r(a)$ de centro $a = (a_1, a_2)$ es el círculo sin borde de centro $a$ y radio $r$:

$$B_r(a) = \{(x, y) \in \mathbb{R}^2 : (x - a_1)^2 + (y - a_2)^2 < r^2 \}$$

![Texto alternativo](https://www.matesfacil.com/topologia/abiertos/bola2.png)

En el espacio real $n = 3$, la bola abierta $B_r(a)$ de centro $a = (a_1, a_2, a_3)$ es el interior de la esfera de centro $a$ y radio $r$:


$$B_r(a) = \{(x_1, x_2, x_3) \in \mathbb{R}^3 : \sqrt{\sum_{1 \leq i \leq 3} (x_i - a_i)^2} < r^2 \}$$

![Texto alternativo](https://www.matesfacil.com/topologia/abiertos/bola3.png)







## **Bola Cerrada**🔵

Sean $a = (a_1, a_2, \ldots, a_n) \in \mathbb{R}^n$ y $r > 0$.
Llamamos bola cerrada de centro $a$ y radio $r$ al conjunto

$$\overline{B}_r(a) := \{ x \in \mathbb{R}^n \mid d(x, a) \leq r \}$$

#### ***La diferencia entre la bola abierta y la cerrada es que la cerrada incluye el borde y la abierta no.***

Por ejemplo, una bola cerrada en el plano es

$$\overline{B}_r(a) = \{ (x, y) \in \mathbb{R}^2 \ |\ (x - a_1)^2 + (y - a_2)^2 \leq r^2 \}$$

![Texto alternativo](https://www.matesfacil.com/topologia/abiertos/cerrada2.png)

Es fácil ver que $B_r(a) \subset \overline{B}_r(a)$


# **Conjunto Cerrado**◼️
Un conjunto $ A \subseteq \mathbb{R}^n $ es **cerrado** si su complementario $ B = \mathbb{R}^n \setminus A $ es un conjunto abierto.

$$ B = \mathbb{R}^n \setminus A := \{ x \in \mathbb{R}^n \mid x \notin A \} $$

- **Un conjunto puede ser abierto y cerrado.**

  En la topología usual, los únicos conjuntos que son abiertos y cerrados son el conjunto vacío y el conjunto total $\mathbb{R}^n$.

  En general, en los espacios conexos, los únicos conjuntos abiertos y cerrados son el vacío y el espacio total.

- **La unión finita de cerrados es un cerrado.**

  Sean $A_1$ y $A_2$ dos conjuntos cerrados y sea $C$ su unión: $C = A_1 \cup A_2$.

  Los complementarios de $A_1$ y $A_2$ son

$$X_1 = \mathbb{R}^n \setminus A_1$$
$$X_2 = \mathbb{R}^n \setminus A_2$$

  Son conjuntos abiertos porque $A_1$ y $A_2$ son cerrados.

  Por tanto, su intersección $Z = X_1 \cap X_2$ también es un abierto (intersección finita de abiertos).

  Por las **leyes de De Morgan**, $Z$ es el complementario de $C$ (la intersección de los complementarios es el complementario de la unión).

  Como $Z$ es el complementario de $C$ y es abierto, entonces $C$ es cerrado.

- **La intersección de cerrados es un cerrado.**

  Caso de intersección finita:

Sean $A$ y $B$ dos conjuntos cerrados. Para probar que su intersección es cerrada, hay que demostrar que su complementario es abierto.

Por De Morgan, el complementario de la intersección de $A$ y $B$ es la unión de los complementarios de $A$ y de $B$. Como los complementarios de $A$ y de $B$ son abiertos, su unión también lo es.


# **Resumen de las propiedades**

**1. Contienen todos sus puntos límite:** Un conjunto cerrado en un espacio métrico contiene todos los puntos límite de cualquier secuencia convergente en él. 🎯

**2. Son complementos de conjuntos abiertos:** Un conjunto cerrado es el complemento de un conjunto abierto. 🔒

**3. Son cerrados bajo operaciones topológicas:** La intersección y unión de conjuntos cerrados sigue siendo un conjunto cerrado. 🔄

**4. Son estables bajo límites:** El límite de una secuencia de conjuntos cerrados es también un conjunto cerrado. 📈

### **Definición de Bola Cerrada**💡

La bola cerrada centrada en un punto $o \in E$ con radio $r$ se define como:

# **Uso del símbolo épsilon ($\epsilon$) en Matemáticas**🤔

El símbolo épsilon ($\epsilon$) tiene varios significados en matemáticas, dependiendo del contexto.

## **1. Análisis Matemático y Cálculo**🤓
En análisis matemático y cálculo, $\epsilon$ (épsilon) se utiliza comúnmente en el contexto de los límites y la definición de la continuidad de funciones.

### **Límites**🔢
La definición formal del límite utiliza $\epsilon$ para describir la precisión con la que una función se aproxima a un valor en particular.

Por ejemplo, decimos que el límite de $f(x)$ cuando $x$ tiende a $a$ es $L$ si, para cada $\epsilon > 0$, existe un $\delta > 0$ tal que:

$$
0 < |x - a| < \delta \implies |f(x) - L| < \epsilon
$$

Esta definición formaliza la idea de que $f(x)$ puede hacerse tan cercano a $L$ como se desee (dentro de $\epsilon$) al elegir $x$ suficientemente cercano a $a$ (dentro de $\delta$).

### **Continuidad**📈
Una función $f$ es continua en un punto $a$ si, para cada $\epsilon > 0$, existe un $\delta > 0$ tal que:

$$
|x - a| < \delta \implies |f(x) - f(a)| < \epsilon
$$

Esto significa que podemos hacer que $f(x)$ esté tan cerca de $f(a)$ como queramos (dentro de $\epsilon$) al elegir $x$ suficientemente cerca de $a$ (dentro de $\delta$).

## **2. Teoría de Números** 1️⃣, 2️⃣, 3️⃣, 4️⃣, 5️⃣, 6️⃣, 7️⃣, 8️⃣, 9️⃣, 🔟
En teoría de números, $\epsilon$ se utiliza a veces para denotar un número arbitrariamente pequeño.

Por ejemplo, al describir la densidad de números primos, podríamos decir que un cierto resultado es verdadero "para todo $\epsilon > 0$".

## **3. Notación Asintótica** 🧮
En análisis asintótico, $\epsilon$ puede representar una pequeña constante positiva en notaciones del tipo $O$, $\Omega$, y $\Theta$. Por ejemplo, podríamos decir que una función $f(n)$ es $O(n^{1+\epsilon})$ para algún $\epsilon > 0$.

## **Ejemplos con $\epsilon$**

### **Ejemplo de Límite**
Supongamos que queremos demostrar que:

$$
\lim_{x \to 2} (3x - 4) = 2
$$

Para probar esto, debemos mostrar que para cada $\epsilon > 0$, existe un $\delta > 0$ tal que si $0 < |x - 2| < \delta$, entonces $|3x - 4 - 2| < \epsilon$.

1. Comenzamos con $|3x - 4 - 2| < \epsilon$:

$$
|3x - 6| < \epsilon
$$

2. Factorizamos $3$:

$$
3|x - 2| < \epsilon
$$

3. Dividimos ambos lados por $3$:

$$
|x - 2| < \frac{\epsilon}{3}
$$

Esto muestra que podemos tomar $\delta = \frac{\epsilon}{3}$. Entonces, siempre que $|x - 2| < \delta$, tenemos $|3x - 4 - 2| < \epsilon$.

### **Ejemplo de Continuidad**
Sea $f(x) = x^2$ y queremos demostrar que $f$ es continua en $a = 1$.

1. Debemos mostrar que para cada $\epsilon > 0$, existe un $\delta > 0$ tal que si $|x - 1| < \delta$, entonces $|x^2 - 1| < \epsilon$.

2. Notamos que:

$$
|x^2 - 1| = |x - 1||x + 1|
$$

3. Si elegimos $\delta \leq 1$, entonces $|x - 1| < \delta \leq 1$ y así $|x + 1| \leq 3$.

4. Entonces, podemos elegir $\delta = \min(1, \frac{\epsilon}{3})$:

$$
|x - 1| < \delta \implies |x - 1||x + 1| < \frac{\epsilon}{3} \cdot 3 = \epsilon
$$

Esto muestra que $f(x) = x^2$ es continua en $a = 1$.

Estos son algunos de los usos más comunes y significativos del símbolo $\epsilon$ en matemáticas.

# **Demostración de que $B(o, r)$ es cerrada en un espacio métrico**

Para demostrar que la bola $B(o, r)$ es cerrada en un espacio métrico $(E, d)$:

1. **Definición de una bola abierta**:
   
   En un espacio métrico $(E, d)$, una bola abierta centrada en $o$ con radio $r$ se define como $B(o, r) = \{ x \in E : d(x, o) < r \}$.

2. **Objetivo de la demostración**:
   
   Queremos demostrar que $B(o, r)$ es cerrada, es decir, su complemento $B(o, r)^c$ es abierto.

3. **Complemento de la bola**:
   
   El complemento de la bola $B(o, r)^c$ está dado por $B(o, r)^c = \{ x \in E : d(x, o) \geq r \}$.

4. **Demostración de que $B(o, r)^c$ es abierto**:
   
   Para demostrar que $B(o, r)^c$ es abierto, tomamos un punto $x \in B(o, r)^c$, lo cual implica que $d(x, o) \geq r$.

5. **Definición de $\varepsilon$**:
   
   Definimos $\varepsilon = d(x, o) - r$. Aquí, $\varepsilon > 0$ porque $d(x, o) > r$.

6. **Construcción de la bola abierta alrededor de $x$**:
   
   Consideramos la bola abierta $B(x, \varepsilon) = \{ y \in E : d(y, x) < \varepsilon \}$.

7. **Verificación de puntos en $B(x, \varepsilon)$**:
   
   Sea $y \in B(x, \varepsilon)$, entonces $d(y, x) < \varepsilon$.

8. **Relación métrica con $o$**:
   
   Ahora evaluamos $d(o, y)$:
   $$d(o, y) \leq d(o, x) + d(x, y) < d(o, x) + \varepsilon$$

9. **Sustitución de $\varepsilon$**:
   
   Sustituyendo $\varepsilon = d(x, o) - r$:
   $$d(o, y) < d(o, x) + d(x, o) - r$$

10. **Conclusión**:
   
    Esto muestra que $d(o, y) < d(o, x) + d(x, o) - r$, lo cual demuestra que $y \in B(o, r)$. Por lo tanto, cualquier punto $y$ en $B(x, \varepsilon)$ también está en $B(o, r)$, lo que indica que $B(x, \varepsilon) \subseteq B(o, r)$.

Esto completa la demostración de que $B(o, r)^c$ es abierto, lo cual implica que $B(o, r)$ es cerrada en el espacio métrico $(E, d)$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la bola
o = np.array([0, 0])  # Centro de la bola
r = 1.5  # Radio de la bola

# Crear una figura y un conjunto de ejes
fig, ax = plt.subplots()

# Dibujar la bola abierta B(o, r)
circle = plt.Circle(o, r, color='blue', alpha=0.5, label='$B(o, r)$')
ax.add_artist(circle)

# Dibujar el punto x fuera de la bola
x = np.array([2.5, 0])
plt.plot(x[0], x[1], 'ro', label='$x \in B(o, r)^c$')

# Dibujar la bola abierta alrededor de x con radio epsilon
epsilon = np.linalg.norm(x - o) - r
circle_epsilon = plt.Circle(x, epsilon, color='red', alpha=0.5, linestyle='--', label='$B(x, \epsilon)$')
ax.add_artist(circle_epsilon)

# Configurar los límites del gráfico
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)

# Dibujar el centro de la bola o
plt.plot(o[0], o[1], 'go', label='$o$')

# Añadir etiquetas y leyenda
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Demostración de que $B(o, r)$ es cerrada')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Mostrar el gráfico
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

# **Teoría de Conjuntos y Topología: Abierto y Cerrado en $\mathbb{R}$** 🔴

## **Propiedades de $\mathbb{R}$ y el Conjunto Vacío**

**$\mathbb{R}$ es abierto.**

El conjunto de los números reales, denotado como $\mathbb{R}$, se considera abierto en la topología estándar de los números reales. Esto significa que para cualquier punto $x \in \mathbb{R}$, existe un intervalo alrededor de $x$ que está completamente contenido dentro de $\mathbb{R}$. Por ejemplo, el intervalo $(x - \epsilon, x + \epsilon)$ para algún $\epsilon > 0$ estará completamente dentro de $\mathbb{R}$.

**$\mathbb{R}^c:=\emptyset$, como $\mathbb{R}$ es abierto, su complemento es cerrado, es decir, $\emptyset$ es cerrado.**

En topología, si un conjunto es abierto, su complemento en el espacio topológico se considera cerrado. Dado que $\mathbb{R}$ es abierto, su complemento, que es $\emptyset$, es automáticamente cerrado.

**$\emptyset$ es abierto,**

El conjunto vacío $\emptyset$ se considera abierto en cualquier espacio topológico. Esto se debe a que no contiene ningún punto, y por lo tanto, no hay puntos de límite que puedan estar fuera de $\emptyset$.

**$\emptyset^C := \mathbb{R}$, entonces $\mathbb{R}$ es cerrado.**

Aquí, $\emptyset^C$ denota el complemento del conjunto vacío $\emptyset$. Como $\emptyset^C$ es igual a $\mathbb{R}$, esto implica que $\mathbb{R}$ es el conjunto más grande en el espacio topológico que contiene a todos los números reales. Por definición en topología, el complemento de un conjunto abierto (como $\emptyset$) es cerrado, lo que implica que $\mathbb{R}$ es cerrado.


**$\mathbb{R}$ y $\emptyset$ son simultánemente abiertos y cerrados.**

En algunos contextos topológicos, existen conjuntos que son tanto abiertos como cerrados. En la topología estándar de $\mathbb{R}$ (la topología usual), $\mathbb{R}$ y $\emptyset$ son ejemplos de conjuntos que cumplen esta propiedad. Esto se debe a las definiciones formales de abiertos y cerrados en topología y a la estructura particular del espacio de los números reales.

# **Demostración de que una bola de centro $o$ y radio $r$ es cerrada**

## **Conjunto Cerrado**

Un conjunto $A$ en un espacio métrico $(E, d)$ es cerrado si contiene todos sus puntos límite. Esto significa que si una secuencia de puntos dentro de $A$ se acerca a un punto $x$ (es decir, $x$ es un límite de la secuencia), entonces $x$ también está en $A$.

**Formalmente**:
- Un conjunto $A$ es cerrado si y solo si contiene todos sus puntos límite.
- Un conjunto $A$ es cerrado si su complemento $A^c$ (todos los puntos que no están en $A$) es un conjunto abierto.

## **Conjunto Abierto**

Un conjunto $B$ en un espacio métrico $(E, d)$ es abierto si, para cualquier punto $y$ en $B$, existe una bola abierta centrada en $y$ que está completamente contenida en $B$.

Esto significa que puedes moverte un poco en cualquier dirección desde cualquier punto dentro del conjunto sin salirte del conjunto.

**Formalmente**:
- Un conjunto $B$ es abierto si, para cada punto $y$ en $B$, existe un $\varepsilon > 0$ tal que la bola $B(y, \varepsilon)$ está contenida en $B$.

- Un conjunto $B$ es abierto si su complemento $B^c$ (todos los puntos que no están en $B$) es un conjunto cerrado.

## **Relación entre Conjuntos Cerrados y Abiertos**

Los conceptos de conjuntos abiertos y cerrados son duales. Esto significa que la definición de uno puede expresarse en términos del otro:

1. **Un conjunto es cerrado si y solo si su complemento es abierto**:
   - Si tienes un conjunto cerrado $A$, su complemento $A^c$ será un conjunto abierto.
   - Esto se debe a que $A$ contiene todos sus puntos límite, lo que implica que $A^c$ no contiene ningún punto límite de $A$, lo que hace que $A^c$ sea abierto.

2. **Un conjunto es abierto si y solo si su complemento es cerrado**:
   - Si tienes un conjunto abierto $B$, su complemento $B^c$ será un conjunto cerrado.
   - Esto se debe a que $B$ no contiene sus puntos límite (puedes moverte ligeramente dentro de $B$ sin salirte), lo que implica que $B^c$ contiene todos esos puntos límite, haciendo que $B^c$ sea cerrado.

# **Ejemplo Visual y Explicación Paso a Paso**

Imaginemos un conjunto $A$ en un espacio métrico. Queremos demostrar que $A$ es cerrado mostrando que su complemento $A^c$ es abierto.

## **Paso a Paso:**

1. **Identificamos el conjunto $A$**:
   - Supongamos que $A$ es una bola cerrada centrada en un punto $o$ con radio $r$, es decir, $A = \{ x \in E \mid d(x, o) \leq r \}$.

2. **Definimos el complemento $A^c$**:
   - El complemento de $A$ incluye todos los puntos $x$ tales que $d(x, o) > r$, es decir, $A^c = \{ x \in E \mid d(x, o) > r \}$.

3. **Queremos mostrar que $A^c$ es abierto**:
   - Tomamos un punto $x$ en $A^c$. Esto significa que $d(x, o) > r$.
   - Definimos una bola abierta alrededor de $x$ con radio $\varepsilon$, donde $\varepsilon = d(x, o) - r$. Como $d(x, o) > r$, $\varepsilon$ es positivo.

4. **Demostramos que esta bola abierta está contenida en $A^c$**:
   - Consideramos cualquier punto $y$ dentro de esta bola abierta $B(x, \varepsilon)$. Esto significa que $d(y, x) < \varepsilon$.
   - Utilizando la desigualdad triangular, tenemos $d(o, y) \leq d(o, x) + d(x, y)$.
   - Sustituyendo $\varepsilon = d(x, o) - r$, obtenemos $d(o, y) \leq d(o, x) + d(x, y) < d(o, x) + \varepsilon = d(o, x) + (d(x, o) - r) = 2d(x, o) - r$.

5. **Conclusión**:
   - Como $d(o, x) > r$, tenemos que $d(o, y) > r$. Esto muestra que $y$ también está en $A^c$.
   - Por lo tanto, $B(x, \varepsilon) \subseteq A^c$, lo que implica que $A^c$ es un conjunto abierto.

De esta manera, hemos demostrado que el conjunto $A$ es cerrado al mostrar que su complemento $A^c$ es abierto.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la bola
o = np.array([0, 0])  # Centro de la bola
r = 1.5  # Radio de la bola

# Crear una figura y un conjunto de ejes
fig, ax = plt.subplots()

# Dibujar la bola cerrada B(o, r)
circle_closed = plt.Circle(o, r, color='cyan', alpha=0.5, label='$B(o, r)$')
ax.add_artist(circle_closed)

# Dibujar el punto x fuera de la bola
x = np.array([3, 0])
plt.plot(x[0], x[1], 'magenta', marker='o', linestyle='None', label='$x \in B(o, r)^c$')

# Dibujar la bola abierta alrededor de x con radio epsilon
epsilon = np.linalg.norm(x - o) - r
circle_epsilon = plt.Circle(x, epsilon, color='yellow', alpha=0.3, linestyle='--', label='$B(x, \epsilon)$')
ax.add_artist(circle_epsilon)

# Configurar los límites del gráfico
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)

# Dibujar el centro de la bola o
plt.plot(o[0], o[1], 'green', marker='o', linestyle='None', label='$o$')

# Añadir etiquetas y leyenda
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Demostración de que $B(o, r)$ es cerrada')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Mostrar el gráfico
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

# Ejemplo en python

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la secuencia
def sequence(n):
    return 1 / n

# Generar términos de la secuencia
N = 100  # Número de términos de la secuencia
terms = np.array([sequence(n) for n in range(1, N + 1)])

# Verificar convergencia
def has_converged(terms, epsilon=1e-6):
    limit = terms[-1]  # Aproximación del límite con el último término
    for term in terms:
        if abs(term - limit) > epsilon:
            return False
    return True

# Graficar la secuencia
plt.plot(range(1, N + 1), terms, marker='o', linestyle='-', color='b')
plt.axhline(y=0, color='r', linestyle='--', label='Límite (0)')
plt.xlabel('n')
plt.ylabel('$a_n$')
plt.title('Secuencia $1/n$')
plt.legend()
plt.grid(True)
plt.show()

# Comprobar convergencia
if has_converged(terms):
    print("La secuencia converge.")
else:
    print("La secuencia no converge.")
```

# Ejemplo en Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las secuencias divergentes
def sequence_n(n):
    return n

def sequence_minus_n(n):
    return -n

def sequence_alternating(n):
    return (-1)**n * n

# Generar los valores de las secuencias
n_values = np.arange(1, 21)  # n desde 1 hasta 20

sequence_n_values = sequence_n(n_values)
sequence_minus_n_values = sequence_minus_n(n_values)
sequence_alternating_values = sequence_alternating(n_values)

# Graficar las secuencias
plt.figure(figsize=(10, 6))

plt.plot(n_values, sequence_n_values, 'o-', label='$a_n = n$')
plt.plot(n_values, sequence_minus_n_values, 'o-', label='$a_n = -n$')
plt.plot(n_values, sequence_alternating_values, 'o-', label='$a_n = (-1)^n \cdot n$')

plt.xlabel('n')
plt.ylabel('$a_n$')
plt.title('Secuencias Divergentes')
plt.legend()
plt.grid(True)
plt.show()
```
