##Parte de un algortimo

1. **Inicio y fin**: Todo algoritmo comienza desde un estado inicial definido y ejecuta una serie de instrucciones ordenadas para llegar a un estado final o de salida.

2. **Exactitud**: Los algoritmos deben definir claramente un orden espec√≠fico y l√≥gico de instrucciones para cada paso, evitando ambig√ºedades o interpretaciones m√∫ltiples.

3. **Secuencia**: Deben seguir una secuencia de pasos ordenados y entendibles, previamente establecidos para resolver el problema o realizar la tarea deseada.

4. **Completitud**: Un algoritmo debe considerar todas las posibles situaciones y condiciones del problema para encontrar una soluci√≥n precisa y correcta.

5. **Finitud**: El n√∫mero de pasos o iteraciones necesarios para ejecutar el algoritmo debe ser finito, de modo que la ejecuci√≥n pueda concluir en un tiempo razonable.

6. **Abstractos**: Los algoritmos representan un modelo o gu√≠a abstracta para la ordenaci√≥n de procesos o la resoluci√≥n de problemas, independientemente de la implementaci√≥n f√≠sica o tecnol√≥gica que se utilice.

##**Sistema binario**
Un n√∫mero binario es un n√∫mero expresado en el sistema de numeraci√≥n de base 2 o sistema de numeraci√≥n binario, un m√©todo para representar n√∫meros que utiliza solo dos s√≠mbolos para los n√∫meros naturales: t√≠picamente "0" y "1".
##**Ejercicio**

1. Dividimos el n√∫mero a convertir, 100 en este caso entre dos. El cociente es 50 y el resto o residuo es 0.
2. Observamos si podemos continuar la divisi√≥n entre 2 logrando un cociente entero. En este caso se puede, cociente 25, residuo 0.
3. Continuamos la divisi√≥n hasta donde sea posible, en este caso al llegar al cociente 1, ya no podemos dividir entre 2 y obtener un cociente entero.
4. El resultado se obtiene colocando de izquierda a derecha, primero el √∫ltimo cociente obtenido, en este caso 1 y luego todos los residuos hasta el primero.
5. Seguimos la direcci√≥n de la flecha para esto.
6. El resultado es 11001002

```python
# Esto es un ejemplo de bloque de c√≥digo en Python
def decimal_a_binario(numero):
    cocientes = []
    while numero > 0:
        cociente = numero // 2
        residuo = numero % 2
        cocientes.append(residuo)
        numero = cociente
    
    # Invertir la lista de residuos para obtener el n√∫mero binario correcto
    binario = ''.join(map(str, cocientes[::-1]))
    return binario

# Ejemplo de uso
numero_decimal = 100
binario_resultado = decimal_a_binario(numero_decimal)
print(f"El n√∫mero binario de {numero_decimal} es: {binario_resultado}")
```

##**Sistema Decimal**
El sistema de numeraci√≥n decimal es el sistema est√°ndar para denotar n√∫meros enteros y no enteros. Es la extensi√≥n a los n√∫meros no enteros del sistema de numeraci√≥n hind√∫-ar√°bigo. La forma de denotar n√∫meros en el sistema decimal a menudo se denomina notaci√≥n decimal.
## **LECTURA DE N√öMEROS DECIMALES**

Leer n√∫meros decimales es muy sencillo. La lectura de n√∫meros decimales se realiza de varias formas. Para verlo, vamos a partir de un ejemplo con el n√∫mero decimal: 45,68.

- Forma 1: Cuarenta y cinco coma sesenta y ocho.
- Forma 2: Cuarenta y cinco con sesenta y ocho.
- Forma 3: Cuarenta y cinco unidades y sesenta y ocho cent√©simas.

```python
# Esto es un ejemplo de bloque de c√≥digo en Python
# Definir una funci√≥n para convertir un n√∫mero decimal en las formas especificadas
def convertir_formas(numero):
    # Obtener la parte entera y la parte decimal
    parte_entera = int(numero)
    parte_decimal = numero - parte_entera
    
    # Forma 1: Cuarenta y cinco coma sesenta y ocho.
    forma_1 = f"{parte_entera} coma {parte_decimal:.2f}"
    
    # Forma 2: Cuarenta y cinco con sesenta y ocho.
    parte_decimal_entera = int(parte_decimal * 100)
    forma_2 = f"{parte_entera} con {parte_decimal_entera}"
    
    # Forma 3: Cuarenta y cinco unidades y sesenta y ocho cent√©simas.
    centesimas = parte_decimal_entera
    forma_3 = f"{parte_entera} unidades y {centesimas} cent√©simas"
    
    # Imprimir resultados
    print("Forma 1:", forma_1)
    print("Forma 2:", forma_2)
    print("Forma 3:", forma_3)

# Ejemplo de uso
numero_decimal = 45.68
```

## **TIPOS DE N√öMEROS DECIMALES**

En este apartado vamos a aprender los tipos de n√∫meros decimales: decimal exacto, decimal peri√≥dico que pueden ser puros o mixtos y decimal no exacto ni peri√≥dico.

### **DECIMAL EXACTO**

Los n√∫meros decimales pueden ser decimal exacto, si est√°n compuestos por un n√∫mero finito de cifras decimales. Por ejemplo 2,5.

### **DECIMAL PERI√ìDICO**

Los n√∫meros decimales tambi√©n pueden ser decimal peri√≥dico. Son aquellos que tienen una cantidad ilimitada de cifras decimales, es decir, la parte decimal llamada

##**Ejemplo**
El n√∫mero 523, por ejemplo, tiene tres cifras. En el sistema decimal, se construye de la siguiente forma, respetando las posiciones correspondientes:

\begin{equation}
(5 \times 10^2) + (2 \times 10^1) + (3 \times 10^0)
