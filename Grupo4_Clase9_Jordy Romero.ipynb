## **üìù Continuidad local y global**


#### **Introducci√≥n**


En el estudio de funciones matem√°ticas, la continuidad es una propiedad fundamental. La continuidad puede ser estudiada de manera global (en toda la extensi√≥n de la funci√≥n) o de manera local (en puntos espec√≠ficos).

***üìå Definici√≥n:*** Una funci√≥n es cont√≠nua cuando podemos dibujarla con un s√≥lo trazo del l√°piz, es decir, sin levantar este del papel.

### **üîªContinuidad en un Punto**
Decimos que una funci√≥n es continua en x=a cuando:

image.png

La condici√≥n implica, estas condiciones:

+ La funci√≥n est√° definida en x=a: ‚àÉ*f(a)*
+ Existe el l√≠mite, y es finito:
$$ \exists \lim_{x \to a} f(x) \in \mathbb{R} $$
Esta condici√≥n implica que los l√≠mites laterales coinciden por ende:
$$ \lim_{x \to a^-} f(x) = \lim_{x \to a^+} f(x) $$

Una funci√≥n sea contin√∫a en un punto implica que la funci√≥n est√° definida en el punto y en un entorno del mismo. Aunque queda fuera del alcance de este nivel, podemos dar una definici√≥n formal bas√°ndonos en estas dos ideas:
+ Formalmente, decimos que una funci√≥n*f(x)* es continua en a cuando, para cualquier entorno de *f(a)* que fijemos, se puede encontrar un entorno de a cuyas im√°genes correspondientes est√©n contenidas en el entorno de *f(a)* . Esto implica:
$$
1. \ \exists f(a) \\
2. \ \forall \varepsilon > 0, \ \exists \delta > 0 \ | \ \text{si} \ |x - a| < \delta \Rightarrow |f(x) - f(a)| < \varepsilon
$$

**‚ú® Donde:**

**Œµ:** Es el valor que define el entorno de f(a) en el eje y

**Œ¥:** Es el valor que define el entorno de a en el eje x

### **‚¨õ Continuidad en un Intervalo Cerrado**
Decimos que una funci√≥n es continua en un intervalo cerrado [a,b] cuando:

+ *f(x)* es continua en el intervalo abierto (a,b). Esto cubre todos los puntos del intervalo salvo los extremos.
+ $$\lim_{x \to \ a+ } f(x) = f(a) \quad \text{y} \quad \lim_{x \to \ a-} f(x) = f(b)$$
Esto contempla la continuidad por la derecha de a y la continuidad por la izquierda de b respectivamente.

### Continuidad local üìü
Una funci√≥n f es continua en un punto x de su dominio si para todo $Œµ$ positivo, existe un $Œ¥$ positivo tal que para cualquier otro punto y en el dominio de f que satisfaga $|x - y| < Œ¥$, se cumple $|f(x) - f(y)| < Œµ$.

En otras palabras, para valores de entrada cercanos a x, los valores de salida de f tambi√©n son cercanos. La continuidad local se puede visualizar como la capacidad de "hacer zoom" en un punto sin observar cambios abruptos en el comportamiento de la funci√≥n.

Por tanto, una funci√≥n $f$ es continua en un punto $x = c$  si se cumplen las siguientes condiciones:

1. $f(c)$ est√° definida (existe).
2. El l√≠mite de $f(x)$ cuando $x$ tiende a $c$ existe.
3. El l√≠mite de $f(x)$ cuando $x$ tiende a $c$ es igual a $f(c)$.

Formalmente, podemos expresar esto como:

$$ \lim_{x \to c} f(x) = f(c) $$


### **Ejemplo 1**

Consideremos la funci√≥n $f(x) = x^2$. Queremos verificar si es continua en $x = 2$.

*Proceso:*

$f(2) = 2^2 = 4$$f(2) = 2^2 = 4$. (La funci√≥n est√° definida en $x = 2$).

$\lim_{x \to 2} f(x) = \lim_{x \to 2} x^2 = 4$ (El l√≠mite existe).

Por lo tanto el $\lim_{x \to 2} f(x) = f(2)$, indica que la funci√≥n es continua en $x = 2$.

**Gr√°fico**



```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 4, 400)
y = x**2

plt.plot(x, y, label='$f(x) = x^2$')
plt.scatter([2], [4], color='red')  # Punto de inter√©s (2,4)
plt.axvline(x=2, color='gray', linestyle='--')
plt.axhline(y=4, color='gray', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gr√°fico de $f(x) = x^2$ en $x = 2$')
plt.legend()
plt.grid(True)
plt.show()
```

Nota: Es continua en $x=2$

### Continuidad Global üåé

Una funci√≥n es continua en un intervalo si es continua en cada punto de ese intervalo. Si una funci√≥n es continua en toda la recta real, decimos que es continua globalmente.

Naturalmente, decimos que una funci√≥n $f : E ‚Üí F$  es continua en un conjunto no vac√≠o $A ‚äÇ E$ cuando es continua en todo punto $x ‚àà A$. Si $f$ es continua en $E$ decimos simplemente que $f$ es continua.

**Para cualquier funci√≥n $f : E ‚Üí F$ las siguientes afirmaciones son equivalentes:**


- (i) f es continua.
- (ii) Para todo abierto $V ‚äÇ F$ , se tiene que $f^{-1}$ (V) es un abierto de $E$.
- (iii) Para todo cerrado $C ‚äÇ F$, se tiene que $f^{-1}$ (C) es un cerrado de $E$.
- (iv) f preserva la convergencia de sucesiones: para toda sucesi√≥n convergente ${x_n}$ de puntos de E, la sucesi√≥n ${ f(x_n)}$ es convergente.

Las caracterizaciones de la continuidad, dadas por las condiciones (ii) y (iii) del √∫ltimo
resultado, se usan a menudo para probar que ciertos subconjuntos de E son abiertos o cerrados.

Por ejemplo, tomando $F$ = $\mathbb{R}$ observamos que si $f : E ‚Üí R$ es una funci√≥n continua, entonces los conjuntos:

- ${x ‚àà E : f(x) > 0}$ = $f^{-1}$ ($\mathbb{R}^+$)
- ${x ‚àà E : f(x) < 0}$ = $f^{-1}$ ($\mathbb{R}^+$)

Por lo tanto, son abiertos, mientras que el conjunto ${x ‚àà E : 0 ‚â§ f(x) ‚â§ 1}$ = $f^{-1}$ $([0,1])$ es cerrado.

Por otro lado, resaltamos que la continuidad de f , no s√≥lo en un punto, sino en todo el espacio E , tiene **car√°cter local**, en el sentido de que se puede comprobar con s√≥lo conocer f localmente: f es continua si, y s√≥lo si, para cada  $x ‚àà E$ existe $U ‚àà U(x)$ tal que $f |_U$ es continua.


```python
def H(x):
  if x < 0:
    return 0
  else:
    return 1
```

Nota: La funci√≥n $H(x)$ es continua en todos los puntos excepto en $x = 0$. En este punto, la funci√≥n experimenta un salto abrupto de 0 a 1. Por lo tanto, la funci√≥n es continua localmente en todos los puntos excepto en x = 0, pero no es continua globalmente.



```python
import matplotlib.pyplot as plt
import numpy as np

def H(x):
  return np.where(x < 0, 0, 1)  # Usamos np.where para aplicar la funci√≥n a cada elemento

x = np.arange(-2, 2, 0.1)
y = H(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('H(x)')
plt.title('Funci√≥n escal√≥n de Heaviside')
plt.grid(True)
plt.show()
```

Nota: La imagen muestra una gr√°fica que indica la funci√≥n escal√≥n de Heaviside, H(x), con √©nfasis en la discontinuidad en $x = 0$.

## **üìñ Espacios M√©tricos y Conjuntos Abiertos**

### üìê **Espacios M√©tricos**

#### *¬øQu√© es un Espacio M√©trico?*

Un **espacio m√©trico** es una colecci√≥n de puntos que tiene una forma de medir la distancia entre cada par de puntos. Esta "forma de medir" se llama **m√©trica**. La m√©trica es una funci√≥n que toma dos puntos y devuelve un n√∫mero que representa la distancia entre ellos.





#### üß©**Propiedades de una M√©trica**

Para que una funci√≥n sea una m√©trica, debe cumplir cuatro propiedades:

1. **No negatividad**: La distancia entre dos puntos siempre es un n√∫mero no negativo.
   - $d(x, y) \geq 0$

2. **Identidad**: La distancia entre un punto y s√≠ mismo es cero.
   - $d(x, x) = 0$

3. **Simetr√≠a**: La distancia de un punto $A$ a un punto $B$ es la misma que la distancia de $B$ a $A$.
   - $d(x, y) = d(y, x)$

4. **Desigualdad Triangular**: La distancia directa entre dos puntos es siempre menor o igual a la suma de las distancias a trav√©s de un tercer punto.
   - $d(x, y) \leq d(x, z) + d(z, y)$


#### üé¢**Ejemplos de Espacios M√©tricos**

1. **L√≠nea Recta (R)**

   En una l√≠nea recta (como una recta num√©rica), podemos usar la distancia absoluta entre dos puntos.
   - M√©trica: $d(x, y) = |x - y|$
   
   **Ejemplo:**
   - Si $x = 2$ y $y = 5$, entonces $d(2, 5) = |2 - 5| = 3$.

2. **Plano (R¬≤)**

   En un plano (como una hoja de papel), podemos usar la distancia euclidiana entre dos puntos.
   - M√©trica: $d((x_1, y_1), (x_2, y_2)) = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$
   
   **Ejemplo:**
   - Si $A = (1, 2)$ y $B = (4, 6)$, entonces $d(A, B) = \sqrt{(4 - 1)^2 + (6 - 2)^2} = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5$.

3. **Espacios de Funciones**

   Podemos definir una m√©trica entre funciones, por ejemplo, tomando la m√°xima diferencia entre sus valores.
   - M√©trica: $d(f, g) = \sup_{x \in [a, b]} |f(x) - g(x)|$
   
   **Ejemplo:**
   - Si $f(x) = x^2$ y $g(x) = x$ en el intervalo $[0, 1]$, entonces $d(f, g) = \max_{x \in [0, 1]} |x^2 - x|$. Aqu√≠, el m√°ximo se alcanza en $x = 1$, por lo que $d(f, g) = |1^2 - 1| = 0$.

- **Estudio de Funciones:** En el an√°lisis, los espacios m√©tricos se utilizan para definir conceptos como continuidad, l√≠mites, y derivadas. Las propiedades de las funciones se exploran en t√©rminos de la m√©trica que define las distancias entre puntos en el dominio y en el codominio.

- **Geometr√≠a de Espacios de Banach y Hilbert:** Los espacios m√©tricos son la base para espacios m√°s especializados como los espacios de Banach y Hilbert, que se utilizan en la geometr√≠a funcional para estudiar espacios vectoriales con m√©tricas inducidas por normas y productos internos.

- **Algoritmos de Optimizaci√≥n:** En la optimizaci√≥n, los espacios m√©tricos se utilizan para definir conceptos como cercan√≠a a un √≥ptimo y para desarrollar m√©todos iterativos como el m√©todo del gradiente descendente.

- **Teoremas de Existencia de √ìptimos:** Los espacios m√©tricos son el marco para teoremas que garantizan la existencia de puntos √≥ptimos bajo ciertas condiciones, como el teorema de Brouwer para funciones continuas en espacios m√©tricos.

- **Espacios de Probabilidad:** En la teor√≠a de la probabilidad, los espacios m√©tricos se utilizan para definir conceptos como la distancia entre distribuciones de probabilidad y para desarrollar teor√≠as de convergencia de variables aleatorias.

- **Espacios M√©tricos en Teor√≠a de la Informaci√≥n:** Los espacios m√©tricos se usan en la teor√≠a de la informaci√≥n para definir distancias entre distribuciones de probabilidad y para medir la divergencia entre eventos aleatorios.

- **M√©todos de Clustering:** En el an√°lisis de datos, los espacios m√©tricos son fundamentales para algoritmos de clustering como k-medias y an√°lisis de agrupamiento basado en distancias.

- **Reducci√≥n de Dimensionalidad:** Los conceptos m√©tricos se utilizan en t√©cnicas de reducci√≥n de dimensionalidad, como el An√°lisis de Componentes Principales (PCA) y t√©cnicas basadas en distancias como el MDS (Multidimensional Scaling).

- **Espacios de Caracter√≠sticas:** En machine learning, los espacios m√©tricos ayudan a definir y analizar espacios de caracter√≠sticas para algoritmos de aprendizaje supervisado y no supervisado.

- **Optimizaci√≥n de Modelos:** Los espacios m√©tricos son fundamentales para la optimizaci√≥n de modelos de machine learning, incluyendo la minimizaci√≥n de funciones de costo y la evaluaci√≥n del desempe√±o de los modelos.

### üßø **Conjuntos Abiertos**

En matem√°ticas, especialmente en topolog√≠a, un **conjunto abierto** es un concepto fundamental. Vamos a explicarlo de una manera simple.

#### *¬øQu√© es un Conjunto Abierto?*

Un conjunto abierto es un conjunto de puntos que no incluye sus fronteras. Imagina que tienes un c√≠rculo dibujado en un papel. Si consideras todos los puntos **dentro** del c√≠rculo, pero **no** los puntos que est√°n exactamente en la circunferencia, tienes un conjunto abierto.




#### ü™î**Ejemplo: C√≠rculo Abierto**

En el gr√°fico de abajo, tenemos un c√≠rculo abierto. La regi√≥n azul clara dentro del c√≠rculo representa todos los puntos que forman parte del conjunto abierto. La l√≠nea azul que forma la circunferencia no es parte del conjunto abierto.



```python
import matplotlib.pyplot as plt
import numpy as np

# Definici√≥n del intervalo y de los puntos
x = np.linspace(-1, 1, 400)
y = np.sqrt(1 - x**2)

fig, ax = plt.subplots()

# Dibujo del c√≠rculo
ax.plot(x, y, color='blue')
ax.plot(x, -y, color='blue')

# Puntos dentro del c√≠rculo
theta = np.linspace(0, 2 * np.pi, 100)
r = 0.5
x_inside = r * np.cos(theta)
y_inside = r * np.sin(theta)

ax.fill(x_inside, y_inside, color='lightblue', alpha=0.5)

# Ajustes del gr√°fico
ax.set_aspect('equal')
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
ax.set_title("Ejemplo de Conjunto Abierto: C√≠rculo Abierto")

plt.show()
```


#### üß©**Propiedades de los Conjuntos Abiertos**

1. **Todo punto dentro del conjunto tiene un "espacio alrededor" que tambi√©n est√° dentro del conjunto.**
   - Si tomas cualquier punto dentro del c√≠rculo abierto, puedes dibujar un c√≠rculo m√°s peque√±o alrededor de ese punto que a√∫n se encuentra completamente dentro del c√≠rculo grande.

2. **No incluye los puntos en el borde.**
   - Si tomas un punto en el borde del c√≠rculo (en la circunferencia), no es parte del conjunto abierto.


#### üé¢**Ejemplo en la Vida Real**

Imagina que tienes un parque circular. Si decimos que el parque es un conjunto abierto, significa que cualquier persona caminando dentro del parque no puede caminar exactamente en la cerca del borde, pero puede moverse libremente en cualquier lugar dentro del parque.

```python
import matplotlib.pyplot as plt
import numpy as np

# Definici√≥n del c√≠rculo grande (el parque)
theta = np.linspace(0, 2 * np.pi, 100)
r = 1
x_park = r * np.cos(theta)
y_park = r * np.sin(theta)

# Definici√≥n de los puntos caminando dentro del parque (c√≠rculo peque√±o)
r_inner = 0.8
theta_inner = np.linspace(0, 2 * np.pi, 100)
x_inner = r_inner * np.cos(theta_inner)
y_inner = r_inner * np.sin(theta_inner)

# Dibujo del parque
fig, ax = plt.subplots()
ax.plot(x_park, y_park, label='Borde del parque', color='blue')
ax.fill(x_inner, y_inner, label='Personas caminando', color='lightblue', alpha=0.5)

# Puntos caminando dentro del parque
for _ in range(10):
    angle = np.random.uniform(0, 2 * np.pi)
    distance = np.random.uniform(0, r_inner)
    x_person = distance * np.cos(angle)
    y_person = distance * np.sin(angle)
    ax.plot(x_person, y_person, 'ro')

# Ajustes del gr√°fico
ax.set_aspect('equal')
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
ax.grid(color='gray', linestyle='--', linewidth=0.5)
ax.set_title("Ejemplo: Parque Circular como Conjunto Abierto")
ax.legend()

plt.show()
```

- La l√≠nea azul representa el borde del parque (circunferencia del c√≠rculo).

- La regi√≥n azul clara muestra el √°rea donde las personas pueden caminar.

- Los puntos rojos representan personas caminando dentro del parque, pero no exactamente en el borde.

### üìê***Espacios M√©tricos***

#### *Definici√≥n de Espacio M√©trico*
Un **espacio m√©trico** es un par ordenado $ (X, d) $ donde $X$ es un conjunto no vac√≠o y $ d $ es una funci√≥n denominada **m√©trica** o **distancia**, que asigna a cada par de puntos $ x, y \in X $ un n√∫mero real $ d(x, y) $, y satisface las siguientes propiedades:

1. **No negatividad**: $ d(x, y) \geq 0\ $para todos $ x, y \in X $ y $ d(x, y) = 0 $ si y solo si $ x = y $.
2. **Simetr√≠a**: $ d(x, y) = d(y, x) $ para todos $ x, y \in X\ $.
3. **Desigualdad triangular**: $ d(x, z) \leq d(x, y) + d(y, z) $ para todos $ x, y, z \in X $.

### Ejemplos de Espacios M√©tricos
1. **Espacio Euclidiano**: $ X = \mathbb{R}^n $ y  

    $$d(x, y) = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}.$$

2. **Espacio Discreto**: $ X $ es cualquier conjunto y

    $$d(x, y) = 1 \text{ si } x \ne y, \text{ y } d(x, y) = 0 \text{ si } x = y.$$

### Propiedades de Espacios M√©tricos
- **Convergencia**: Una secuencia $ (x_n) $ en $ X $ converge a $ x \in X $ si para todo $ \epsilon > 0 $, existe un $ N \in \mathbb{N} $

    $$d(x_n, x) < \epsilon \text{ para todo } n \geq N.$$

- **Completitud**: Un espacio m√©trico $ X $ es completo si toda sucesi√≥n de Cauchy en $ X $ converge a un punto en $ X $. Una sucesi√≥n $ (x_n) $ es de Cauchy si para todo $ \epsilon > 0 $, existe un $ N \in \mathbb{N} $ tal que $ d(x_n, x_m) < \epsilon $ para todo $ n, m \geq N $.


### üßø***Conjuntos Abiertos***

#### *Definici√≥n de Conjunto Abierto*
Un subconjunto $ U \subset X $ de un espacio m√©trico $ (X, d) $ es **abierto** si para todo $ x \in U $, existe un $ \epsilon > 0 $ tal que la bola abierta $ B(x, \epsilon) = \{ y \in X : d(x, y) < \epsilon \} $ est√° contenida en $ U $.


### Propiedades de los Conjuntos Abiertos
- **Uni√≥n de Conjuntos Abiertos**: La uni√≥n arbitraria de conjuntos abiertos es un conjunto abierto.
- **Intersecci√≥n de Conjuntos Abiertos**: La intersecci√≥n finita de conjuntos abiertos es un conjunto abierto.
- **Espacios M√©tricos y Topolog√≠a**: La colecci√≥n de todos los conjuntos abiertos en un espacio m√©trico $ X $ forma una topolog√≠a sobre  $ X $ , llamada la **topolog√≠a m√©trica**.

### Ejemplos de Conjuntos Abiertos
1. En el espacio euclidiano $ \mathbb{R}^n $, cualquier bola abierta $ B(x, \epsilon) $ es un conjunto abierto.
2. En el espacio discreto, cada subconjunto es abierto.





### üé´***Relaci√≥n entre Espacios M√©tricos y Conjuntos Abiertos***

Bolas Abiertas y Cerradas
- Una **bola abierta** de radio $ \epsilon $ centrada en $ x \in X $ se denota por $ B(x, \epsilon) $ y se define como $ B(x, \epsilon) = \{ y \in X : d(x, y) < \epsilon \} $.

- Una **bola cerrada** de radio $ \epsilon $ centrada en $ x \in X $ se denota por $ \overline{B}(x, \epsilon) $ y se define como $ \overline{B}(x, \epsilon) = \{ y \in X : d(x, y) \leq \epsilon \} $.


### Conjuntos Cerrados
Un subconjunto $ A \subset X $ es **cerrado** si su complemento $ X \setminus A $ es abierto. En un espacio m√©trico, los conjuntos cerrados pueden caracterizarse como aquellos que contienen todos sus puntos de acumulaci√≥n.

### Frontera de un Conjunto
La **frontera** de un conjunto $ A \subset X $, denotada $ \partial A $, es el conjunto de puntos que est√°n en el cierre de $ A $ pero no en el interior de $ A $.

### ‚úçÔ∏è***Ejemplos y Aplicaciones***

#### üï≥Ô∏è**Ejemplo 1:** Verificaci√≥n de Conjuntos Abiertos en Python
En este ejemplo, implementaremos una funci√≥n en Python que verifica si un conjunto dado en un espacio m√©trico es abierto o no, utilizando la m√©trica est√°ndar en el espacio euclidiano de los Reales.

```python
import numpy as np

def es_conjunto_abierto(X, U, epsilon=1e-6):
    """
    Verifica si el conjunto U es abierto en el espacio m√©trico X.

    Args:
    - X: Conjunto de puntos en el espacio m√©trico (aqu√≠ se asume el espacio euclidiano R^n).
    - U: Conjunto que se quiere verificar si es abierto.
    - epsilon: Tolerancia para la verificaci√≥n de la m√©trica (default: 1e-6).

    Returns:
    - True si U es abierto, False si no lo es.
    """
    for u in U:
        abierto = False
        for x in X:
            if np.linalg.norm(u - x) < epsilon:
                abierto = True
                break
        if not abierto:
            return False
    return True

# Ejemplo de uso
X = np.array([[0, 0], [1, 1], [2, 2], [3, 3]])  # Espacio m√©trico R^2
U1 = np.array([[0.5, 0.5], [1.5, 1.5]])  # Conjunto cerrado
U2 = np.array([[1, 1], [2, 2]])  # Conjunto abierto

print("¬øU1 es abierto?", es_conjunto_abierto(X, U1))  # False
print("¬øU2 es abierto?", es_conjunto_abierto(X, U2))  # True

```

#### üï≥Ô∏è**Ejemplo 2:** Visualizaci√≥n de Conjuntos Abiertos en un Espacio M√©trico
En este ejemplo, vamos a visualizar gr√°ficamente conjuntos abiertos en un espacio m√©trico utilizando la biblioteca matplotlib en Python.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

def visualizar_conjunto_abierto(X, U, epsilon=0.5):
    """
    Visualiza un conjunto abierto U en el espacio m√©trico X (R^2).

    Args:
    - X: Conjunto de puntos en el espacio m√©trico (aqu√≠ se asume el espacio euclidiano R^2).
    - U: Conjunto abierto que se desea visualizar.
    - epsilon: Radio para dibujar las bolas abiertas (default: 0.5).
    """
    fig, ax = plt.subplots()

    # Dibujar todos los puntos en X
    ax.scatter(X[:, 0], X[:, 1], color='blue', label='Espacio m√©trico X')

    # Dibujar el conjunto U
    for u in U:
        circle = Circle(u, epsilon, color='green', alpha=0.2)
        ax.add_patch(circle)

    ax.set_aspect('equal')
    ax.set_title('Visualizaci√≥n de Conjunto Abierto en Espacio M√©trico')
    ax.legend()
    plt.xlabel('Coordenada X')
    plt.ylabel('Coordenada Y')
    plt.grid(True)
    plt.show()

# Ejemplo de uso
X = np.array([[0, 0], [1, 1], [2, 2], [3, 3]])  # Espacio m√©trico R^2
U = np.array([[1, 1], [2, 2]])  # Conjunto abierto

visualizar_conjunto_abierto(X, U)
```
