### **3. üìà Teorema Fundamental del C√°lculo ‚ú®**

Este teorema establece una conexi√≥n profunda üîó entre la integraci√≥n y la diferenciaci√≥n, mostrando que son procesos inversos üîÑ. Se compone de dos partes:

1Ô∏è‚É£ **Primera parte:** Afirma que toda funci√≥n continua $f(x)$ en un intervalo cerrado $[a, b]$ tiene una **funci√≥n antiderivada** $F(x)$. La funci√≥n antiderivada es aquella cuya derivada es $f(x)$.

2Ô∏è‚É£ **Segunda parte:** Proporciona un m√©todo para evaluar la **integral definida** $\int_a^b f(x) \,dx$ de una funci√≥n continua $f(x)$ en el intervalo $[a, b]$. Esta evaluaci√≥n se realiza mediante la f√≥rmula:

$$\int_a^b f(x) \,dx = F(b) - F(a)$$

Es decir, el valor de la integral definida es la diferencia entre la funci√≥n antiderivada evaluada en los extremos del intervalo de integraci√≥n.

---

### **4. Ejemplo: C√°lculo del √Årea bajo una Curva üìâ**

Consideremos la funci√≥n $f(x) = x^2$ en el intervalo $[1, 2]$. Queremos calcular el √°rea sombreada bajo la curva $f(x)$ entre $x = 1$ y $x = 2$.

**Soluci√≥n:**

1. **Encontrar la funci√≥n antiderivada:** La funci√≥n antiderivada de $f(x) = x^2$ es $F(x) = \dfrac{x^3}{3}$.

2. **Aplicar el teorema fundamental del c√°lculo:**

$$\int_1^2 x^2 \,dx = F(2) - F(1) = \left( \dfrac{2^3}{3} \right) - \left( \dfrac{1^3}{3} \right) = \dfrac{7}{3}$$

**Resultado:** El √°rea sombreada bajo la curva $f(x) = x^2$ entre $x = 1$ y $x = 2$ es de $\dfrac{7}{3}$ unidades cuadradas.

---

### **5.üìè ¬øCu√°l es la Interpretaci√≥n Geom√©trica del Teorema Fundamental? üìê**

![Texto alternativo](https://media1.tenor.com/m/X9SGirjkxxEAAAAC/tenor.gif)

üîçEl teorema fundamental del c√°lculo no solo establece una relaci√≥n matem√°tica, sino que tambi√©n posee una profunda interpretaci√≥n geom√©trica.üìê

üìè La integral definida representa el √°rea neta acumulada entre la curva $f(x)$ y el eje $x$ desde el punto $x = a$ hasta el punto $x = b$.

üìà La funci√≥n antiderivada, por otro lado, representa la familia de funciones primitivas de $f(x)$, cada una con una constante aditiva diferente.

üî¢ Al evaluar la funci√≥n antiderivada en los extremos del intervalo de integraci√≥n, se obtiene la diferencia de alturas de las curvas primitivas en dichos puntos, lo que corresponde al √°rea neta acumulada.






```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the function f(x) and its antiderivative F(x)
def f(x):
    return x**2  # Example function: f(x) = x^2

def F(x):
    return (x**3) / 3  # Antiderivative of f(x): F(x) = (x^3) / 3

# Define the interval [a, b]
a, b = 1, 3

# Compute the area under the curve using definite integral
area, _ = quad(f, a, b)

# Generate x values for plotting
x_values = np.linspace(0, 4, 400)
y_values = f(x_values)

# Generate values for antiderivative plot
F_values = F(x_values)

# Plot the function and the area under the curve
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = x^2', color='blue')
plt.fill_between(x_values, y_values, where=((x_values >= a) & (x_values <= b)), color='lightblue', alpha=0.6, label=f'√Årea = {area:.2f}')

# Plot antiderivative lines
plt.plot(x_values, F_values, label='F(x) = (x^3)/3', color='green', linestyle='--')
plt.scatter([a, b], [F(a), F(b)], color='red')
plt.text(a, F(a), f'F(a)={F(a):.2f}', fontsize=10, verticalalignment='bottom', horizontalalignment='right')
plt.text(b, F(b), f'F(b)={F(b):.2f}', fontsize=10, verticalalignment='bottom', horizontalalignment='right')

# Highlight the difference in heights of antiderivative
plt.vlines([a, b], [0], [F(a), F(b)], colors='red', linestyles='dotted')

# Customize the plot
plt.title('√Årea bajo la curva f(x) = x^2 y funci√≥n antiderivada F(x) = (x^3)/3')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.legend()
plt.grid(True)
plt.xlim(0, 4)
plt.ylim(0, 30)

# Show the plot
plt.show()
```

### **üîÑ OPERACIONES ENTRE CONJUNTOS**

| **Operaci√≥n**                 | **Descripci√≥n**                                                                                           | **Notaci√≥n Matem√°tica**                                               |
|-------------------------------|-----------------------------------------------------------------------------------------------------------|----------------------------------------------------------------------|
| **‚à™ ‚û° "Uni√≥n"** üåê            | La uni√≥n de dos conjuntos A y B incluye todos los elementos que est√°n en A, en B o en ambos.              | $x ‚àà A ‚à™ B$ ‚û° $x ‚àà A \lor x ‚àà B$ <br> $A ‚à™ B$ = ${ x ‚àà Œ© : (x ‚àà A) \lor (x ‚àà B)}$ |

### **OPERACIONES ENTRE CONJUNTOS EN PYHTON**
