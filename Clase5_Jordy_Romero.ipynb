##**Funci√≥n Inyectiva**
En matem√°ticas, una funci√≥n inyectiva es una funci√≥n f que asigna elementos distintos de su dominio a elementos distintos; es decir, $x‚ÇÅ ‚â† x‚ÇÇ$ implica $f ‚â† f$. En otras palabras, cada elemento del codominio de la funci√≥n es la imagen de, como m√°ximo, un elemento de su dominio.

##**Criterio de la recta horizontal**
Tambi√©n se puede determinar si una funci√≥n es inyectiva a trav√©s de su representaci√≥n
gr√°fica en el sistema de coordenadas. Se utiliza el criterio de la recta horizontal, el cual,
consiste en trazar (como dice el nombre), rectas horizontales. Si la recta corta a la gr√°fica en
un solo punto, la funci√≥n es inyectiva, y si corta en m√°s de un punto la funci√≥n no es
inyectiva.

##**Funci√≥n Sobreyectiva**
Una funci√≥n es sobreyectiva cuando todos los elementos del codominio son imagen de al menos
un elemento del dominio, es decir, el codominio es igual al recorrido.
En lenguaje algebraico, una funci√≥n f de A en B es sobreyectiva cuando;
$$Rec (f) = B$$
Si se representara una funci√≥n sobreyectiva, en un diagrama sagital, de una funci√≥n $f$ de $A$ en $B$,
ver√≠amos que todos los elementos $B$ tienen al menos una flecha desde el conjunto $A$.
Tambi√©n se conoce como funci√≥n epiyectiva.

##**Funci√≥n Biyectiva**
Una funci√≥n es biyectiva cuando es inyectiva y sobreyectiva a la vez, es decir, cuando todas las
im√°genes tienen una sola preimagen y no existen elementos del codominio que no tengan una
preimagen.
En lenguaje algebraico, una funci√≥n f de A en B es biyectiva si se cumple que, siendo $x$ e $y$
elementos de $A$:
$$f (x) = f (y)‚Üíx = y$$
$$Rec (f) = B$$
Si se representa una funci√≥n biyectiva, en un diagrama sagital, de una funci√≥n $f$ de $A en B$,
ver√≠amos que todos los elementos B tienen s√≥lo una flecha desde el conjunto $A$.

###**Funci√≥n Inversa**
```python
# Definici√≥n de la funci√≥n original f(x)
def f(x):
    return 2*x + 3

# Funci√≥n para encontrar la funci√≥n inversa f^{-1}(y)
def f_inversa(y):
    return (y - 3) / 2

# Ejemplo de uso
x = 5
print(f"f({x}) =", f(x))

# Encontrar f^{-1}(y) para y = 13
y = 13
print(f"f_inversa({y}) =", f_inversa(y))
```
###**Funci√≥n Compuesta**
```python
# Definici√≥n de las funciones f(x) y g(x)
def f(x):
    return 2*x + 3

def g(x):
    return x**2

# Definici√≥n de la funci√≥n compuesta (f ‚àò g)(x)
def funcion_compuesta(x):
    return f(g(x))

# Ejemplo de uso
x = 2
resultado = funcion_compuesta(x)
print(f"El resultado de (f ‚àò g)({x}) es:", resultado)
```
