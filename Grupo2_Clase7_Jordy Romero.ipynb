### üîç **¬øQu√© es el L√≠mite de una sucesi√≥n?** üîç
![Texto alternativo](https://media.tenor.com/MVK4B9LYZGAAAAAi/ijustboughtlandinthemetaverse.gif)

Formalmente, una sucesi√≥n es una funci√≥n que asigna n√∫meros naturales a un conjunto de n√∫meros, com√∫nmente denotados como $\{a_n\}_{n=1}^{\infty}$. El l√≠mite de una sucesi√≥n, si existe, es el valor al que se aproximan los t√©rminos de la sucesi√≥n a medida que el √≠ndice $n$ tiende a infinito. Se representa matem√°ticamente como:

$$ \lim_{n \to \infty} a_n = L $$

donde $L$ es el l√≠mite. Intuitivamente, esto significa que para cualquier valor arbitrariamente peque√±o $\epsilon > 0$, existe un √≠ndice natural $N$ tal que para todo $n > N$, se cumple la desigualdad:

$$ |a_n - L| < \epsilon $$


```python
import matplotlib.pyplot as plt
import numpy as np

# Definici√≥n de la sucesi√≥n a_n (en este caso, una sucesi√≥n que converge a 0.5)
def sucesion(n):
    return 1 / (n + 1)

# Definici√≥n del l√≠mite L
L = 0.5

# Valores de n para la gr√°fica
n_values = np.arange(1, 21)  # Valores de n desde 1 hasta 20

# T√©rminos de la sucesi√≥n a_n
sucesion_values = [sucesion(n) for n in n_values]

# Gr√°fica de la sucesi√≥n y su l√≠mite
plt.figure(figsize=(10, 6))
plt.plot(n_values, sucesion_values, marker='o', linestyle='-', color='b', label='$a_n$')
plt.hlines(L, n_values[0], n_values[-1], colors='r', linestyles='dashed', label='$L$')
plt.xlabel('n')
plt.ylabel('$a_n$')
plt.title('Aproximaci√≥n de una sucesi√≥n a su l√≠mite')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### üåü **Propiedades de los l√≠mites de sucesiones** üåü

Las propiedades b√°sicas que gobiernan los l√≠mites de sucesiones son las siguientes:

* **Propiedad aditiva:** Si $\lim_{n \to \infty} a_n = L$ y $\lim_{n \to \infty} b_n = M$, entonces se cumple que $\lim_{n \to \infty} (a_n + b_n) = L + M$.
* **Propiedad multiplicativa:** Si $\lim_{n \to \infty} a_n = L$ y $c$ es un n√∫mero real, entonces se cumple que $\lim_{n \to \infty} c \cdot a_n = c \cdot L$.
* **Propiedad del cociente:** Si $\lim_{n \to \infty} a_n = L$ y $\lim_{n \to \infty} b_n = M$, donde $M \neq 0$, entonces se cumple que $\lim_{n \to \infty} \frac{a_n}{b_n} = \frac{L}{M}$.
* **Teorema del l√≠mite:** Si $f$ es una funci√≥n continua en un intervalo cerrado $[a, b]$ y $x_0 \in [a, b]$, entonces se cumple que $\lim_{n \to \infty} f(x_n) = f(x_0)$, donde $\{x_n\}_{n=1}^{\infty}$ es una sucesi√≥n de n√∫meros reales que converge a $x_0$.

---


```python
import matplotlib.pyplot as plt
import numpy as np

# Definici√≥n de la sucesi√≥n a_n y b_n para las propiedades
def sucesion_a(n):
    return 1 / (n + 1)

def sucesion_b(n):
    return 0.5 / (n + 1)

# Valores de n para la gr√°fica
n_values = np.arange(1, 21)  # Valores de n desde 1 hasta 20

# T√©rminos de las sucesiones a_n y b_n
sucesion_a_values = [sucesion_a(n) for n in n_values]
sucesion_b_values = [sucesion_b(n) for n in n_values]

# Valores de las propiedades de los l√≠mites
L = 0.5
M = 0.2
c = 2

# Gr√°fica de las propiedades de los l√≠mites
plt.figure(figsize=(12, 8))

# Propiedad aditiva: a_n + b_n
plt.subplot(2, 2, 1)
plt.plot(n_values, sucesion_a_values, marker='o', linestyle='-', color='b', label='$a_n$')
plt.plot(n_values, sucesion_b_values, marker='s', linestyle='-', color='g', label='$b_n$')
plt.plot(n_values, np.add(sucesion_a_values, sucesion_b_values), marker='d', linestyle='-', color='r', label='$a_n + b_n$')
plt.hlines(L + M, n_values[0], n_values[-1], colors='m', linestyles='dashed', label='$L + M$')
plt.xlabel('n')
plt.ylabel('Valores')
plt.title('Propiedad aditiva')
plt.legend()
plt.grid(True)

# Propiedad multiplicativa: c * a_n
plt.subplot(2, 2, 2)
plt.plot(n_values, sucesion_a_values, marker='o', linestyle='-', color='b', label='$a_n$')
plt.plot(n_values, np.multiply(c, sucesion_a_values), marker='s', linestyle='-', color='g', label='$c \cdot a_n$')
plt.hlines(c * L, n_values[0], n_values[-1], colors='m', linestyles='dashed', label='$c \cdot L$')
plt.xlabel('n')
plt.ylabel('Valores')
plt.title('Propiedad multiplicativa')
plt.legend()
plt.grid(True)

# Propiedad del cociente: a_n / b_n
plt.subplot(2, 2, 3)
plt.plot(n_values, sucesion_a_values, marker='o', linestyle='-', color='b', label='$a_n$')
plt.plot(n_values, sucesion_b_values, marker='s', linestyle='-', color='g', label='$b_n$')
plt.plot(n_values, np.divide(sucesion_a_values, sucesion_b_values), marker='d', linestyle='-', color='r', label='$\\frac{a_n}{b_n}$')
plt.hlines(L / M, n_values[0], n_values[-1], colors='m', linestyles='dashed', label='$\\frac{L}{M}$')
plt.xlabel('n')
plt.ylabel('Valores')
plt.title('Propiedad del cociente')
plt.legend()
plt.grid(True)

# Teorema del l√≠mite: f(x_n) -> f(x_0)
plt.subplot(2, 2, 4)
x_values = np.linspace(0, 2, 100)
y_values = np.sin(np.pi * x_values)  # Funci√≥n seno como ejemplo
x_n = np.array([1.5, 1.75, 1.9])  # Ejemplo de sucesi√≥n x_n
plt.plot(x_values, y_values, color='b', label='$f(x)$')
plt.scatter(x_n, np.sin(np.pi * x_n), color='r', label='$f(x_n)$')
plt.hlines(np.sin(np.pi * 1.5), 0, 2, colors='m', linestyles='dashed', label='$f(x_0)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Teorema del l√≠mite')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### üîÑ **Convergencia y divergencia de sucesiones** üîÑ

Una sucesi√≥n se considera convergente si posee un l√≠mite finito. Si el l√≠mite es infinito, la sucesi√≥n se clasifica como divergente. Tambi√©n existe el caso en que la sucesi√≥n no tenga un l√≠mite, en cuyo caso se denomina oscilante.

---

```python
import matplotlib.pyplot as plt
import numpy as np

# Funci√≥n para generar una sucesi√≥n convergente
def sucesion_convergente(n):
    return 1 / (n + 1)

# Funci√≥n para generar una sucesi√≥n divergente
def sucesion_divergente(n):
    return n + 1

# Funci√≥n para generar una sucesi√≥n oscilante
def sucesion_oscilante(n):
    return (-1) ** n

# Valores de n para la gr√°fica
n_values = np.arange(1, 21)  # Valores de n desde 1 hasta 20

# T√©rminos de las sucesiones
sucesion_convergente_values = [sucesion_convergente(n) for n in n_values]
sucesion_divergente_values = [sucesion_divergente(n) for n in n_values]
sucesion_oscilante_values = [sucesion_oscilante(n) for n in n_values]

# Gr√°fico de convergencia y divergencia de sucesiones
plt.figure(figsize=(10, 6))

# Sucesi√≥n convergente
plt.plot(n_values, sucesion_convergente_values, marker='o', linestyle='-', color='b', label='Convergente')

# Sucesi√≥n divergente
plt.plot(n_values, sucesion_divergente_values, marker='s', linestyle='-', color='g', label='Divergente')

# Sucesi√≥n oscilante
plt.plot(n_values, sucesion_oscilante_values, marker='d', linestyle='-', color='r', label='Oscilante')

plt.xlabel('n')
plt.ylabel('Valores')
plt.title('Convergencia y Divergencia de Sucesiones')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

### ‚ûï **Series** ‚ûï

![Texto alternativo](https://content.gnoss.ws/imagenes/Usuarios/ImagenesCKEditor/e39dea24-3293-4ccc-971a-48f63fc6c932/dc92efa2-ac10-480d-9027-c424f5eba151.gif)


Una serie se define como la suma de los t√©rminos de una sucesi√≥n. Se representa matem√°ticamente como:

$$ \sum_{n=1}^{\infty} a_n $$

donde $a_n$ es el t√©rmino n-√©simo de la sucesi√≥n. La serie converge si la sucesi√≥n de sus sumas parciales converge. En otras palabras, si la sucesi√≥n $\{S_n\}_{n=1}^{\infty}$, donde $S_n = \sum_{k=1}^{n} a_k$, converge a un l√≠mite $L$, entonces la serie converge a $L$. Se representa como:

$$ \sum_{n=1}^{\infty} a_n = L $$

#### üìè **Criterios de convergencia de series** üìè

Existen diversos criterios matem√°ticos para determinar si una serie converge o diverge. Entre los m√°s comunes se encuentran:

* **Criterio de la raz√≥n:** La serie $\sum_{n=1}^{\infty} a_n$ converge si y solo si $ \lim_{n \to \infty} \left| \frac{a_{n+1}}{a_n} \right| < 1$.
* **Criterio de la ra√≠z:** La serie $\sum_{n=1}^{\infty} a_n$ converge si y solo si $ \lim_{n \to \infty} \sqrt[n]{|a_n|} < 1$.
* **Criterio de la prueba:** La serie $\sum_{n=1}^{\infty} a_n$ converge si existe una sucesi√≥n positiva convergente $\{b_n\}_{n=1}^{\infty}$ tal que para todo $n$, se cumple la desigualdad:

$$ |a_n| \leq |b_n| $$

|                           | Concepto                                                                               |
|---------------------------|----------------------------------------------------------------------------------------|

## **üßä Organizador Gr√°fico - Operaciones L√≥gicas üßä**

```python
from graphviz import Digraph

# Operaciones l√≥gicas
p = True
q = False
resultado_and = p and q
resultado_or = p or q
resultado_not = not p

# Imprimir resultados
print("AND l√≥gico:", resultado_and)  # Y l√≥gico
print("OR l√≥gico:", resultado_or)  # O l√≥gico
print("NOT l√≥gico:", resultado_not)  # Negaci√≥n l√≥gico

# Crear el diagrama de flujo
dot = Digraph()

# Nodo inicial
dot.node('A', 'Inicio\n(p=True, q=False)', color='black', fillcolor='lightblue', style='filled', shape='ellipse')

# Operaciones
dot.node('B', 'AND l√≥gico\np and q = {}'.format(resultado_and), color='black', fillcolor='lightgreen', style='filled')
dot.node('C', 'OR l√≥gico\np or q = {}'.format(resultado_or), color='black', fillcolor='lightcoral', style='filled')
dot.node('D', 'NOT l√≥gico\nnot p = {}'.format(resultado_not), color='black', fillcolor='lightpink', style='filled')

# Enlaces
dot.edge('A', 'B', color='blue')
dot.edge('A', 'C', color='green')
dot.edge('A', 'D', color='red')

# Renderizar el gr√°fico
dot.render('operaciones_logicas_colores', format='png', cleanup=True)

# Mostrar el gr√°fico en la notebook
from IPython.display import Image
Image(filename='operaciones_logicas_colores.png')
```
