###**Ecuaci√≥n cuadr√°tica**
```python
# Ejemplo de c√≥digo Python
import math

def resolver_ecuacion_cuadratica(a, b, c):
    # Calcular el discriminante
    discriminante = b**2 - 4*a*c
    
    if discriminante > 0:
        # Dos soluciones reales
        raiz1 = (-b + math.sqrt(discriminante)) / (2*a)
        raiz2 = (-b - math.sqrt(discriminante)) / (2*a)
        return raiz1, raiz2
    elif discriminante == 0:
        # Una soluci√≥n real
        raiz = -b / (2*a)
        return raiz,
    else:
        # Soluciones complejas
        parte_real = -b / (2*a)
        parte_imaginaria = math.sqrt(-discriminante) / (2*a)
        raiz1 = complex(parte_real, parte_imaginaria)
        raiz2 = complex(parte_real, -parte_imaginaria)
        return raiz1, raiz2

# Ejemplo de uso
a = 1
b = -3
c = 2
soluciones = resolver_ecuacion_cuadratica(a, b, c)
print("Las soluciones de la ecuaci√≥n cuadr√°tica son:", soluciones)
```
###**Balanza Comercial**
```python
def calcular_balanza_comercial(exportaciones, importaciones):
    # Calcular la balanza comercial
    balanza_comercial = exportaciones - importaciones
    
    # Determinar el estado de la balanza comercial
    if balanza_comercial > 0:
        estado = "super√°vit"
    elif balanza_comercial < 0:
        estado = "d√©ficit"
    else:
        estado = "equilibrio"
    
    return balanza_comercial, estado

# Datos de ejemplo
exportaciones = 50000  # valor de las exportaciones en millones de USD
importaciones = 45000  # valor de las importaciones en millones de USD

# Calcular y mostrar la balanza comercial
balanza, estado = calcular_balanza_comercial(exportaciones, importaciones)
print(f"Balanza Comercial: {balanza} millones de USD ({estado})")
```
###**N√∫mero Factorial**
```python
def factorial_recursivo(n):
    # Caso base: el factorial de 0 o 1 es 1
    if n == 0 or n == 1:
        return 1
    # Caso recursivo: n * factorial de (n-1)
    else:
        return n * factorial_recursivo(n-1)
# Ejemplo de uso
numero = 5
resultado = factorial_recursivo(numero)
print(f"El factorial de {numero} es: {resultado}")
```
