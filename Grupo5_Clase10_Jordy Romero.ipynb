# **Definici贸n de Espacio M茅trico** 

Un **espacio m茅trico** es un conjunto $M$ junto con una funci贸n $d: M \times M \rightarrow \mathbb{R}$ que define una distancia entre cada par de puntos en $M$. La funci贸n $d$ se llama **m茅trica** y debe satisfacer las siguientes propiedades para todos los $x, y, z \in M$:

1. **No negatividad**: $d(x, y) \geq 0$ y $d(x, y) = 0$ si y solo si $x = y$.
2. **Simetr铆a**: $d(x, y) = d(y, x)$.
3. **Desigualdad triangular**: $d(x, z) \leq d(x, y) + d(y, z)$.

Matem谩ticamente, un espacio m茅trico se denota como $(M, d)$.

## **Ejemplos de Espacios M茅tricos**

1. **Espacio Euclidiano**: El conjunto $\mathbb{R}^n$ con la distancia euclidiana definida por:

   $$ d(x, y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2} $$

2. **Espacio Discreto**: El conjunto $M$ con la distancia $d$ definida por:

   $$ d(x, y) = \begin{cases}
      0 & \text{si } x = y \\
      1 & \text{si } x \neq y
   \end{cases} $$


# **Distancia y Propiedades de la M茅trica**


La **distancia** en un espacio m茅trico $(M, d)$ es una funci贸n $d$ que cumple con las propiedades mencionadas anteriormente.

## **Propiedades de la M茅trica**

1. **No negatividad **:

   $$ d(x, y) \geq 0 $$

   Esta propiedad asegura que la distancia entre dos puntos nunca es negativa.
   
2. **Identidad del Indiscernible М**:

   $$ d(x, y) = 0 \iff x = y $$

   Esto implica que la distancia entre un punto y s铆 mismo es cero y, si la distancia entre dos puntos es cero, entonces los puntos son id茅nticos.
   
3. **Simetr铆a**:

   $$ d(x, y) = d(y, x) $$

   La distancia desde $x$ a $y$ es la misma que desde $y$ a $x$.
   
4. **Desigualdad Triangular **:

   $$ d(x, z) \leq d(x, y) + d(y, z) $$
   
   Esta propiedad establece que la distancia directa entre dos puntos $x$ y $z$ es menor o igual que la suma de las distancias desde $x$ a $y$ y de $y$ a $z$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos puntos para demostrar las propiedades
x = np.array([0, 0])
y = np.array([1, 1])
z = np.array([2, 0])

# Funci贸n de distancia euclidiana
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))

# Propiedades de la m茅trica
d_xy = euclidean_distance(x, y)
d_xz = euclidean_distance(x, z)
d_yz = euclidean_distance(y, z)

# Graficar los puntos
plt.figure(figsize=(10, 10))

# Graficar puntos y sus etiquetas
plt.scatter(*x, color='red')
plt.text(x[0], x[1], 'x', fontsize=12, ha='right')

plt.scatter(*y, color='blue')
plt.text(y[0], y[1], 'y', fontsize=12, ha='right')

plt.scatter(*z, color='green')
plt.text(z[0], z[1], 'z', fontsize=12, ha='right')

# Graficar las distancias
plt.plot([x[0], y[0]], [x[1], y[1]], 'k--', label=f'd(x, y) = {d_xy:.2f}')
plt.plot([x[0], z[0]], [x[1], z[1]], 'k--', label=f'd(x, z) = {d_xz:.2f}')
plt.plot([y[0], z[0]], [y[1], z[1]], 'k--', label=f'd(y, z) = {d_yz:.2f}')

# A帽adir t铆tulos y leyendas
plt.title('Propiedades de la M茅trica')
plt.legend()
plt.grid(True)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Mostrar gr谩ficos de propiedades
plt.show()

# Verificar las propiedades
print("Propiedades de la M茅trica:")
print(f"1. No negatividad: d(x, y) = {d_xy} >= 0")
print(f"2. Identidad del Indiscernible: d(x, x) = {euclidean_distance(x, x)} == 0")
print(f"3. Simetr铆a: d(x, y) = {d_xy} == d(y, x) = {euclidean_distance(y, x)}")
print(f"4. Desigualdad Triangular: d(x, z) = {d_xz} <= d(x, y) + d(y, z) = {d_xy + d_yz}")
```

# **Demostraci贸n de que $B(o, r)$ es cerrada en un espacio m茅trico**

Para demostrar que la bola $B(o, r)$ es cerrada en un espacio m茅trico $(E, d)$:

1. **Definici贸n de una bola abierta**:
   
   En un espacio m茅trico $(E, d)$, una bola abierta centrada en $o$ con radio $r$ se define como $B(o, r) = \{ x \in E : d(x, o) < r \}$.

2. **Objetivo de la demostraci贸n**:
   
   Queremos demostrar que $B(o, r)$ es cerrada, es decir, su complemento $B(o, r)^c$ es abierto.

3. **Complemento de la bola**:
   
   El complemento de la bola $B(o, r)^c$ est谩 dado por $B(o, r)^c = \{ x \in E : d(x, o) \geq r \}$.

4. **Demostraci贸n de que $B(o, r)^c$ es abierto**:
   
   Para demostrar que $B(o, r)^c$ es abierto, tomamos un punto $x \in B(o, r)^c$, lo cual implica que $d(x, o) \geq r$.

5. **Definici贸n de $\varepsilon$**:
   
   Definimos $\varepsilon = d(x, o) - r$. Aqu铆, $\varepsilon > 0$ porque $d(x, o) > r$.

6. **Construcci贸n de la bola abierta alrededor de $x$**:
   
   Consideramos la bola abierta $B(x, \varepsilon) = \{ y \in E : d(y, x) < \varepsilon \}$.

7. **Verificaci贸n de puntos en $B(x, \varepsilon)$**:
   
   Sea $y \in B(x, \varepsilon)$, entonces $d(y, x) < \varepsilon$.

8. **Relaci贸n m茅trica con $o$**:
   
   Ahora evaluamos $d(o, y)$:
   $$d(o, y) \leq d(o, x) + d(x, y) < d(o, x) + \varepsilon$$

9. **Sustituci贸n de $\varepsilon$**:
   
   Sustituyendo $\varepsilon = d(x, o) - r$:
   $$d(o, y) < d(o, x) + d(x, o) - r$$

10. **Conclusi贸n**:
   
    Esto muestra que $d(o, y) < d(o, x) + d(x, o) - r$, lo cual demuestra que $y \in B(o, r)$. Por lo tanto, cualquier punto $y$ en $B(x, \varepsilon)$ tambi茅n est谩 en $B(o, r)$, lo que indica que $B(x, \varepsilon) \subseteq B(o, r)$.

Esto completa la demostraci贸n de que $B(o, r)^c$ es abierto, lo cual implica que $B(o, r)$ es cerrada en el espacio m茅trico $(E, d)$.

- Un conjunto $A$ es cerrado si su complemento $A^c$ (todos los puntos que no est谩n en $A$) es un conjunto abierto.

Un conjunto $B$ en un espacio m茅trico $(E, d)$ es abierto si, para cualquier punto $y$ en $B$, existe una bola abierta centrada en $y$ que est谩 completamente contenida en $B$.

**Formalmente**:
- Un conjunto $B$ es abierto si, para cada punto $y$ en $B$, existe un $\varepsilon > 0$ tal que la bola $B(y, \varepsilon)$ est谩 contenida en $B$.

- Un conjunto $B$ es abierto si su complemento $B^c$ (todos los puntos que no est谩n en $B$) es un conjunto cerrado.

Los conceptos de conjuntos abiertos y cerrados son duales. Esto significa que la definici贸n de uno puede expresarse en t茅rminos del otro:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las secuencias divergentes
def sequence_n(n):
    return n

def sequence_minus_n(n):
    return -n

def sequence_alternating(n):
    return (-1)**n * n

# Generar los valores de las secuencias
n_values = np.arange(1, 21)  # n desde 1 hasta 20

sequence_n_values = sequence_n(n_values)
sequence_minus_n_values = sequence_minus_n(n_values)
sequence_alternating_values = sequence_alternating(n_values)

# Graficar las secuencias
plt.figure(figsize=(10, 6))

plt.plot(n_values, sequence_n_values, 'o-', label='$a_n = n$')
plt.plot(n_values, sequence_minus_n_values, 'o-', label='$a_n = -n$')
plt.plot(n_values, sequence_alternating_values, 'o-', label='$a_n = (-1)^n \cdot n$')

plt.xlabel('n')
plt.ylabel('$a_n$')
plt.title('Secuencias Divergentes')
plt.legend()
plt.grid(True)
plt.show()
```
