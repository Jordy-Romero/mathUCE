**1.	Inyectividad (Función Inyectiva):**

**•**Una función f: A→Bf: A \ Bf:A→B es inyectiva si para cada elemento en el conjunto de llegada (codominio) BBB, existe a lo sumo un elemento en el conjunto de partida (dominio) AAA que es mapeado a ese elemento.


**•**Formalmente, f es inyectiva si f(x1)= f(x2) f(x_1) = f(x_2)f(x1)=f(x2) implica x1=x2x_1 = x_2x1=x2 para todo x1,x2∈Ax_1, x_2 \in Ax1,x2∈A.


**•**En otras palabras, dos elementos distintos en el dominio no pueden tener la misma imagen en el codominio.


**Ejemplo:** La función f(x)=x2f(x) = x^2f(x)=x2 no es inyectiva en R/{R}R porque, por ejemplo, f(2) = f(−2)=4f(2) = f(-2) = 4f(2)=f(−2)=4.


**2.	Sobreyectividad (Función Sobreyectiva):**

**•**	Una función f: A→Bf: A \rightarrow Bf: A→B es sobreyectiva si cada elemento en el conjunto de llegada BBB tiene al menos un elemento en el conjunto de partida AAA que es mapeado a él.


**•**	Formalmente, f es sobreyectiva si para todo y∈By \in By∈B, existe al menos un x∈Ax \in Ax∈A tal que f(x)=yf(x) = yf(x)=y.


**•**	En términos simples, la función cubre (o alcanza) todos los elementos del codominio.


**Ejemplo:** La función f(x) = exf(x) = e^xf(x)=ex, donde f: R→R+f: \{R} \rightarrow \{R}^+f:R→R+, es sobreyectiva porque cada número positivo en R+\{R}^+R+ tiene un preimagen en R/{R}R.


**3.	Biyectividad (Función Biyectiva):**


**•**	Una función f:A→Bf: A \rightarrow Bf:A→B es biyectiva si es tanto inyectiva como sobreyectiva.


**•**	Es decir, cada elemento del conjunto de llegada BBB tiene exactamente un elemento correspondiente en el conjunto de partida AAA, y cada elemento en BBB es mapeado por exactamente un elemento en AAA.


**•**	En términos simples, una función es biyectiva si es una correspondencia uno a uno y sobre entre AAA y BBB.


**Ejemplo:** La función f(x)=2x+1f(x) = 2x + 1f(x)=2x+1 donde f:R→Rf: \{R} \rightarrow \{R}f:R→R es biyectiva porque es tanto inyectiva (cada valor de xxx tiene una única imagen) como sobreyectiva (cada número real es alcanzado por algún xxx).


Para entenderlo de mejor manera.

**•	Inyectiva:** Cada elemento del dominio se une a un único elemento del codominio.


**•	Sobreyectiva:** Todos los elementos del codominio son alcanzados por al menos un elemento del dominio.


**•	Biyectiva:** Es inyectiva y sobreyectiva al mismo tiempo.

Una función f: A→Bf: es inyectiva (o uno a uno) si y solo si, para cada par de elementos distintos x1, x2 en el dominio de A, sus imágenes bajo f son distintas en el codominio B. En otras palabras, f(x1) =f(x2) f(x1) = f(x2) f(x1) = f(x2) implica que         x1= x2.


**Ejemplo de función inyectiva:**

 **La función f: R→R(f) definida por f (x)=2x+3f(x) = 2x + 3f(x)=2x+3.**


**•**	Para verificar la inyectividad, asumimos que f(x1) =f(x2) f(x_1) = f(x_2) f(x1) =f(x2).


**•**	Entonces, 2x1+3=2x2+32x_1 + 3 = 2x_2 + 32x1+3=2x2+3.


**•**	Restando 3 de ambos lados: 2x1=2x22x_1 = 2x_22x1=2x2.


**•**	Dividiendo por 2: x1=x2x_1 = x_2x1=x2.


Como f(x1) =f(x2) f(x_1) = f(x_2) f(x1) =f(x2) implica que x1=x2x_1 = x_2x1=x2, la función f(x)=2x+3f(x) = 2x + 3f(x)=2x+3 es inyectiva.

**Sobreyectividad**


**Una función f: A→Bf: es sobreyectiva si y solo si, para cada elemento y en el codominio BBB, existe al menos un elemento x en el dominio A tal que f(x)=yf(x) = yf(x)=y. En otras palabras, la imagen de f cubre todo el codominio B.**


**Ejemplo de función sobreyectiva:**


**Consideremos la función h: R→R definida por h(x)=x3h(x) = x^3h(x)=x3.**


**•**	Para cada y en {R}R, queremos encontrar un x en {R}R tal que h(x)=yh(x) = yh(x)=y.


**•**	Tomemos x=y3x = {y}x=3y.


**•**	Entonces, h(y3) =(y3)3=y = {y}) ^3 = yh(3y) =(3y)3=y.
Como para cada y en {R}R existe un x en {R}R tal que h(x)=yh(x) = yh(x)=y, la función h(x)=x3h(x) = x^3h(x)=x3 es sobreyectiva.


**Ejemplo de función no sobreyectiva:**


**Consideremos la función k: R→R definida por k(x)=exk(x) = e^xk(x)=ex.**


**•**	Queremos encontrar un x en {R}R tal que k(x)=−1k(x) = -1k(x)=−1.


**•**Sin embargo, exe^xex nunca toma valores negativos para ningún x en {R}R.
Por lo tanto, no todos los elementos de{R}R (el codominio) tienen una preimagen en el dominio {R}R, y k(x)=exk(x) = e^xk(x)=ex no es sobreyectiva.


**Biyección**


**Una función f: A→Bf: es biyectiva si es tanto inyectiva como sobreyectiva. Esto significa que f establece una correspondencia uno a uno y sobre entre los elementos de A y B.**


**Ejemplo de función biyectiva:**


**Consideremos la función l: R→Rl: definida por l(x)=2x+3l(x) = 2x + 3l(x)=2x+3.**


**•**	Como vimos antes, esta función es inyectiva.


**•** Para verificar la sobreyectividad, tomemos cualquier y en{R}R y resolvamos l(x)=yl(x) = yl(x)=y:


**•**	2x+3=y2x + 3 = y2x+3=y

**•**	2x=y−32x = y - 32x=

**•**	x=y−32x = \ {y - 3}{2}x=2y−3

Para cada y en {R}R, existe un x en {R}R tal que l(x)=yl(x) = yl(x)=y. Así, la función l(x)=2x+3l(x) = 2x + 3l(x)=2x+3 es sobreyectiva.
